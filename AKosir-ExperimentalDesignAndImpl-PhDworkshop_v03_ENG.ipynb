## ■ What is experimental design

The elements of the experimental design are as follows:
##### 1. Response (output, dependent) variable $y$:
Criteria function, performance indicator, etc., which gives the main result. In our case, the MM exposure.

##### 2. Controlled factor (s)  $F_e$:
Independent variables that we **control** in the experiment. 
In our case, this is a persons' gender. 

##### 3. Nuisance factors:
These are adverse effects on the experimental result that are unavoidable. We want to neutralize them in several ways;

##### 4. Input noise  $\varepsilon$:
This is the noise with which we present the unknown disturbing factors and the real noise resulting from the randomness of human behavior, the noise of the sensors, etc.


<img style="float: center; width: 50%" src="https://raw.githubusercontent.com/andrejkk/UPK_DataImgs/master/ExperimentalDesignBox.png">


#### Def. The Design of Experiment (DOE) is a process for designing experiments, which includes the selection of criteria, the design of procedures in the experiment, and the design analysis of the results obtained, leading to valid and objective conclusions.

In doing so:
- the procedures also include the determination of criteria for the selection of test persons, the required number of test persons, etc. They also include the exact design of the course of the experiment;
- before the experiment, we make a **statistical design** for data analysis, which will answer the questions
- each experiment covers only a part of the area - it is representative only for the area that the input data (test persons) represent or. represent;

## ■ Experimental and nonexperimental design

#### Experimental design

This is an experiment where we **compare** two or more variants
- user groups
- system variant
- situations
- ... <br>
with each other. To do this, we need to control at least one of the **factors**, that is, to control its value (level).

It is important to consider other **factors** (influences on the result) appropriately.


#### Nonexperimental design

This is an experiment in which we observe the results **without factor control** and do not interfere in the course of the factor control. We call it **observational study**.

### Statistical design of experiments

A data analysis plan that leads to statistical testing of a selected hypothesis, such as
- ANOVA

### Selected terminology

Since the basic scheme is ANOVA, the terminology also derives from their terminology.

##### Regarding factor control
- fixed effect: the experimental factor is controlled, assumed the same for the group
- random effect: the value of the controlled factor is random (but known), assumed different for every individual item
- model with mixed effects: we have factors with fixed and random effects


##### Regarding the number of controll factors
- one way: the design has one controlled factor
- two way: the design has two controlled factors
- ...


##### Blocking and group factors
- blocking factor: division of subjects into two or more groups prior to the experiment by an explicit characteristics;
- group factor: arbitrary division of subjects into two or more groups
  
  
##### What is randomisation and what is replication
- Randomisation: random selection of controlled factor values (levels)
- Replication: repeatin experiment with purposuly varied factors


##### Experiments and quasi-experiment
- randomisation of factor levels and samples are required for experimental desing. Otherwise it is a quasi-experiment only.

## ■ Design ANOVA  

$\def\ovr#1{{\overline{#1}}}$


#### Controlled and nuisance factors

Types:
- fixed effect: the experimental factor is controlled
- random effect: the value of the experimental factor is random
- mixed effects: we have factors with fixed and random effects


Controlled factor value:
- discrete values
- two or more values

#### Assumptions

- Independence of observations
- Normality: the distributions of the residuals of predicted variable are normal.
- Homogeneity: equality of variances, called homoscedasticity—the variance of data in groups should be the same.

If normality is not met (you test it), go to Kruskal-Wallis ANOVA. 

#### Null hypothesis and test

We cover only the option **one factor** with **fixed effect**.

Null hypothesis:
$$ H_{0}=[\ovr{y}_{G_1}=\cdots=\ovr{y}_{G_I}], $$

Test: F-test.


$\def\o#1{{\overline{#1}}}$
$\def\SS{{\text{SS}}}$
$\def\Var{{\text{Var}}}$

###### Notations
- $n$ is the number of experiments performed at each value of the controlled factor $F_e$.
- $a$ is the number of values of the controlled factor $F_e \in \{1, 2, \ldots, a \}$
- $y_{ij}$ is the $j$ th result of the experiment, $i=1, \ldots, n$, for the value of the factor $F_e=i$
- $y_{i.}=\sum_{j=1}^n y_{ij}$ sums per experiment
- $\o{y}_{i.}=y_{i.} / n$ average per experiment
- $y_{. j}=\sum_{i=1}^n y_{ij}$ sums by factor values
- $\o{y}_{. j}=y_{. j} / n$ average by factor values
- $y_{..}=\sum_{i, j} y_{ij}$ total
- $\o{y}_{..}=y_{..} / n$ total average

##### Sum of squares
- total sum of squares
$$\SS_{tot}=\sum_{i=1}^a \sum_{j=1}^n (y_{ij} - \o{y_{..}}) ^ 2$$
- the sum of the squares of the factor
$$\SS_{trt}=n \sum_{i=1}^a (\o{y}_{i.} - \o{y_{..}}) ^ 2$$
- Then the total sum of squares can be divided by the sum of the factor and the error $\SS_{err}$, i.e.
$$\SS_{tot}=\SS_{trt} + \SS_{err}$$


##### Statistics F

Statistics
$$F_0=\frac{SS_{trt} / (a-1)}{SS_{err} / (n (a-1))}$$
assuming $H_0$ is distributed after the distribution of $F (a-1, n (a-1))$, from which we calculate the $p$ value.

#### Variants
 - one way: the design has one controlled factor
 - two way: the design has two controlled factors

```python
## Title: ANOVA design
# Goal: undesrtand the implementaiton of the ANOVA experimental design 
# Hint: 

import numpy as np
import pandas as pd 
import seaborn as sns 
import scipy.stats as stats

# Load data 
post_fn = 'https://raw.githubusercontent.com/andrejkk/UPK_DataImgs/master/ANOVAtestData.csv'
data_df = pd.read_csv(post_fn, sep=';', encoding='utf8')
```

```python
# Plot categories 
sns.catplot(data=data_df, x='G', y='y')
```

```python
data_df
```

```python
# Plot and test assumptions
ax = sns.boxplot(data=data_df, x='G', y='y', color='#99c2a2')
ax = sns.swarmplot(data=data_df, x='G', y='y', color='#7d0013')
```

```python
# Assumptions
X_G1 = data_df[data_df['G']==1]['y']
X_G2 = data_df[data_df['G']==2]['y']
X_G3 = data_df[data_df['G']==3]['y']


# Normality
p_G1 = stats.shapiro(X_G1)[1]
p_G2 = stats.shapiro(X_G2)[1]
p_G3 = stats.shapiro(X_G3)[1]
print ('Normality', p_G1, p_G2, p_G3)
# All p > 0.05 => OK

# Homogenous variance
p_G12 = stats.levene(X_G1, X_G2)[1]
p_G13 = stats.levene(X_G1, X_G3)[1]
p_G23 = stats.levene(X_G2, X_G3)[1]
print ('Homogenous variance:', p_G12, p_G13, p_G23)

# All p > 0.05 => OK
```

```python
# ANOVA for normal data
X_G1 = data_df[data_df['G']==1]['y']
X_G2 = data_df[data_df['G']==2]['y']
X_G3 = data_df[data_df['G']==3]['y']

fvalue, pvalue = stats.f_oneway(X_G1, X_G2, X_G3)
print(fvalue, pvalue)

# p < 0.05 => there are differences
```

## What if data is not normal?

1. Transformations - if there is a theoretical reason:
- logaritmic
- square root
- angular (arcsine) 
- power

2. Use Kruskal-Wallis ANOVA

```python
# ANOVA for non-normal normal data
X_G1 = data_df[data_df['G']==1]['y']
X_G2 = data_df[data_df['G']==2]['y']
X_G3 = data_df[data_df['G']==3]['y']

fvalue, pvalue = stats.kruskal(X_G1, X_G2, X_G3)
print(fvalue, pvalue)

# p < 0.05 => there are differences
```

## ■ Design Latin square

$\def\ovr#1{{\overline{#1}}}$

Test: F-test.




##### Statistics F

Statistics
$$F_0=\frac{MS}{MSE}$$
is distributed after the distribution of $F (n-1, (n-1)(n-2))$, from which we calculate the $p$ value.

## ■ Sample size

#### Main observation:
- different for 1. Hypthese testing and 2. Parameter estimation
- for each stat task separatelly and then maximal required is used


#### Hypothese testing 

Apriori power analysis:
- effect size $\text{ef}$:
    - t-test: $ef=\frac{\overline{x}_1-\overline{x}_2}{\sigma_p}$
    - correlation: $ef=|r|$
    - ... 
- Stat. power = $Pr[\hat{\neg H_0}|_{\neg H_0}] = f(n, \text{ef})$, use G*Power to estimate sample size $n$


Link to [G*Power](https://www.psychologie.hhu.de/arbeitsgruppen/allgemeine-psychologie-und-arbeitspsychologie/gpower) tool.


#### Parameter estimation 

Input info is the upper limit to the confidence interval ($Pr[x\in CI]=\alpha$)
$$ CI = [\hat{\overline{x}} - z_\alpha \frac{\hat{\s}}{\sqrt{n}}, \hat{\overline{x}} - z_\alpha\frac{\hat{\s}}{\sqrt{n}}] $$
and 
$$ |CI| = 2 z_\alpha\frac{\hat{\s}}{\sqrt{n}} \leq \Delta P, $$
where $z_\alpha = 1.96$,
and
$$ 4 z_\alpha^2 \frac{\hat{\s}^2}{\Delta P^2} < n. $$

## ■ Questionnaire and measurement characteristics


#### Problem

Questionnaires can give **useless results** because the answers to the questions are not related to the real situation. There are several reasons for this:
1. The terms in the questions or the questions themselves **are not well defined**: terms in human language can have several meanings at the same time and if they are not specified, the answers refer to different terms, objects, etc.
2. It is not clear in the question what it refers to - no reference is given: the question can refer to the quality of the service, interface, content, etc .;
3. Questions can be offensive or distracting to the test person
Consequently, it is imperative that we be skeptical about the applicability of individual issues.

#### Solution

Measurement characteristics can detect questions that **do not work**, that is, they do not give meaningful information. The point is that the measurement characteristics indicate the properties of the answers that must always be valid (eg stability in case of multiple questions, etc.) and thus good measurement characteristics are **a necessary condition for the usability of the answers**, but not a sufficient condition.

Therefore:
1. Estimate measurement characteristics
2. If OK, use the data, otherwise repeat user experiment

## ■ Reliability

#### Def .: Reliability refers to the consistency of a measure, that is, whether the same concept / phenomenon would measure sufficiently similarly when repeatedly measured.


#### Types of reliability
1. Reliability over time - reliability test-retest: whether multiple testing gives similar results
2. Internal consistency: are the answers between the questions sufficiently related
3. Consistency between research (researchers)



#### Internal consistency - split half method

Answers to the question **Are the answers / data stable?**

##### The process of determining parallel shapes
$\def\s{{\sigma}}$
In this method for the given data:
1. Test subjects are randomly divided into two halves
2. We calculate the correlations between them $r_ {12}$
3. According to the Spearman-Brown formula, calculate the reliability coefficient
   - for default the same standard deviations
  $$ r_{tt} = \frac{2 r_{12}} {1 + r_{12}}$$
   - for different standard deviations
  $$ r_{tt} = \frac{4 \s_1 \s_2 r_{12}} {\s_1^2 + \s_2^2 + 2\s_1 \s_2 r_ {12}}, $$
  where $\s_1$ and $\s_2$ are the standard deviations of individual data.



#### Internal consistency - Cronbach's alpha

Answers to the question **Do all questions / data together measure a coherent concept?**

##### Cronbach's alpha determination process
The results (answers) to the $K$ questions $X_i$ are summed in
$$ Y = X_1 + \cdots + X_K, $$
calculate the standard deviations of individual results $ \ s_ {X_i} $ and the sum of $ Y_ {Y} $, and calculate
$$ \alpha = \frac {K-1}{K} \left(1- \frac{\sum_{i = 1}^K \s_{X_i}^2} {\s_Y^2} \right) $$


#### Advice
You need to do it on your data. The right reliability coefficient is difficult to select.
Literature: 2016 Cho Making Reliability Reliable: A Systematic Approach to Reliability Coefficients, Organizational researc mothods, 2016.

### Observational study: model building and association analysis

The results of an observational study give

### Experimental method: hypothesis testing


Experimental design to data for **hypothesis testing**

- from **research question** we form **working hypothesis**
  - our example of a meausring multimedia exposure:
    - research question: dose gender affects the MM exposure one receives?
    - working hypothesis: the gender has no effect on MM exposure
- on the basis of the working hypothesis we form **null hypothesis**
  - our example of Multimedia exposure:
    - null hypothesis (e)
    $$ H_0 = [\overline{y}_M = \overline{y}_F], $$
    where $\overline{y}_M$ is the average exposure of males and $\overline{y}_F $ is the average exposure of females.

#### Steps / elements

Vloga za etično presojo imam naslednje elemente: 
1. _Osnovni podatki o vlogi:_ vlagatelj, naslov, področje raziskave ipd. 
2. _Uvod:_ vlagatelj poda izhodišča raziskovalnega dela, raziskovalna vprašanja, hipoteze in druge relevantne uvodne informacije; 
3. _Udeleženci:_ število in opis udeležencev, način pridobivanja udeležencev (dobrodošla je navedba vključitvenih in izključitvenih kriterijev) in njihova morebitna korist od udeležbe;   
4. _Raziskovalni načrt:_ opis raziskovalnega načrta (npr. anketa, eksperiment, intervju, fokusna skupina itd) ter navedba odvisnih (rezultati postopka) in neodvisnih spremenljivk (kontrola eksperimenta in drugo);
5. _Postopek:_ poleg samega zaporedja korakov vlagatelj poda še seznam instrumentov in pripomočkov (vprašalniki, senzorji ipd), opis eksperimentalnih nalog udeležencev in izdelana navodila udeležencem.
6. _Soglasje za sodelovanje:_ dokončan pisni obrazec za formalno soglasje k sodelovanju udeležencev, opustitev pisnega soglasja, morebitni razgovor o raziskavi in umik soglasja; 
7. _Varovanje in zaupnost podatkov:_ Vlagatelj zagotovi varovanje zajetih podatkov v vseh fazah od merjenja do shranjevanja in obdelave, torej popolno anonimnost, psevdoanonimizacijo podatkov (če je potrebno zaradi nadaljnje obdelave podatkov) ali polno anonimizacijo podatkov, ter preprečitev prepoznavnosti podatkov;
8. _Ocena morebitnih tveganj in koristi sodelovanja:_ tveganja za testne udeležence morajo biti jasno opredeljena, prav tako njihove morebitne koristi (slednje bi lahko vplivalo na izbor udeležencev in posledično na same rezultate). 
9. _Priloge:_ potrebne priloge kot je izdelano soglasje za sodelovanje itd.
