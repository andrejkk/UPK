## Vodila pri izbiri vsebine in osnovne predpostavke


#### Vodila

- človekov (uporabnikov) čas in človekova prozornost je več vredna kot strojna (storitvena)
- kaj postaja relevantno v praksi: podjetja
- kaj je zanemarjeno pri študijih
- poleg teoretičnih osnov tudi praktične spretnosti 
- to ni predmet iz: strojnega učenja, načrtovanja uporabniške izkušnje
- kaj nimajo drugi predmeti
- druga stopnja je magisterij: zahtevnejše vsebine
- tehnologije:
    - Juyter
    - Google in Microsoft on ostale spletne storitve
    

#### Osnovne predpostavke

- konzumiranje multimedijskih vsebin je dvosmerna komunikacija
- prilagajanje uporabniku je smiselno
- prilagajanje uporabniku je s prihajajočimi tehnologijami izvedljivo
- socialna inteligenca storitev je prihodnost multimedijskih komunikacij


#### Osnovna zgodba

Sodobne storitve (v zvezi z mulimedijo in ostale) so socialno neinteligentne. Dodajanje socialne inteligence bo v veliki meri izboljšalo uporabnikovo dobrobit. 

Bistvo je minimizacija kraje pozornosti. 

TEDx: Kaj nam umetna prinaša socialna inteligenca: https://www.youtube.com/watch?v=YtmAcupMYK8&t=281s
