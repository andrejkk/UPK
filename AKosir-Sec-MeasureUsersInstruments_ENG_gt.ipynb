# ■ Introduction: Why psychological constructs in technology?


Psychological constructs are of the utmost importance when measuring users outside of real time, which is carried out on the basis of a user's self -employment in the form of answers to questions.They are made up of an instrument (a list of questions by the measurement - user answers) and specific structures from one or more dimensions or.Aspects that are evaluated in numeric form based on these answers.E.g.The construct of Applicability has one dimension of "Applicability", which is evaluated on the basis of answers to 10 carefully selected questions.These questions are selected based on so -called operationalization and one or more experiments on real measurements, followed by a multivariate statistical analysis.The purpose of the latter is to assess or this set of questions 1. It measures what is intended for (validity) and 2. This measures stable and reliable (reliability).

Why do we not ask the user in an instant "Do you find this service good?" But do we ask him a larger number of indirect questions?Simply because human natural language is not accurate enough to get stable and valid answers.Experiments have shown that when asked "service is good", users think of extremely different aspects, from the shape and structure of the interface to the specific features of the service itself.Therefore, the answers are very scattered and are useless to direct the concrete development of applications and services.The solution is to ask the user more better questions and then analyze the answers.Because both questions and answers are related to each other, they are summarized by dimension (multivariate statistics) at the minimum number of dimensions that then make up the construct.We add operational significance to these dimensions or rocks and consequently become readable for humans.

The importance of psychological constructs in technology is exceptional.Selected set of user states and behavior Konstuct 1. Defines and 2. Makes it measurable.Such a construct is e.g.The Applicability of the system.What is the Applicability of the instrument and the construct was unknown.defined in a way that would be useful in the development of services and applications.Today, however, under the term Applicability, we understand what this construct evaluates!However, realistic measurement (measurability), of course, requires a prerequisite for any Applicability in real technological development.

Given the written, it is understandable that making your own questionnaire for practical use is always a wrong decision.The right decision is to select an appropriate existing validated inacatrument and construct.As the experience of the construction of new instruments and structures shows, it requires extensive experiments, several times in several consecutive performances and demanding analysis in the field of multivariate statistics.The laws of his own questionnaire are doomed to fail in terms of validity (what measures) and reliability (gives stable results).

# ■ what is an instrument and what is a construct


#### Construct

The construct is the theoretical framework of human onbag, traits or opinion processes, we are interested in behavior or properties relevant to user interaction with smart systems.Constructs are usually made up of several dimensions, indicate individual aspects of properties or behavior and give them with individual numbers.The construct indicates a model that has several independent dimensions.If the dimension is single, it is called a rock.

An essential feature of structures is their connection to the instruments that make the construct measurable day.

A large personality construct (see below) describes the basic characteristics of the user's personality and consists of five dime.Each covers its own aspect and its intensity is given with numbers between 0 and 100 (or between 0 and 1).


##### Def.: Psychological construct is a tool used to facilitate understanding human behavior.It identifies important aspects of chosen behavior and provides a way of measuring these aspects.


##### Dimensions:
The term described by the psychological construct can cover several (at least partially) independent characteristics.We call them dimensions.Psychological construct has dimensions
$$ d_1, \ldots, d_n, $$
who are independent of each other and each has their own interpretation.

In the model of the Big5,  $n = 5$ dimensions, and these dimensions are


| Dimension | Description |
|-----------|------|
| Agreeableness | maintaining positive relationships with others |
| Extraversion | interest in social events, activity, energy |
| Conscientiousness | responsibility, adherence to rules, perseverance, attentiveness |
| Openness | openness to new experiences, cultural sophistication |
| Neuroticism | emotional control, anxiety |



The dimensions of the construct are usually not really independent.

# ■ Selected constructs in interaction with (smart) systems


Psychological constructs chosen the framework of the user's condition and behavior are just defined and made measurable.

Important constructs in user interaction with (smart) systems are:
1. **Applicability**: How easy is to use the system.

2. **Satisfaction**: How the user responds emotionally after using the system.

3. **Cognitive load**: The user becomes acquainted with the use of a product or service and agrees: cognitive effort: the mental effort required during the interaction.

4. **Affective and emotional state**: User's affective and emotional response to the system (positive or negative).

5. **Engagement**: user's involvement and immersion in interaction.

6. **Motivation**: Factors that encourage users to use the system.

7. **Specific Constructs**:
    - the quality of the spoken interface
    - an interaction with a smart home
    - ...

8. **Accessibility**: useful for users with special needs.

9. **Detected control**: Feeling of system control.

10. **Frustration**: frustration: negative emotional response caused by problems in the system.

11. **Trust**: Trust in the functionality and security of the system.

12. **Aesthetics**: Visual attraction and beauty of the interface.

# ■ Applicability: System Applicability SUS

#### Measures: System Applicability of a service, device or system. Serves as a first assessment of whether a service is useful for a given group of users.

#### Scale: Real number that gives the system Applicability of the service

#### Instrument

<img style="float: center; width: 350px; margin: 20px 20px 50px 0;" src="https://raw.githubusercontent.com/andrejkk/ORvTK_SlidesImgs/master/Instrument-ApplicabilitySUS.png">

#### Scoring

Answers are on a scale of 1 - 5. The answer to the $i$ question is denoted by $qa_i$.
The equation for determining the score is
$$ SUS = 2.5\left(20 + (qa_1 + qa_3 + qa_5 + qa_7 + qa_9) - (qa_2 + qa_4 + qa_6 + qa_8 + qa_{10})\right). $$

<br>

Source: [System Applicability](https://www.Applicability.gov/how-to-and-tools/methods/system-Applicability-scale.html)

# ■ Personality Measurement: Big 5

#### Measures: personality. There are no direct guidelines for Applicability in user customization. Personality is used in creating user profiles.

#### Construct: Five-Dimensional Personality Model

|Dimension |Description |
|:--- |:--- |
|**Agreeableness**|Agreeableness refers to the ability to get along with other people and helps us overcome the frustrations associated with living in a group. |
|**Extraversion**|Extraverts are highly interested in social events, are very active and energetic. |
|**Conscientiousness**|Conscientiousness encompasses traits such as responsibility, adherence to rules, perseverance, attentiveness |
|**Openness**|It focuses primarily on an individual's cultural sophistication and openness to new experiences. |
|**Neuroticism**|Individuals with high neuroticism are less able to control their emotions and are more anxious. |

#### Instrument: 44 questions (Big Five inventory, BFI)

The instrument consists of 44 questions from different sub-areas.

<img style="float: center; width: 350px; margin: 20px 20px 50px 0;" src="https://raw.githubusercontent.com/andrejkk/ORvTK_SlidesImgs/master/Instrument-BFI_44.png">

This instrument (44 questions) is most suitable for determining personality in the context of service adaptation. There are several other instruments (e.g. NEO-PI-R with 240 questions and TIPI with 10 questions) that are less suitable or unsuitable for use in our context.

#### Scoring

Answers are coded 1 - 5. The answers to "negative" questions must be reversed, i.e. 5->1, 4->2, 3->3, 2->4, 1->5. Then, for each dimension, the number of points of the corresponding questions is added up as shown in the table below.

|Dimension |Positive questions| Negative questions |
|:--- |:--- |:--- |
|**Agreeableness**| 7, 17, 22, 32, 42 | 2, 12, 27, 37 |
|**Extraversion**| 1, 11, 16, 26, 36 | 6, 21, 31 |
|**Conscientiousness**| 3, 13, 28, 33, 38 | 8, 18, 23, 43 |
|**Openness**| 5, 10, 15, 20, 25, 30, 40, 44 | 35, 41 |
|**Neuroticism**| 4, 14, 19, 29, 39 | 9, 24, 34 |

Note: some systems give the average of the points instead of the sum.

<br><br>
Source: [Center for Psychodiagnostic Tools](https://www.center-pds.si/)

#### Usage
The dimensions Agreeableness and Openness have a weak influence on users' decision-making.

# ■ Mood measurement: Positive and Negative Affect Scale (PANAS)

#### Measures: positive and negative affect.

#### Applicability:
- assesses the user's perceived affect after the experiment.
- the sign of affect influences the perception of the difficulty of a task and the success of solving interaction tasks

#### Construct

|Dimension|Description|
|:--- |:--- |
|**Positive Affect (PA)**|Measures the level of feelings of enthusiasm, activity and alertness |
|**Negative Affect (NA)**|Measures negative emotional states such as contempt, disgust, fear, etc. |

#### Instrument

<img style="float: center; width: 350px; margin: 20px 20px 50px 0;" src="https://raw.githubusercontent.com/andrejkk/ORvTK_SlidesImgs/master/Instrument-PANAS.png">

#### Scoring

Answers are scored 1 - 5. We label them with $qa_1, \ldots, qa_{20}$. The scores are determined by
$$ \text{PA} = \sum_{i=1}^{10} qa_i, \qquad \text{NA} = \sum_{i=11}^{20} qa_i. $$

# ■ Engagement Measurement: UES

#### Measures: user engagement in using the system

#### Applicability:
- Assessment of user engagement of existing solutions
- Planning of solution improvements


#### Construct

|Dimension|Description|
|:--- |:--- |
|**Focused attention (FA)**| Feeling absorbed in the interaction and losing track of time |
|**Perceived usability (PU)**| Negative affect experienced as a result of the interaction and the degree of control and effort expended |
|**Aesthetic appeal (AE)**| Attractiveness and visual appeal of the interface |
|**Reward (RW)**| Positive feelings and satisfaction users get from interacting with the system |

#### Instruments and Scoring

The questions in both instruments are coded on a Likert scale "Strongly disagree, Disagree, Neither agree nor disagree, Agree, Strongly agree" with codes 1 2 3 4 5.

###### UES-LF Instrument

Longer form of the instrument with 30 questions.

<img style="float: center; width: 350px; margin: 20px 20px 50px 0;" src="https://raw.githubusercontent.com/andrejkk/ORvTK_SlidesImgs/master/Instrument-UES-LF.png">

The scores for individual dimensions are determined by:
1. Reverse the answers to questions (1->5, ..., 5->1) to questions PU-1, ..., PU-6, PU-8, and RW-3;
2. Average the values ​​of the corresponding answers:
- FA: average FA.1 - FA.7
- PU: average PU.1 - PU.8
- AE: average AE.1 - AE.5
- RW: average RW.1 - RW.10

#### UES-SF Instrument
Shorter form of the instrument with 12 questions.

<img style="float: center; width: 350px; margin: 20px 20px 50px 0;" src="https://raw.githubusercontent.com/andrejkk/ORvTK_SlidesImgs/master/Instrument-UES-SF.png">

The points of individual dimensions are determined by:
1. Reverse the answers to questions (1->5, ..., 5->1) to questions PU-S1, PU-S2 and PU-S3;
2. Average the values ​​of the corresponding responses:
- FA: average of FA.S1 - FA.S3
- PU: average of PU.S1 - PU.S3
- AE: average of AE.S1 - AE.S3
- RW: average of RW.S1 - RW.S3

<br><br>
Source: O'Brien et. al. A practical approach to measuring user engagement with the refined user engagement scale (UES) and new UES short form, International Journal of Human-Computer Studies, 2018.

# ■ Measuring an emotional response: SAM

#### Measures: Affective status

### Applicability:
- Measuring the user's affective and emotional state



#### Construct


|Dimension|Description|
|:--- |:--- |
|**Valence**| Pleasantness or unpleasantness a person feels in response to a stimulus. It reflects how positive or negative the emotion is. |
|**Arousal**| Intensity of the emotional response, specifically whether the person feels activated or calm. |
|**Dominance**| The level of control or power the person feels in the situation. |



### Instrument

The instrument is graphic, the original form is shown here (Valence is named Peasure)

Points are determined directly from the user's choice on the instrument and depends on the dimension itself.

##### 1. Valence:
1 = Very unhappy or unpleSant (Extremely negative emotion), ..., 5 = Neutral, ..., 9 = Very Happy or Pleasant (Extremely Positive Emotion)


##### 2. Arousal
1 = Extremely Calm or Relaxed, ..., 5 = Neutral Arousal, ..., 9 = Extremely Excited or Aroused


##### 3. Dominance
1 = Feeling Highly Submissive or Powerless, ..., 5 = Neutral Dominance, ..., 9 = Feeling Highly Dominant or and Control

# ■ Calculation of construct or scale scores

The result of a measurement with a psychological instrument is determined by calculating scale scores or points for an individual dimension of a multidimensional construct. These are calculated directly from the respondents' answers, which are represented by numerical values ​​of the Likert scale. For example, from "unintelligible" = 1 to "understandable" = 7.

The instructions for determining the points are part of the instrument itself. Usually this is a simple averaging of agreements, but it may also require matrix calculation or even an optimization procedure.

### Construct model

The construct model is obtained directly from the results of the analysis of the experiment, which gives the construct itself. Its essential part is the loading matrix.

##### User responses

A matrix of user responses $U$ to questions $Q$,
$$ X = [a_{uq}], \qquad u\in U, q\in Q, $$
where $U$ is the set of test users and Q is the set of their responses to the questions.

##### The model is given by a weight matrix

The weight matrix
$$ L = [\lambda_{qi}], \qquad q\in Q, i\in F, $$
where Q is the set of their responses to the questions and $F$ is the set of factors.

#### Calculation of scores (sores)

##### 1. Reversal of responses

Average of selected responses for the selected dimension or scale, with some responses reversed (e.g. 5 ->1, 4->2, ... 1 ->5).

##### 2. Matrix calculus

The matrix of scores $S=[s_{ui}], u\in U, i\in F$, is given by
$$ S = X.C^{\top}.L, $$
where $C$ is the covariance matrix of the responses $X$.

It can be shown that simplifying this formula leads to averaging the responses and that averaging the responses gives less accurate scores than this formula.

```python
## Estimate quality of speech interface 
# User Goal Orientation = UGO: 8 items, a = .92
# Customer Service Behaviors = CSB: 8 items, a = .89
# Speech Characteristics = SC: 5 items
# Verbosity = V: 4 items, a = .69


import numpy as np
import pandas as pd 
from scipy.optimize import minimize


# @brief reshape ND array to 1D as required by binary FA
# @arg nS number of samples
# @arg nQ number of questions
def reshape_NDto1D(a_nparr, nS, nQ):
    return a_nparr.T.reshape(nS*nQ)


# @brief reshape 1D array back to ND as required by binary FA
# @arg nS number of samples
# @arg nQ number of questions
def reshape_1DtoND(a_nparr, nS, nQ):    
    return a_nparr.reshape(nQ, nS).T


# @brief Get linear scores using direct optimisation. 
# @arg qa_mat anwsers matrix, users raws, questions columns
# @arg la_mat loadings  
# @arg cls0_trsh=0.5 ignored
# @ret S_star resulting scores
def get_linear_scores_D(qa_mat, la_mat, cls0_trsh=0.5):
    
    nS, nQ = qa_mat.shape # Number of samples and number of questions
    _, nF = la_mat.shape # Number of factors


    reg_la = 1.0 # Regularisation constant

    # Cost function    
    def cost_f(Sc_1D):
        Sc = reshape_1DtoND(Sc_1D, nS, nF)
        err_mat = qa_mat - Sc @ la_mat.T
        cost_err = np.linalg.norm(err_mat) + reg_la*np.abs(1 - np.linalg.norm(Sc))
        return cost_err


    # Minimsation
    S0 = np.ones((nS, nF)) / (1.0*np.sqrt(nS)) # To start with |S0|_2 = 1
    S0_1D = reshape_NDto1D(S0, nS, nF)
    
    
    res = minimize(cost_f, S0_1D) #, method='nelder-mead',
               #options={'xatol': 1e-8, 'disp': True})
    
    S_star = reshape_1DtoND(res.x, nS, nF) # This are scores    
    
    return S_star


# ---------------------------------------------------------------------------
# Load data 
post_fn = 'https://raw.githubusercontent.com/andrejkk/UPK_DataImgs/master/2020-2021_PredPost-vprasalnik(dvogovor)(Responses).csv'
data_df = pd.read_csv(post_fn, header=0, sep=';', encoding='utf8')

#qs_weights_fn = 'https://raw.githubusercontent.com/andrejkk/UPK_DataImgs/master/QualityOfConvSys_Weights.cvs'
#qs_weights_df = pd.read_csv(qs_weights_fn, header=0, sep=';', encoding='utf8')

qs_weights_df = pd.DataFrame(
[[0.799, 0, 0, 0],
[0, 0, 0, 0.706],
[0.8, 0, 0, 0],
[0, 0.711, 0, 0],
[0.805, 0, 0, 0],
[0, 0.758, 0, 0],
[0.628, 0, 0, 0],
[0, 0, 0, 0.701],
[0, 0.736, 0, 0],
[0.831, 0, 0, 0],
[0, 0.739, 0, 0],
[0.834, 0, 0, 0],
[0.858, 0, 0, 0],
[0, 0.726, 0, 0],
[0, 0, 0, 0.655],
[0, 0, 0.585, 0],
[0.834, 0, 0, 0],
[0, 0, 0.808, 0],
[0.794, 0, 0, 0],
[0, 0, 0.797, 0],
[0, 0.599, 0, 0],
[0, 0, 0, 0.73],
[0, 0.648, 0, 0],
[0, 0, 0.658, 0],
[0, 0.668, 0, 0]])




# Selectors 
post_qs_inds = list(range(11,36))
data_qs_df = data_df.iloc[:, post_qs_inds].drop([13])


# Estimate latent variables
qa_mat = data_qs_df.to_numpy()
la_mat = qs_weights_df.to_numpy()
speech_quality_est_np = get_linear_scores_D(qa_mat, la_mat) # was data_qs_df.dot(qs_weights_df.to_numpy())

speech_quality_est_df = pd.DataFrame(speech_quality_est_np)
speech_quality_est_df.columns = ['UGO', 'CSB', 'SC', 'V']
```

# ■ Problem of questionnaires and measurement characteristics


#### Problem

Questionnaires can produce **useless results * Because the ridicule of the questions are not related to the real situation.There are more reasons for this:
1. Concepts in questions or questions themselves **are not well defined**: terms in human language can carry several meanings at the same time and if they are not specified, the answers refer to different concepts, objects, etc
2. The question is not clear what the reference is - the reference is not given: the questions may be related to the quality of the service, the interface, the content, etc.;
3. Questions can be offensive or disturbing for a test person
As a result, it is imperative that we are skeptical of individual issues.

#### Solution

The measurement characteristics detect questions that ** do not work **, do not give meaningful answers.It is that measurement characteristics are perceived by the properties of answers that always have to apply (eg stability to multiple questions, etc.), and thus good measurement carcateristics ** is a requirement for the Applicability of answers **, but not a sufficient condition.

# ■ Reliability

#### Def.: Reliability refers to the consistency of a measure, i.e. whether the same concept/phenomenon would be measured sufficiently similarly when measured repeatedly.

#### Types of reliability
1. Reliability over time - test-retest reliability: does repeated testing give similar results (_not applicable to the case of a two-way conversation with Alexa_)
2. Internal consistency: are the answers between questions sufficiently related (_essential for the case of a two-way conversation with Alexa_ - see below)
3. Consistency between studies (researchers) (_not applicable to the case of a two-way conversation with Alexa_)

#### Ad 2: Internal consistency - split half method

Answers to the question **Are the answers/data stable?**

###### Determination process
$\def\s{{\sigma}}$
In this method for the given data:
1. The test subjects are randomly divided into two halves
2. We calculate the correlations between them $r_{12}$
3. We calculate the reliability coefficient according to the Spearman-Brown formula
- for assumed equal standard deviations
$$ r_{tt} = \frac{2 r_{12}}{1 + r_{12}} $$
- for different standard deviations
$$ r_{tt} = \frac{4 \s_1 \s_2 r_{12}}{\s_1^2 + \s_2^2 + 2 \s_1 \s_2 r_{12}}, $$
where $\s_1$ and $\s_2$ are the standard deviations of the individual data.

#### Ad 2: Internal consistency - Cronbach alpha

Answers to the question **Do all questions/data together measure a coherent concept?**

The starting point is the ratio of true variability $\sigma_T^2$ and measured variability $\sigma_T^2 + \sigma_\varepsilon^2$, i.e.
$$ \frac{\sigma_T^2}{\sigma_T^2 + \sigma_\varepsilon^2}, $$
which is estimated with statistical formulas such as Cronbach's $\alpha$, among others

It applies to unidimensional data, therefore for each dimension separately.

##### The process of determining Cronbach's $\alpha$
The results (answers) to $K$ questions $X_i$ are summed up in
$$ Y = X_1 + \cdots + X_K, $$
we calculate the standard deviations of the individual results $\s_{X_i}$ and the sum $Y_{Y}$, and calculate
$$ \alpha = \frac{K-1}{K} \left(1-\frac{\sum _{i=1}^K\s_{X_i}^2}{\s_Y^2}\right) $$

```python
## Measurement characteristics
import numpy as np
import pandas as pd 
import random

# Load data 
post_fn = 'https://raw.githubusercontent.com/andrejkk/UPK_DataImgs/master/2020-2021_PredPost-vprasalnik(dvogovor)(Responses).csv'
#post_fn = 'https://raw.githubusercontent.com/andrejkk/UPK_DataImgs/master/PostQs_Responses.csv'
data_df = pd.read_csv(post_fn, header=0, sep=';', encoding='utf8').drop([13])


# Selectors 
pre_qs_inds = list(range(5,11))
post_qs_inds = list(range(11,36))


# ==============================================================================================
## Validitiy
# We estimate is OK

# ==============================================================================================
## Reliability

# Parallel forms - inter method reliability: divide test tasks to two parts and corelate anwsers 
# Split half
np.random.shuffle(post_qs_inds)
perm_inds_1, perm_inds_2 = post_qs_inds[0:12], post_qs_inds[12:24] 
anws_1, anws_2 = data_df.iloc[:, perm_inds_1], data_df.iloc[:, perm_inds_2]


# Get correlation coefficient
half1_pd = anws_1.sum(axis=1)
half2_pd = anws_2.sum(axis=1)
corr = np.corrcoef(half1_pd, half2_pd)[0,1]


# ==============================================================================================
print ('====================================================================================')
print ('== For instruments')
# Spearman–Brown prediction formula
# Equal variances
req_tt = 2*corr / (1.0 + corr)
print ('Equal var, r_tt=', req_tt)

# Non-equal variances
sd1, sd2 = np.std(half1_pd), np.std(half2_pd) 
rneq_tt = 4*sd1*sd2*corr / (sd1**2 + sd2**2 +  2*sd1*sd2*corr)
print ('Std 1 = ', sd1)
print ('Std 2 = ', sd2)
print ('Not equal var, r_tt=', rneq_tt)


## ==============================================================
# Internal consistency - Cronbach alpha
# Verifies the instrument as a whole

post_qs_df = data_df.iloc[:, 11:36]
_,K = post_qs_df.shape
X = post_qs_df.sum(axis=1)
ss_X = np.var(X, ddof=1)
ss_Yi = np.var(post_qs_df, ddof=1)
ss_Y = ss_Yi.sum()

Cronb_a = (K/(K-1))*(1.0 - ss_Y/ss_X)
print ('Cronbach Alpha =', Cronb_a)


```
