# ■ Uvod: kaj je instrument in kaj je konstrukt

#### Konstrukt

Konstrukt je oznaka za dano področje obnašanja ali lastnosti. Nas zanimajo obnašanja ali lastnosti, relevantna za uporabniško interakcijo s pametninimi sistemi. Konstrukti so navadno sestavljeni iz večih dimenzij, označujejo posamezne vidike lastnosti ali obnašanja in jih podajamo s posameznimi števili. 

Npr. konstrukt osebnosti Velikih (glejte spodaj) opisuje osnovno značilnosti osebnosti uporabnika in je sestavljen iz petih dimezij. Vsaka pokriva svoj vidik in njegovo intenzivnost podamo s številom med 0 in 100 (ali med 0 in 1). 


##### Def.: 

Psihološki konstrukt je orodje, ki se uporablja za lažje razumevanje človeškega vedenja. Identificira pomembne vidike izbranega obnašanja in nudi način merjenja teh vidikov. 


##### Dimenzije:
Pojem, ki ga psihološki konstrukt opisuje, lahko zajema več (vsaj delno) neodvisnih značilnosti. Imenujemo jih dimenzije. Psihološki konstrukt ima dimenzije 
$$ d_1, \ldots, d_N, $$
ki imajo svojo interpretacijo. 

Npr. pri modelu osebnosti Velikih 5 je $N=5$ dimenzij te dimenzije so
$$ d_1=\text{Sprejemljivost},\quad\ldots\quad d_5=\text{Nevroticizem}. $$

Dimenzije konstrukta navadno niso zares neodvisne.

```python
## Zajem osebnosti Velikih 5
import ipywidgets as widgets
import numpy as np


# Definicija vprašalnika
m = 50
tab_contents = ["Q" + str(i) for i in range(1,m+1)]

# Vprašanja
list_of_questions = ["Am the life of the party", "Feel little concern for others.", "Am always prepared.", "Get stressed out easily.",
             "Have a rich vocabulary.", "Don't talk a lot.", "Am interested in people.", "Leave my belongings around.",
             "Am relaxed most of the time.", "Have difficulty understanding abstract ideas.", "Feel comfortable around people.",
             "Insult people.", "Pay attention to details.", "Worry about things.", "Have a vivid imagination.",
             "Keep in the background.", "Sympathize with others' feelings.", "Make a mess of things.", "Seldom feel blue.",
             "Am not interested in abstract ideas.", "Start conversations.", "Am not interested in other people's problems.",
             "Get chores done right away.", "Am easily disturbed.", "Have excellent ideas.", "Have little to say.",
             "Have a soft heart.", "Often forget to put things back in their proper place.", "Get upset easily.",
             "Do not have a good imagination.", "Talk to a lot of different people at parties.",
             "Am not really interested in others.", "Like order.", "Change my mood a lot.", "Am quick to understand things.",
             "Don't like to draw attention to myself.", "Take time out for others.", "Shirk my duties.", "Have frequent mood swings.",
             "Use difficult words.", "Don't mind being the center of attention.", "Feel others' emotions.", "Follow a schedule.",
             "Get irritated easily.", "Spend time reflecting on things.", "Am quiet around strangers.", "Make people feel at ease.",
             "Am exacting in my work.", "Often feel blue.", "Am full of ideas."]

# Možni odgovori
options = ["Very Inaccurate", "Moderately Innacurate", "Neither Accurate nor Innacurate", "Moderately Accurate", "Very Accurate"]
list_of_options = [options] * m

 

# "radio" gumbi
children = [widgets.RadioButtons(options=opt) for opt in list_of_options]


# Implementacija vprašalnika - uporaba widgeta
# sestava: labela(vprašanje, widget (radio gumbi))
questions = [widgets.HBox([widgets.Label(name), child])
            for name, child in zip(list_of_questions, children)]


# Poimenuj vprašanja za lažjo orientacijo
tab = widgets.Tab()

tab.children = questions
for i in range(len(questions)):
    tab.set_title(i, tab_contents[i])
```

```python
## Izvedba merjenja
tab
```

```python
## Definicija modela osebnosti - določitev uteži za posamezne lastnosti

# Pretvorba opisnih odgovorov v številske
map_dict = {k:v for k, v in zip(options, [0, 1, 2, 3, 4])}
aQ = [map_dict[x.value] for x in children]


# Definicija uteži - to je modela
Sprejemljivost_d1 = (100 / 40) * (aQ[0] + 4 - aQ[5] + aQ[10] + 4 - aQ[15] + aQ[20] + 4 - aQ[25] + aQ[30] + 4 - aQ[35] + aQ[40] + 4 - aQ[45])
Ekstravertnost_d2 = (100 / 40) * (aQ[6] + 4 - aQ[1] + aQ[16] + 4 - aQ[11] + aQ[26] + 4 - aQ[21] + aQ[36] + 4 - aQ[31] + aQ[46] + 4 - aQ[41])
Odprtost_d3 = (100 / 40) * (aQ[2] + 4 - aQ[7] + aQ[12] + 4 - aQ[17] + aQ[22] + 4 - aQ[27] + aQ[32] + 4 - aQ[37] + aQ[42] + 4 - aQ[47])
Vestnost_d4 = (100 / 40) * (aQ[8] + 4 - aQ[3] + aQ[18] + 4 - aQ[13] + aQ[28] + 4 - aQ[23] + aQ[38] + 4 - aQ[33] + aQ[48] + 4 - aQ[43])
Nevroticizem_d5 = (100 / 40) * (aQ[4] + 4 - aQ[9] + aQ[14] + 4 - aQ[19] + aQ[24] + 4 - aQ[29] + aQ[34] + 4 - aQ[39] + aQ[44] + 4 - aQ[49])
```

```python
## Izpis
print ("Izmerjene dimenzije Velikih 5 (lastnosti) so:")
print ("Sprejemljivost:", Sprejemljivost_d1) 
print ("Ekstravertnost:", Ekstravertnost_d2)
print ("Odprtost:", Odprtost_d3)
print ("Vestnost:", Vestnost_d4)
print ("Nevroticizem:", Nevroticizem_d5)
```

# ■ Merjenje obremenitve: NASA TLX, konstrukt

#### Meri: obremenitev med aktivnostjo (med uporabniškim eksprerimentom). Za prilagajanje uporabnikom je zanimivih več dimenzij, predvsem "mental demand", "effort" in "frustration level".


#### Konstrukt: model obremenitve s šestimi dimenzijami


|Dimenzija|opis|
|:--- |:--- |
|**Mental Demand**|How much mental and perceptual activity was required? Was the task easy or demanding, simple or complex?|
|**Physical Demand**|How much physical activity was required? Was the task easy or demanding, slack or strenuous?|
|**Temporal Demand**|How much time pressure did you feel due to the pace at which the tasks or task elements occurred? Was the pace slow or rapid?|
|**Overall Performance**|How successful were you in performing the task? How satisfied were you with your performance?|
|**Effort**|How hard did you have to work (mentally and physically) to accomplish your level of performance?|
|**Frustration Level**|How irritated, stressed, and annoyed versus content, relaxed, and complacent did you feel during the task?|

# ■ Merjenje kvalitete pogovora: SUISQ, konstrukt

```python
## Estimate quality of speech interface 
# User Goal Orientation = UGO: 8 items, a = .92
# Customer Service Behaviors = CSB: 8 items, a = .89
# Speech Characteristics = SC: 5 items
# Verbosity = V: 4 items, a = .69


import numpy as np
import pandas as pd 
from scipy.optimize import minimize


# @brief reshape ND array to 1D as required by binary FA
# @arg nS number of samples
# @arg nQ number of questions
def reshape_NDto1D(a_nparr, nS, nQ):
    return a_nparr.T.reshape(nS*nQ)


# @brief reshape 1D array back to ND as required by binary FA
# @arg nS number of samples
# @arg nQ number of questions
def reshape_1DtoND(a_nparr, nS, nQ):    
    return a_nparr.reshape(nQ, nS).T


# @brief Get linear scores using direct optimisation. 
# @arg qa_mat anwsers matrix, users raws, questions columns
# @arg la_mat loadings  
# @arg cls0_trsh=0.5 ignored
# @ret S_star resulting scores
def get_linear_scores_D(qa_mat, la_mat, cls0_trsh=0.5):
    
    nS, nQ = qa_mat.shape # Number of samples and number of questions
    _, nF = la_mat.shape # Number of factors


    reg_la = 1.0 # Regularisation constant

    # Cost function    
    def cost_f(Sc_1D):
        Sc = reshape_1DtoND(Sc_1D, nS, nF)
        err_mat = qa_mat - Sc @ la_mat.T
        cost_err = np.linalg.norm(err_mat) + reg_la*np.abs(1 - np.linalg.norm(Sc))
        return cost_err


    # Minimsation
    S0 = np.ones((nS, nF)) / (1.0*np.sqrt(nS)) # To start with |S0|_2 = 1
    S0_1D = reshape_NDto1D(S0, nS, nF)
    
    
    res = minimize(cost_f, S0_1D) #, method='nelder-mead',
               #options={'xatol': 1e-8, 'disp': True})
    
    S_star = reshape_1DtoND(res.x, nS, nF) # This are scores    
    
    return S_star


# ---------------------------------------------------------------------------
# Load data 
post_fn = 'https://raw.githubusercontent.com/andrejkk/UPK_DataImgs/master/2020-2021_PredPost-vprasalnik(dvogovor)(Responses).csv'
data_df = pd.read_csv(post_fn, header=0, sep=';', encoding='utf8')

#qs_weights_fn = 'https://raw.githubusercontent.com/andrejkk/UPK_DataImgs/master/QualityOfConvSys_Weights.cvs'
#qs_weights_df = pd.read_csv(qs_weights_fn, header=0, sep=';', encoding='utf8')

qs_weights_df = pd.DataFrame(
[[0.799, 0, 0, 0],
[0, 0, 0, 0.706],
[0.8, 0, 0, 0],
[0, 0.711, 0, 0],
[0.805, 0, 0, 0],
[0, 0.758, 0, 0],
[0.628, 0, 0, 0],
[0, 0, 0, 0.701],
[0, 0.736, 0, 0],
[0.831, 0, 0, 0],
[0, 0.739, 0, 0],
[0.834, 0, 0, 0],
[0.858, 0, 0, 0],
[0, 0.726, 0, 0],
[0, 0, 0, 0.655],
[0, 0, 0.585, 0],
[0.834, 0, 0, 0],
[0, 0, 0.808, 0],
[0.794, 0, 0, 0],
[0, 0, 0.797, 0],
[0, 0.599, 0, 0],
[0, 0, 0, 0.73],
[0, 0.648, 0, 0],
[0, 0, 0.658, 0],
[0, 0.668, 0, 0]])




# Selectors 
post_qs_inds = list(range(11,36))
data_qs_df = data_df.iloc[:, post_qs_inds].drop([13])


# Estimate latent variables
qa_mat = data_qs_df.to_numpy()
la_mat = qs_weights_df.to_numpy()
speech_quality_est_np = get_linear_scores_D(qa_mat, la_mat) # was data_qs_df.dot(qs_weights_df.to_numpy())

speech_quality_est_df = pd.DataFrame(speech_quality_est_np)
speech_quality_est_df.columns = ['UGO', 'CSB', 'SC', 'V']
```

# ■ Izračun točk konstrukta ali skale




### Statistična ocena konstrukta ali skale


##### Odgovori uporabnikov

Matrika odgovorov uporabnikov $U$ na vprašanja $Q$,
$$ X = [a_{uq}], \qquad u\in U, q\in Q, $$
kjer je $U$ množica testnih uporabnikov in Q množica njihovih odgovorov na vprašanja.

##### Model je podan z matriko uteži

Matrika uteži 
$$ L = [\lambda_{qi}], \qquad q\in Q, i\in F, $$
kjer je Q množica njihovih odgovorov na vprašanja in $F$ množica faktorjev. 


##### Izračun točk (angl. sores)

Matrika točk $S=[s_ui], u\in U, i\in F$, dana z 
$$ S = X.C^{\top}.L, $$
kjer je $C$ kovariančna matrika odgovorov $X$. 




### Poenostavitve: povprečje odgovorov

Vrdnosti dimenzij so povprečja odgovorov. To je poenostavitev zgodnje formule, kjer predpostavimo, da so vse kovariance enake in vse faktorske uteži enake.

# ■ Problem vprašalnikov in merske karakteristike


#### Problem

Vprašalniki lahko dajejo **neuporabne rezultate** ker odgoovri na vprašanja niso povezani z realnim stanjem. Razlogov za to je več:
1.Pojmi v vprašanjih ali vprašanja sama **niso dobro definirani**: izrazi v človeškem jeziku lahko hkrati nosijo več pomenov in če ti niso specificirani, se odgovori nanašajo na različne pojme, objekte itd
2. V vprašanju ni jasno, na kaj se nanaša - referenca ni podana: vprašnje se lahko nanaša na kvaliteto storitve, vmesnika, vsebine itd;
3. Vprašanja so lahko za testno osebo žaljiva ali moteča
Posledično je nujno, da smo pri uporanbosti posameznih vprašanj skeptični.

#### Rešitev

Merske karakteristike zaznajo vprašanja, ki **ne delujejo**, to je ne dajejo smiselnih odgovorov. Gre za to, da merske karakteristike zaznavajo lastnosti odgovorov, ki vedno morajo veljati (npr. stabilnost ob večkratnih vprašanjih ipd) in s tem so dobre merske karkateristike **potreben pogoj za uporabnost odgovorov**, ne pa tudi zadosten pogoj.

# ■ Zanesljivost

#### Def.: Zanesljivost se nanaša na konsistentnost mere, to je ali bi isti pojem/pojav ob večkratnem merjenju izmerila dovolj podobno.


#### Vrste zanesljivosti
1. Zanesljivost v času - zanesljivost test-retest: ali večkratno testiranje da podobne rezultate (_ni uporabno za primer dvogovora z Alexo_)
2. Notranja konsistentost: ali so odgoovri med vprašanji dovolj povezani (_bistveno za primer dvogovora z Alexo_ - glejte spodaj)
3. Konsistentnost med raziskavami (raziskovalci) (_ni uporabno za primer dvogovora z Alexo_)


#### Notranja konsistentost - metoda paralelnih oblik (angl. split half)

Odgovori na vprašanje **Ali so odgovori/podatki stabilni?**

##### Potek določanja
$\def\s{{\sigma}}$
Pri tej metodi za dane podatke:
1.Testne osebe naključno razdelimo na dve polovici
2. Izračunamo korelacije med njimi $r_{12}$
3. Po Spearman-Brownovi formuli izračunamo koeficient zanesljivosti
   - za privzete enake standardne odklone
  $$ r_{tt} = \frac{2 r_{12}}{1 + r_{12}} $$
   - za različne standardne odklone
  $$ r_{tt} = \frac{4 \s_1 \s_2 r_{12}}{\s_1^2 + \s_2^2 + 2 \s_1 \s_2 r_{12}}, $$
  kjer sta $\s_1$ in $\s_2$ standardna odklona posameznih podatkov. 



#### Notranja konsistentost - Cronbach alpha

Odgovori na vprašanje **Ali vsa vprašanja/podatki skupaj merijo koherenten pojem?**

Izhodišče je razmerje prave variabilnosti $\sigma_T^2$ in izmerjene variabilnosti $\sigma_T^2 + \sigma_\varepsilon^2$, torej
$$ \frac{\sigma_T^2}{\sigma_T^2 + \sigma_\varepsilon^2}, $$
ki jo ocenimo s statističnimi formulami kot je med drugim Cronbachova $\alpha$

##### Potek določanja Cronbachove $\alpha$
Rezultate (odgovore) na $K$ vprašanj $X_i$ seštejemo v 
$$ Y = X_1 + \cdots + X_K, $$
izračunamo standardne odklone posameznih rezultatov $\s_{X_i}$ in vsote $Y_{Y}$, ter izračunamo
$$ \alpha = \frac{K-1}{K} \left(1-\frac{\sum _{i=1}^K\s_{X_i}^2}{\s_Y^2}\right) $$

```python
## Measurement characteristics
import numpy as np
import pandas as pd 
import random

# Load data 
post_fn = 'https://raw.githubusercontent.com/andrejkk/UPK_DataImgs/master/2020-2021_PredPost-vprasalnik(dvogovor)(Responses).csv'
#post_fn = 'https://raw.githubusercontent.com/andrejkk/UPK_DataImgs/master/PostQs_Responses.csv'
data_df = pd.read_csv(post_fn, header=0, sep=';', encoding='utf8').drop([13])


# Selectors 
pre_qs_inds = list(range(5,11))
post_qs_inds = list(range(11,36))


# ==============================================================================================
## Validitiy
# We estimate is OK

# ==============================================================================================
## Reliability

# Parallel forms - inter method reliability: divide test tasks to two parts and corelate anwsers 
# Split half
np.random.shuffle(post_qs_inds)
perm_inds_1, perm_inds_2 = post_qs_inds[0:12], post_qs_inds[12:24] 
anws_1, anws_2 = data_df.iloc[:, perm_inds_1], data_df.iloc[:, perm_inds_2]


# Get correlation coefficient
half1_pd = anws_1.sum(axis=1)
half2_pd = anws_2.sum(axis=1)
corr = np.corrcoef(half1_pd, half2_pd)[0,1]


# ==============================================================================================
print ('====================================================================================')
print ('== For instruments')
# Spearman–Brown prediction formula
# Equal variances
req_tt = 2*corr / (1.0 + corr)
print ('Equal var, r_tt=', req_tt)

# Non-equal variances
sd1, sd2 = np.std(half1_pd), np.std(half2_pd) 
rneq_tt = 4*sd1*sd2*corr / (sd1**2 + sd2**2 +  2*sd1*sd2*corr)
print ('Std 1 = ', sd1)
print ('Std 2 = ', sd2)
print ('Not equal var, r_tt=', rneq_tt)


## ==============================================================
# Internal consistency - Cronbach alpha
# Verifies the instrument as a whole

post_qs_df = data_df.iloc[:, 11:36]
_,K = post_qs_df.shape
X = post_qs_df.sum(axis=1)
ss_X = np.var(X, ddof=1)
ss_Yi = np.var(post_qs_df, ddof=1)
ss_Y = ss_Yi.sum()

Cronb_a = (K/(K-1))*(1.0 - ss_Y/ss_X)
print ('Cronbach Alpha =', Cronb_a)


```
