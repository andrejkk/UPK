### Chapter Objectives and Learning Outcomes

#### Objectives

The objective of this chapter is to learn about the basic concepts of service adaptation itself. We will learn about them together with important examples of personalized services.
- to learn about approaches to recommending content and services
- to understand the operation and behavior of basic algorithms

#### Learning Outcomes
Understanding and active use of the concepts of adaptation, distance between users and content, ranking, and individual personalized services.
