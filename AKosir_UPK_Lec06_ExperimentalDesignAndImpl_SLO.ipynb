## ■ Kaj je načrt poskusa

Elementi načrta poskusa so naslednji:
##### 1. Odvisna spremenljivka (odziv, angl. response) $c$: 
Kriterijska funkcija, indikator uspešnosti ipd, ki podaja glavni rezultat. V našem primeru uspešnost pogovora. 

##### 2. Kontroliran(i) faktor(ji) $F_e$: 
Neodvisne spremenljivke, ki jih v poskusu **kontroliramo**. V našem primeru je to komunikacijska naprava, torej ima kontroliran faktor vrednosti (nivoje) {PC, Alexa Echo}.   

##### 3. Moteči (angl. nuisance) faktorji: 
To so nezaželjeni vplivi na eksperimentalni rezultat, ki so neizogibni. Želimo jih nevtralizirat na več načinov;

##### 4. Vhodni šum $\varepsilon$:
To je šum, s katerim predstavimo nepoznane moteče faktorje in resnični šum, ki izhaja naključnosti človeškega obnašanja, šuma senzorjev ipd.

Pri tem: 
- postopki vključujejo tudi določitev kriterijev za izbor tesnih oseb, potebno število testnih oseb ipd. Vključujejo tudi natančen načrt poteka eksperimenta
- pred poskusom izdelamo **statistični načrt** za analizo podatkov, ki bo odgovorila za zastavljena vprašanja
- vsak eksperiment pokriva le del območja - reprezentativen je le za območje, ki ga vhodni podatki (testne osebe) predstavljajo oz. reprezentirajo;

```python
import numpy as np
from scipy.stats import binom
import matplotlib.pyplot as plt


# Plot binomial distributio
fig, ax = plt.subplots(1, 1)

n = 1000
p = 0.5
x = np.arange(binom.ppf(0.0001, n, p), binom.ppf(0.9999, n, p))
ax.plot(x, binom.pmf(x, n, p), 'bo', ms=8, label='binom pmf')
ax.vlines(x, 0, binom.pmf(x, n, p), colors='b', lw=5, alpha=0.5)


# Read critical value
alpha = 0.05
q1 = 1 - alpha/2
q2 = alpha/2
print ('Lower critical value:', binom.ppf(q1, n, p, loc=0))
print ('Upper critical value:', binom.ppf(q2, n, p, loc=0))
```

```python
## Perform statistical test
import scipy as sc
# Generate data


# Get p-value
p_in = 0.5

k = 480
x = [k, n-k]
p_val = sc.stats.binom_test(x, p=p_in)
print ('p value for k =', k, 'is: ', p_val)

k = 470
x = [k, n-k]
p_val = sc.stats.binom_test(x, p=p_in)
print ('p value for k =', k, 'is: ', p_val)

k = 465
x = [k, n-k]
p_val = sc.stats.binom_test(x, p=p_in)
print ('p value for k =', k, 'is: ', p_val)

k = 420
x = [k, n-k]
p_val = sc.stats.binom_test(x, p=p_in)
print ('p value for k =', k, 'is: ', p_val)

k = 20
x = [k, n-k]
p_val = sc.stats.binom_test(x, p=p_in)
print ('p value for k =', k, 'is: ', p_val)
```

## ■ Ničelna hipoteza, p-vrednost in stopnja tveganja $\alpha$


Glede na primer definiramo:

#### Ničelna hipoteza $H_0$

Hipoteza $H_0$ predpostavlja, da **ni učinka**, torej da ni razlike med skupinami, napravami, algoritmi, da je kovanec raven itd. 


#### p-vrednost: 

$$ p = P[\mbox{Rezultat poskusa je za dobljen rezultat poskusa ali več oddaljen od ničelne hipoteze $H_0$}\big|_{H_0}]. $$


#### Stopnja tveganja $\alpha$ in sklep o hipotezi 

Izberemo stopnjo tveganja $\alpha$, tipčno $\alpha=0.05$, in je spodnja meja verjetnosti, ki jo še sprejmemo ob interpretaciji poskusa.  
Sklepamo takole:
- če je $p<\alpha$, hipotezo $H_0$ zavrnemo;
- če je $p\geq\alpha$, hipotezo $H_0$ sprejmemo;

Če ničelno hipotezo $H_0$ zavrnemo, je odklon (razlika) **statistično značilna**.

#### Alternativna hipoteza, eno in dvostranski test

Bistvena je določitev tudi glede alternative ničelni hipotezi, ki neposredno vpliva na določitev p vrednosti. 

Alternativne hipoteze so lahko
- $H_1$ je negacija $H_0$: ker je to navadno oblike $H_1=[X\ne x_0]$, je to dvostranski test
- $H_1$ je le del negacija $H_0$: ker je to navadno oblike $H_1=[X < x_0]$ ali $H_1=[X > x_0]$, je to enostranski test

## Statistična moč in velikost učinka 


#### Statistična moč

Statistična moč je verjetnost, da ničelno hipotezo $H_0$ zavernemo ko ta ne drži, torej $P[\neg\hat{H}_0|_{\neg H_0}]$ ali
$$ pw = 1 - \beta. $$
To je torej verjetnost, da ničelno hipotezo res zavrnemo ko ta ne drži.

Odvisna je od velikosti vzorca in velikosti učinka. Uporabna je za **določanje velikosti vzorcev pred izvedbo poskusa** in za ugotavljanje **dosežene moči** po poskusu.  

Če je moč testa npr. $pw=0.8$, praktično to pomeni, da bomo resničen učinek (velikost učinka ni nič, to je $H_0$ ne drži) zares zaznali z verjtnostjo $0.8$. Torej ga bi pri ponavljanju poskusa zgrešili z verjetnostjo $1-0.8=0.2$. Pri ponavljanju testa npr. $N=60$ krat bi to bilo približno $N(1-pw) = 60\cdot 0.2 = 12$ krat.   


Inženirsko je statistična moč **občutljivost**:
- Merimo zelo majhno napetost $u=10^{-5}V$;
- Izberemo $H_0=[U=0V]$
- občutljiv merilni instrument za napetost zazna neničelno vrednost in $H_0$ zavrne - ima veliko moč. 


#### Velikost učinka (angl. effect size)

Velikost učinka je mera za velikost odklona od ničelne hipoteze. Odvisna je od posameznega testa.

## ■ Test deležev


#### Opis
Test deležev testira ali je eksperimentalno dobljen delež enak znani vrednosti $p_0$. 

#### Ničelna hipoteza $H_0$

Ničelna hipoteza je 
$$ H_0=[k/n = p_0], $$ 
kjer je $p_0$ znan delež in $k/n$ eksperimentalno dobljen delež ($n$ je število izvedb, $k$ število uspešnih izvedb).


#### Privzetki

Ni privzetkov. 



#### Ekspermentalni podatki in p-vrednost

Eksperimentalni podatek je dobljen delež $k/n$.

```python
## Test of proportions

# Data generation
import numpy as np
from scipy import stats
import random


# Generate data
random.seed()
p0 = 0.38
p1 = 0.4
n = 100
x_lst = np.zeros(n)

for i in range(n):
    if random.random() < p1:
        x_lst[i] = 1

# Get p-value
n = len(x_lst)
k = np.sum(x_lst) 
p_val = stats.binom_test(k, n=n, p=p0)        
print ('Proportion test, p-val =', p_val)
```

## ■ t-test

#### Opis
t-test testira razlike povprečij dveh Normalno porazdeljenih spremenljivk $X_1$ in $X_2$. 

#### Ničelna hipoteza $H_0$

Ničelna hipoteza je 
$$ H_0=[E(X_1) = E(X_2)], $$ 
kjer je $E(X)$ matematično upanje (pričakovana vrednost) slučajne spemenljivke $X$.


#### Privzetki

Slučajni spremenljivki $X_1$ in $X_2$ sta porazdelejni Normalno. Preverimo z Lillieforsovim testom. 


#### Variante

Obstaja 
- t-test za odvisne vzorce: spremenljivki $X_1$ in $X_2$ predstavljate isti objekt (osebo, ...), ki je izmerjen v različnih časih; 
- t-test za neodvisne vzorce: spremenljivki $X_1$ in $X_2$ predstavljata različne objekte; 

#### Ekspermentalni podatki in p-vrednost

Eksperimentalni podatki sta vzorca slučajnih spremenljivk $X_1$ (torej $x_{11}, \ldots, x_{1n}$) in $X_2$ (torej $x_{21}, \ldots, x_{2n}$). 
Testna statistika je porazdeljena po Studentovi $t$ porazdelitvi. Za določanje p vrednosti glejte primer spodaj.

```python
## T-test example

# Data generation
import numpy as np
from scipy import stats
import random


## Independent samples

# Generate data
random.seed()
mu1, sg1 = 1, 0.8 # Params of normal distributions
mu2, sg2 = 1.2, 0.8 # Params of normal distributions
n1, n2 = 100, 120
x1_lst = np.zeros(n1)
x2_lst = np.zeros(n2)

la = 10.0
for i in range(n1):
    x1_lst[i] = random.gauss(mu1, sg1) + la*random.uniform(mu1-sg1, mu2+sg1)
for i in range(n2):
    x2_lst[i] = random.gauss(mu2, sg2) + la*random.uniform(mu2-sg2, mu2+sg2)

    
# Test are assumptions true
a1, s1 = np.mean(x1_lst), np.std(x1_lst)
a2, s2 = np.mean(x2_lst), np.std(x2_lst)
st1, p1 = stats.kstest((x1_lst-a1)/s1, 'norm')
st2, p2 = stats.kstest((x2_lst-a2)/s2, 'norm')
print ('X1 Normal?, p=', p1)
print ('X2 Noirmal?, p=', p2)
    
# Get p-value
t_stat, p_val = stats.ttest_ind(x1_lst, x2_lst)        
print ('Independent t-test, p-val =', p_val)



## Dependent samples = paired test

# Generate data
random.seed()
mu1, sg1 = 1, 0.8
mu2, sg2 = 1.2, 0.8
n = 100
x1_lst = np.zeros(n)
x2_lst = np.zeros(n)

for i in range(n1):
    x1_lst[i] = random.gauss(mu1, sg1)
    x2_lst[i] = random.gauss(mu2, sg2)    

# Get p-value
t_stat, p_val = stats.ttest_rel(x1_lst, x2_lst)        
print ('Dependent t-test, p-val =', p_val)


```

## ■ F-test

$\def\o#1{{\overline{#1}}}$
$\def\SS{{\mbox{SS}}}$
$\def\Var{{\mbox{Var}}}$
#### Def.: F-test primerja pričakovane vrednosti med dvema (to je t-test) ali več podgrupama in testira ničelno hipotezo, da so pričakovane vrednosti vseh grup enake. 

Ime **analiza avriance** (angl ANalysis Of VAriance, ANOVA) je nekoliko zavajajoče in ničelna hipoteza se ne tiče varianc, ampak pričakovanih vrednosti eksperimentalnih grup. 

#### Vsota kvadratov (angl. sums of squares, SS)

Jedro razmisleka statističnih eksperimentalnih načtov je zapis variance rezultata $Y$ z vsotami kvadratov **na dva načina**, kjer drugi način vključuje odklone od pričakovanih vrednosti po eksperimentalnih grupah.   

1. Variance rezultata
$$ \Var_y = \frac{1}{n-1} \sum_{i=1}^n (y_i - \o{y})^2 = \frac{1}{n-1} \SS_{tot},  $$
kjer je $n$ število vseh meritev (rezultatov), $\Var_y$ celotna varianca rezultata, $\SS_{tot}$ pa celotna vsota kvadratov.


2. Vzemimo, da imamo eksperimentalne grupe $G_1, \ldots, G_a$, v katerih so pričakovane vrednosti $E(y)|_{G_j} = \o{y}_{G_j}$, $j=1,\ldots,a$. Vsoto kvadratov se da preurediti tako, da ločemo zapišemo vsoto kovadratov celote, znotraj skupin in napako
$$ \SS_{tot} = \SS_{treat} + \SS_{err}, $$
kjer je $\SS_{tot}$ pa celotna vsota kvadratov, $\SS_{treat}$ vsota kvadratov faktorja (varianca med skupinami) in $\SS_{err}$ vsota kvadratov napak (varianca znotraj skupin). 


#### Ničelna hipoteza $H_0$ in F-porazdelitev

Izkaže se, da če velja ničelna hipoteza
$$ H_{0} = [\o{y}_{G_1} = \cdots = \o{y}_{G_a}], $$
potem je statistika
$$ \frac{\SS_{treat}/(I-a)}{\SS_{err}/(n-a)} $$
porazdeljena po $F(I-1, n-a)$ porazdelitvi s prostostnimi stopnjami $a-1$ in $n-a$. 

To je osnova za izračun $p$ vrednosti. 


<p style="margin-bottom:2cm;"></p>
<div style="width:100%;text-align:right;font-weight:bold;font-size:1.2em;"> 15 </div>

```python
## ANOVA example

# Data generation
import numpy as np
from scipy import stats
import random


## Independent samples

# Generate data
random.seed()
mu1, sg1 = 1, 0.8 # Params of normal distributions
mu2, sg2 = 1.2, 0.8 # Params of normal distributions
mu3, sg3 = 2.0, 0.8 # Params of normal distributions
n1, n2, n3 = 100, 120, 100
x1_lst = np.zeros(n1)
x2_lst = np.zeros(n2)
x3_lst = np.zeros(n3)

la = 0.0
for i in range(n1):
    x1_lst[i] = random.gauss(mu1, sg1) + la*random.uniform(mu1-sg1, mu2+sg1)
for i in range(n2):
    x2_lst[i] = random.gauss(mu2, sg2) + la*random.uniform(mu2-sg2, mu2+sg2)
for i in range(n3):
    x3_lst[i] = random.gauss(mu3, sg3) + la*random.uniform(mu3-sg3, mu3+sg3)

    
    
# Test are assumptions true
a1, s1 = np.mean(x1_lst), np.std(x1_lst)
a2, s2 = np.mean(x2_lst), np.std(x2_lst)
a3, s3 = np.mean(x3_lst), np.std(x3_lst)

st1, p1 = stats.kstest((x1_lst-a1)/s1, 'norm')
st2, p2 = stats.kstest((x2_lst-a2)/s2, 'norm')
st2, p3 = stats.kstest((x3_lst-a3)/s3, 'norm')
print ('X1 Normal?, p=', p1)
print ('X2 Normal?, p=', p2)
print ('X3 Normal?, p=', p3)
    
    
# Get p-value
t_stat, p_val = stats.f_oneway(x1_lst, x2_lst, x2_lst)        
print ('One way ANOVA, F-test, p-val =', p_val)



```

## ■ Načrt ANOVA 

$\def\ovr#1{{\overline{#1}}}$

#### Kontrolirani in moteči faktorji

Vrste:
- fiksen učinek (angl. fixed effect): eksperimentalni faktor kontroliramo
- slučajen (angl. random effect): vrednost eksperimentalnega faktorja je naključna
- model z mešainmi učinki (angl. mixed effect): imamo faktorje s fiksnimi in slučajnimi učinki


Zaloga vrednosti: 
- diskretne vrednosti
- dve ali več vrednosti

#### Pogoji

Spremenljivke so porazdeljene normalno. 



#### Ničelna hipoteza in test

Pokrijemo le varianto **en faktor** s **fiksnim učinkom**.

Ničelna hipoteza:
$$ H_{0} = [\ovr{y}_{G_1} = \cdots = \ovr{y}_{G_I}],  $$

Test: F-test.

Pri tem je:

$\def\o#1{{\overline{#1}}}$
$\def\SS{{\mbox{SS}}}$
$\def\Var{{\mbox{Var}}}$
###### Oznake
- $n$ je število izvedb poskusa pri vsaki vrednosti od kontroliranega faktorja $F_e$. 
- $a$ je število vrednosti kontroliranega faktorja $F_e \in \{1, 2, \ldots, a\}$
- $y_{ij}$ je $j$-ti rezultat poskusa, $i=1,\ldots, n$, pri vrednosti faktorja $F_e=i$
- $y_{i.} = \sum_{j=1}^n y_{ij}$ vsote po poskusih
- $\o{y}_{i.} = y_{i.}/n$ povprečje po poskusih
- $y_{.j} = \sum_{i=1}^a y_{ij}$ vsote po vrednostih faktorja
- $\o{y}_{.j} = y_{.j}/n$ povprečje po vrednostih faktorja
- $y_{..} = \sum_{i,j} y_{ij}$ celotna vsota
- $\o{y}_{..} = y_{..}/n$ celotno povprečje

##### Vsote kvadratov
- totalna vsota kvadratov
$$ \SS_{tot} = \sum_{i=1}^a \sum_{j=1}^n (y_{ij} - \o{y_{..}})^2 $$
- vsota kvadratov faktorja
$$ \SS_{trt} = n \sum_{i=1}^a (\o{y}_{i.} - \o{y_{..}})^2 $$
- Tedaj je mogoče totalno vsoto kvadratov razdeliti na vsoto faktorja in napake $\SS_{err}$, to je
$$ \SS_{tot} = \SS_{trt} + \SS_{err} $$


##### Statistika F

Statistika
$$ F_0 = \frac{SS_{trt}/(a-1)}{SS_{err}/(n(a-1))} $$
je porazdeljena po porazdelitvi $F(a-1, n(a-1))$, od koder izračunamo $p$ vrednost. 

#### Varinate
 - enorazsežna (angl. one way): načrt ima en kontroliran faktor
 - dvorazsežna (angl. two way): načrt ima dva kontrolirana faktorja

<p style="margin-bottom:1cm;"></p>
<div style="width:100%;text-align:right;font-weight:bold;font-size:1.2em;"> 17 </div>
<img src="https://raw.githubusercontent.com/andrejkk/ORvTK_SlidesImgs/master/footer_full.jpg">

```python
# Define functions
def get_ANOVA_SS(data_df):
    a,n = data_df.shape
    N = a*n
    
    y_pp = 1.0*data_df.sum().sum()
    y_ip = 1.0*data_df.sum(axis=1)
    
    # SS total
    SS_T = (data_df**2).sum().sum() - (y_pp**2)/N

    # SS treatment
    SS_trt = (1.0/n)*(y_ip**2).sum() - (y_pp**2)/N
    
    # SS error
    SS_err = SS_T - SS_trt
   
    # Report
    return SS_trt, SS_err, SS_T
```

```python
## ANOVA

import numpy as np
import pandas as pd 
from scipy.stats import f

# Load data 
post_fn = 'https://raw.githubusercontent.com/andrejkk/UPK_DataImgs/master/ANOVAtestData31.csv'
data_df = pd.read_csv(post_fn, sep=';', encoding='utf8')


# Summs of squares
SS_trt, SS_err, SS_T = get_ANOVA_SS(data_df)

print (SS_trt, SS_err, SS_T)

# F-stat
a,n = data_df.shape
N = a*n
F_0 = (SS_trt/(a-1)) / (SS_err/(N-a))

# P-value
pVal = 1-f.cdf(F_0, a-1, N-a)

print ('p value = ', pVal)
```

Shema
- vrednosti kontroliranega faktorja so $F_e\in\{A=1, B=2, C=3, D=4\}$
- vrednosti dveh motečih faktorjev so dane po vrsticah in stolpcih, torej na mestu $ij$ je $F_1 = i$ in $F_2=j$


#### Ničelna hipoteza in test

Ničelna hipoteza:
$$ H_{0} = [\ovr{y}_{G_1} = \cdots = \ovr{y}_{G_N}],  $$


Statistični test je F-test. 

##### Oznake
- $p$ dimenzija Latinskega kvadrata, $N=p^2$
- $y_{ijk}$ izzid poskusa v vrstici $i$, stolpcu $k$ in vrednosti faktorja $k$
- $y_{.j.}$, $y_{i..}$, $y_{..k}$ kot pri oznakah ANOVA


##### Vsote kvadratov
- vsota kvadratov faktorja
$$ \SS_{trt} = \frac{1}{p}\sum_{j=1}^p y_{.j.}^2 - \frac{y_{...}^2}{N} $$
- vsota kvadratov vrstic
$$ \SS_{row} = \frac{1}{p}\sum_{i=1}^p y_{i..}^2 - \frac{y_{...}^2}{N} $$
- vsota kvadratov stolpcev
$$ \SS_{col} = \frac{1}{p}\sum_{k=1}^p y_{..k}^2 - \frac{y_{...}^2}{N} $$
- totalna vsota kvadratov
$$ \SS_{tot} = \sum_{i=1}^p \sum_{j=1}^p \sum_{k=1}^p y_{ijk}^2 - \frac{y_{...}^2}{N} $$
- vsota kvadratov napake
$$ S_{err} =  \SS_{tot} - (\SS_{row} + \SS_{col} + \SS_{trt}). $$

##### Statistika F

Tedaj je statistika
$$ F_0 = \frac{SS_{trt}/(p-1)}{SS_{err}/((p-1)(p-2))} $$
porazdelejna po poraudelitvi $F(p-1, (p-1)(p-2))$, kar uporabimo za izračun p vrednosti. 


#### Variante

- za kontrolo **treh** faktorjev lahko uporabimo Graeco-Latinski kvadrat
- obstajajo ne kvadratni načrti (Latinski pravokotnik)
- obstajajo načrti z več Latinskimi kvadrati hkrati


#### Opombe

- Pred izdelavo Latninskega kvadrata je potrebno randomizirati vrednosti (nivoje) vseh zajetih faktorjev.

<p style="margin-bottom:0.5cm;"></p>
<div style="width:100%;text-align:right;font-weight:bold;font-size:1.2em;"> 18 </div>

```python
## Latin square

import numpy as np
from scipy.stats import f

# Defien functions
def get_LatinSq_SS(LSQ_data, LSQ_des):
    p = len(LSQ_data)
    N = p*p
    inds = range(p)
    
    y_ppp = sum([sum(LSQ_data[i]) for i in inds]) # Average all
    #print ('y_ppp=', y_ppp)

    y_Rpp, y_pCp = {}, {} # Row and column averages 
    LSQ_data_T = np.array(LSQ_data).T.tolist()
    for i in inds:
        y_Rpp[i] = sum(LSQ_data[i])
        y_pCp[i] = sum(LSQ_data_T[i])
        #print ('y_Rpp[i]', i, y_Rpp[i])
        #print ('y_pCp[i]', i, y_pCp[i])


    y_ppT = {} # Letter = treatment effects
    for s in LSQ_des:
        y_ppT[s] = sum([LSQ_data[ii[0]][ii[1]] for ii in LSQ_des[s]])
        #print ('y_ppT[s]', s, y_ppT[s])

    # Sums of squares
    SS_tot = sum([LSQ_data[i][j]**2 for i in inds for j in inds]) - y_ppp**2/N
    SS_trt = (1.0/p)*sum([y_ppT[s]**2 for s in LSQ_des.keys()]) - y_ppp**2/N
    SS_row = (1.0/p)*sum([y_Rpp[i]**2 for i in inds]) - y_ppp**2/N
    SS_col = (1.0/p)*sum([y_pCp[i]**2 for i in inds]) - y_ppp**2/N
    SS_err = SS_tot - (SS_row + SS_col + SS_trt)


    # Report
    return SS_trt, SS_row, SS_col, SS_err, SS_tot
    
```

```python
## ================
# Example 1

LSQ_des = {}
LSQ_des['A'] = [[0,1], [1,0], [2,2], [3,3]]
LSQ_des['B'] = [[0,3], [1,2], [2,1], [3,0]]
LSQ_des['C'] = [[0,2], [1,3], [2,0], [3,1]]
LSQ_des['D'] = [[0,0], [1,1], [2,3], [3,2]]


# Data - After all randomisations
LSQ_data = [[192, 195, 292, 249],
            [190, 203, 218, 210],
            [214, 139, 245, 163],
            [221, 152, 204, 134]] 


# Get sums of squares
SS_trt, SS_row, SS_col, SS_err, SS_tot = get_LatinSq_SS(LSQ_data, LSQ_des)
print ('SS_row=', SS_row)
print ('SS_err=', SS_err)

# Compute stat
p = len(LSQ_data)
F_trt = (SS_trt/(p-1))/(SS_err/((p-2)*(p-1)))
F_row = (SS_row/(p-1))/(SS_err/((p-2)*(p-1)))
F_col = (SS_col/(p-1))/(SS_err/((p-2)*(p-1)))

# Report
print ('Treatment')
print ('F_trt=', F_trt)
print ('F_trt pval=', 1-f.cdf(F_trt, p-1, (p-2)*(p-1)))
print 
print ('Rows')
print ('F_row=', F_row)
print ('F_row pval=', 1-f.cdf(F_row, p-1, (p-2)*(p-1)))
print 
print ('Cols')
print ('F_col=', F_col)
print ('F_col pval=', 1-f.cdf(F_col, p-1, (p-2)*(p-1)))
```

#### Problem

Vprašalniki lahko dajejo **neuporabne rezultate** ker odgoovri na vprašanja niso povezani z realnim stanjem. Razlogov za to je več:

3. Vprašanja so lahko za testno osebo žaljiva ali moteča
Posledično je nujno, da smo pri uporanbosti posameznih vprašanj skeptični.

#### Rešitev

Merske karakteristike zaznajo vprašanja, ki **ne delujejo**, to je ne dajejo smiselnih odgovorov. Gre za to, da merske karakteristike zaznavajo lastnosti odgovorov, ki vedno morajo veljati (npr. stabilnost ob večkratnih vprašanjih ipd) in s tem so dobre merske karkateristike **potreben pogoj za uporabnost odgovorov**, ne pa tudi zadosten pogoj.

## ■ Zanesljivost

#### Def.: Zanesljivost se nanaša na konsistentnost mere, to je ali bi isti pojem/pojav ob večkratnem merjenju izmerila dovolj podobno.


#### Vrste zanesljivosti
1. Zanesljivost v času - zanesljivost test-retest: ali večkratno testiranje da podobne rezultate (_ni uporabno za primer dvogovora z Alexo_)
2. Notranja konsistentost: ali so odgoovri med vprašanji dovolj povezani (_bistveno za primer dvogovora z Alexo_ - glejte spodaj)
3. Konsistentnost med raziskavami (raziskovalci) (_ni uporabno za primer dvogovora z Alexo_)


#### Notranja konsistentost - metoda paralelnih oblik (angl. split half)

Odgovori na vprašanje **Ali so odgovori/podatki stabilni?**

##### Potek določanja
$\def\s{{\sigma}}$
Pri tej metodi za dane podatke:
1. Testne osebe naključno razdelimo na dve polovici
2. Izračunamo korelacije med njimi $r_{12}$
3. Po Spearman-Brownovi formuli izračunamo koeficient zanesljivosti
   - za privzete enake standardne odklone
  $$ r_{tt} = \frac{2 r_{12}}{1 + r_{12}} $$
   - za različne standardne odklone
  $$ r_{tt} = \frac{4 \s_1 \s_2 r_{12}}{\s_1^2 + \s_2^2 + 2 \s_1 \s_2 r_{12}}, $$
  kjer sta $\s_1$ in $\s_2$ standardna odklona posameznih podatkov. 



#### Notranja konsistentost - Cronbach alpha

Odgovori na vprašanje **Ali vsa vprašanja/podatki skupaj merijo koherenten pojem?**

##### Potek določanja
Rezultate (odgovore) na $K$ vprašanj $X_i$ seštejemo v 
$$ Y = X_1 + \cdots + X_K, $$
izračunamo standardne odklone posameznih rezultatov $\s_{X_i}$ in vsote $Y_{Y}$, ter izračunamo
$$ \alpha = \frac{K-1}{K} \left(1-\frac{\sum _{i=1}^K\s_{X_i}^2}{\s_Y^2}\right) $$

#### Eksperimentalni načt: testiranje hipotez

Eksperimentalni načrt da podatke za **testiranje hipotez**
- iz **raziskovalnega vprašanja** oblikujemo **delovno hipotezo**
  - naš primer pogovora z Alexo: 
    - raziskovalno vprašanje: ali vmesnik (Echo ali simulator) vpliva da percepcijo kvalitete dvogovora?
    - delovna hipoteza: vmesnik vpliva na percepcijo kvalitete dvogovora
- na podlagi delovne hipoteze oblikujemo **ničelno hipotezo**
  - naš primer pogovora z Alexo:
    - ničelna hipoteza(e)
    $$ H_0 = [\overline{y}_S = \overline{y}_E], $$
    kjer je $\overline{y}_S$ povprečje rezultatov skupine s simulatorjem in $\overline{y}_E$ povprečje rezultatov skupine z napravo Echo.

```python
# Define functions
def get_ANOVA_SS(data_df):
    a,n = data_df.shape
    N = a*n
    
    y_pp = 1.0*data_df.sum().sum()
    y_ip = 1.0*data_df.sum(axis=1)
    
    # SS total
    SS_T = (data_df**2).sum().sum() - (y_pp**2)/N

    # SS treatment
    SS_trt = (1.0/n)*(y_ip**2).sum() - (y_pp**2)/N
    
    # SS error
    SS_err = SS_T - SS_trt
   
    # Report
    return SS_trt, SS_err, SS_T
```
