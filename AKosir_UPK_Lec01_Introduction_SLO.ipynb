## ■ Kaj je modeliranje uporabnikov in kaj je prilagajanje uporabniku?

### Modeliranje uporabnikov

Modeliranje uporabnikov je reprezentacija uporabnikovih preferenc, preteklih aktivnosti in obnašanja, ter potreb v modelih strojnega in statističnega učenja. Tako lahko pametni sistem predvideva uporabnikove akcije in mu sklano s temi predvidevanji prilagaja storitev. 

##### Ključni gradniki so:
1. Uporabniški profili: strukturiran nabor informacij o uporabniku, kot so demografski podatki, preference, navade in cilji.
Vedenjski podatki: Podatki o tem, kako uporabniki komunicirajo s sistemom, vključno z vzorci klikanja, časom, porabljenim za naloge, in ogledano vsebino.

2. Preference in interesi: Dolgoročne preference, kot so priljubljeni žanri ali teme, ter kratkoročni interesi, ki se lahko spreminjajo glede na kontekst.
Uporabniški cilji: Cilji: kaj želi uporabnik doseči, na primer naučiti se spretnosti, kupiti izdelek ali poiskati informacije.

##### Vrste uporabniških modelov:

1. Statični modeli: med uporabo pametne aplikacije se ne spreminjajo in daljši časo ostajajo enaki. Zgrajeni so na demografskih informacijah in na že znani zgodovini uporabnikovih aktivnosti.  

2. Dinamični modeli: med uporabo pametne aplikacije se spreminjajo, pri čemer uporabljajo podatke o trenutnem stanju in aktivnostih te seanse. 


Primeri so personalizirano iskanje, aplikacije za podporo rekteativnemu štportu, priporočilni sistemi itd. 



### Prilagajanje uporabnikov

Prilagajanje uporabniku zmožnost pametnega sistema, da storitev prilagaja uporabniku na podlagi uporabnikovega modela. Cilj je prilagojena storitev, ki je za uporabnika bolj učinkovita (izboljšana ponudba) in je bolj prijetna za uporabo (boljša uporabniška izkušnja). 

##### Vrste prilagajanja:

1. Prilagajanje vsebine: sistem prilagodi vsebino storitve, na primer priporoči drugačno športno vadbo, drug film ipd. 

2. Prilagajanje vmesnika: prilagoditev izgleda in konfiguracije vmesnika, načinov interakcije (npr. poenostavitev vmesnika za uporabnike začetnike).

3. Prilagajanje nalog: prilagajanje poteka nalog, ki jih ob uporabi pametnega sistema mora opraviti uporabnik (poenostavitve, ...), prilagajanje predstavitve nalog ipd. 


##### Mehanizmi prilagajanja:

1. Sistemi, ki temeljijo na pravilih: vnaprej določen nabor pravil, ki določajo potek storitve. 

2. Modeli strojnega učenja: modeli, ki temeljijo na podatkih zgodovine uporabnika in skupine uporabnikov. Npr. ocena primernosti vsebine (filma, vadbe, ...) za posameznega uporabnika. 


Primeri aplikacij so pametni dom za podporo starejšim, personalizirani iskalnik, izobraževalna orodja, priporočilni sistem za hrano ipd. 


##### Odnos med modeliranjem in prilagajanjem uporabnikom:

Pri modeliranju gre za uporaben model njegovega obnašanja in preferenc, kar omogoča tudi razumevanje njegovih aktivnosti. Tako je ta model osnova za prilagajanje, ki je adaptacija pametne storitve uporabniku.

## ■ Delitev uporabnikove pozornosti 


<h5>
Def.: Pozornost je proces s katerim aktivno spremljamo omejeno količino vseh dražljajevin informacij, ki jih zaznamo s svojimi čutili, jih imamo shranjene v spominu ali jih pridobimo z ostalimi kognitivnimi procesi. (Sternberg, 2006) 
</h5>
V poenostavljenem jeziku je pozornost postopek, pri katerem "se z mislimi osredotočamo na trenutno dejavnost". 

<h5>
Def.: Deljena pozornost nam omogoča, da istočasno opravljamo več nalog oz. da lahko svojo pozornost osredotočamo na več dejavnikov hkrati, svojo pozornost pa nadzorovano lahko preusmerjamo od enega dejavnika/opravila k drugemu.
</h5>
V našem okviru nas zanima prehajanje pozornosti med storitvijo (MM vsebino) in upravljanjem storitve. 

<h5>
Def.: Kraja pozornosti je pojav pri komunikaciji med uporabnikom in storitvijo, pri čemer je delež deljene pozornosti za upravljanje storitve nepotrebno velik. 
</h5>
Izraz nepotrebno velik pomeni, da obstaja realna tehnološka možnost, da je delež pozornosti za upravljanje storitve manjši. 

##### Merjenje 

* Izven realnega časa: vprašalniki
* V realnem času: sledenje oči

## ■ Kontekst uporabnika


<h5>
Def.: Kontekst je opis tistih vidikov situacije uporabnika, ki so relevantni za uporabnikove preference in odločitve glede same vsebine storitve. 
</h5>

- Kaj je kontekst v komunikacijskem sistemu
    - Kar vpliva na izvedbo komunikacije
    - Med drugim: čas v dnevu, dan, vreme, čustveno stanje, socialno stanje, ...


- Formalno ima kontekst tri dimenzije:
    - Čas
    - Prostor
    - Aktvnost, <br>
  ki imajo drevesno strukturo v smislu hierahičnega dodajanja detajlov.


- Človekova komunikacija je vsebinsko in protokolno odvisna od konteksta:
    - Kontekst vpliva na želje in na učinke

    - precej zapostavljeno področje v raziskovanju

    - izjemno kompleksne razmere z velikim številom faktorjev

## ■ Izzivi na področju prilagajanja uporabnikov

#### 1. Zasebnost in varnost podatkov
Zbiranje in analiziranje podatkov o uporabnikih predstavlja veliko tveganje za zasebnost. S predpisi, kot sta GDPR in CCPA, je vse težje najti ravnovesje med personalizacijo in zasebnostjo uporabnikov.

Rešitev: Izvajanje tehnik za ohranjanje zasebnosti, kot so diferencialna zasebnost, šifriranje in anonimizacija, pomaga, vendar lahko omejijo natančnost modelov.

#### 2. Kakovost in razpoložljivost podatkov
V primeru, da je treba zagotoviti kakovostne podatke, je treba zagotoviti, da bodo ti podatki na voljo v vseh oblikah, ki jih je treba zagotoviti: Natančni modeli uporabnikov so odvisni od visokokakovostnih in izčrpnih podatkov. Vendar so lahko podatki nepopolni, nedosledni ali pristranski, kar vodi do popačenih modelov.

Rešitev: Potrebne so boljše tehnike zbiranja podatkov in postopki za obravnavo manjkajočih ali šumnih podatkov, kar pa zahteva vire in infrastrukturo.

#### 3. Prilagodljivost in dinamično vedenje uporabnikov
Izziv: Uporabnikove preference in vedenje se sčasoma spreminjajo, zato lahko statični modeli hitro zastarajo.

Rešitev: Dinamični, prilagodljivi modeli, ki se lahko posodabljajo v realnem času, so nujni, vendar to povečuje računalniško zapletenost in zahteve po virih.

#### 4. Razložljivost in preglednost

Izziv: Razlagalnost in razumljivost je treba zagotoviti z naslednjimi metodami: 1: Veliko uporabniških modelov, zlasti tistih, ki temeljijo na algoritmih strojnega učenja, je „črnih skrinjic“, zato je težko razumeti, kako pridejo do določenih zaključkov.

Rešitev: V primeru, da je treba za izračun podatkov uporabiti model, je potrebno, da je mogoče ugotoviti, ali so rezultati razumljivi: Razvoj razložljivih modelov in zagotavljanje preglednosti brez žrtvovanja učinkovitosti modela je uravnoteženo dejanje.

#### 5. Skalabilnost

Izziv: Z rastjo sistemov in povečanjem števila uporabnikov mora biti uporabniško modeliranje razširjeno, da lahko učinkovito obdeluje velike količine podatkov, ne da bi se pri tem poslabšala zmogljivost.

Rešitev: Učinkoviti algoritmi in porazdeljeno računalništvo lahko pomagajo, vendar so stroški skaliranja lahko visoki.

#### 6. Nepristranskost in pravičnost

Izziv: Modeli uporabnikov lahko nehote uvedejo ali ohranijo pristranskost, kar vodi v nepravično obravnavo določenih skupin.

Rešitev: Tehnike za odkrivanje in zmanjševanje pristranskosti se sicer izboljšujejo, vendar imajo še vedno težave z odkrivanjem vseh virov nepravičnosti v modelih.

#### 7. Integracija podatkov med platformami

Izziv: Uporabniki pogosto komunicirajo z več napravami in platformami. Povezovanje podatkov iz različnih virov za oblikovanje enotnega uporabniškega modela je težavno.

Rešitev: Zaradi različnih oblik podatkov in omejitev zasebnosti sta standardizacija in integracija podatkov med platformami stalni izziv.

#### 8. Privolitev in nadzor uporabnikov

Izziv: Zagotavljanje, da imajo uporabniki nadzor nad svojimi podatki in da razumejo, kako se uporabljajo, je vse večja skrb.

Rešitev: Oblikovanje sistemov, ki nudijo jasne mehanizme soglasja uporabnikov in hkrati omogočajo učinkovito personalizacijo, je zapleteno in pogosto premalo uporabljeno.

#### 9. Multimodalno povezovanje podatkov

Izziv: Uporabniki ustvarjajo različne podatke (besedilo, slike, govor itd.), zato je povezovanje teh različnih vrst podatkov v enoten skladen model izziv.

Rešitev: V primeru, da je treba uporabiti večpredstavnostne podatke, je to zelo zahtevno: Razvoj tehnik za obdelavo multimodalnih podatkov brez izgube pomembnih kontekstualnih informacij je še vedno v razvoju.

#### 10. Zavedanje konteksta

Izziv: Vedenje uporabnika je odvisno od konteksta (npr. lokacije, časa, naprave), zato je zajemanje konteksta za izboljšanje napovedi težavno.

Rešitev: V primeru, da se kontekst pojavi, je treba ga preveriti: Izdelava modelov, ki se zavedajo konteksta, zahteva natančno zaznavanje in interpretacijo okolice, kar zahteva veliko virov.
Reševanje teh izzivov bo zahtevalo napredek na področju tehnologije, politike in interdisciplinarnega sodelovanja.
