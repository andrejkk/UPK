## ■ Delitev uporabnikove pozornosti 


<h5>
Def.: Pozornost je proces s katerim aktivno spremljamo omejeno količino vseh dražljajevin informacij, ki jih zaznamo s svojimi čutili, jih imamo shranjene v spominu ali jih pridobimo z ostalimi kognitivnimi procesi. (Sternberg, 2006) 
</h5>
V poenostavljenem jeziku je pozornost postopek, pri katerem "se z mislimi osredotočamo na trenutno dejavnost". 

<h5>
Def.: Deljena pozornost nam omogoča, da istočasno opravljamo več nalog oz. da lahko svojo pozornost osredotočamo na več dejavnikov hkrati, svojo pozornost pa nadzorovano lahko preusmerjamo od enega dejavnika/opravila k drugemu.
</h5>
V našem okviru nas zanima prehajanje pozornosti med storitvijo (MM vsebino) in upravljanjem storitve. 

<h5>
Def.: Kraja pozornosti je pojav pri komunikaciji med uporabnikom in storitvijo, pri čemer je delež deljene pozornosti za upravljanje storitve nepotrebno velik. 
</h5>
Izraz nepotrebno velik pomeni, da obstaja realna tehnološka možnost, da je delež pozornosti za upravljanje storitve manjši. 

##### Merjenje 

* Izven realnega časa: vprašalniki
* V realnem času: sledenje oči

## ■ Kontekst uporabnika


<h5>
Def.: Kontekst je opis tistih vidikov situacije uporabnika, ki so relevantni za uporabnikove preference in odločitve glede same vsebine storitve. 
</h5>

- Kaj je kontekst v komunikacijskem sistemu
    - Kar vpliva na izvedbo komunikacije
    - Med drugim: čas v dnevu, dan, vreme, čustveno stanje, socialno stanje, ...


- Formalno ima kontekst tri dimenzije:
    - Čas
    - Prostor
    - Aktvnost, <br>
  ki imajo drevesno strukturo v smislu hierahičnega dodajanja detajlov.


- Človekova komunikacija je vsebinsko in protokolno odvisna od konteksta:
    - Kontekst vpliva na želje in na učinke
    - Kontekst vpliva na potek odločanja 


- Testna množica Ldos-CoMoDa: uporablja več kot 80 skupin po svetu
    - Izbor filmov
    - 12 kontekstnih spremenljivk
    - Ocena osebnosti v VAD prostoru

## 1.3 Kriteriji in mere uspešnosti

- Uporabnost storitve: **brez uporabne vrednosti prilagajanje nima smisla**
- Prilagajanje: glavni kriterij je visoka funkcionalnost ob minimalna kraja pozornosti
- ne gradimo nadomestka za človeško bližino!



- **ključno** domensko odvisne mere uspešnosti: 
    - sistem prilagajanja ne gradimo le na računalniških merah uspešnosti, ampak tudi na učinkih na realno življenje
    - precej zapostavljeno področje v raziskovanju



- Problem **dolgoročnih** učinkov 
    - izjemno kompleksne razmere z velikim številom faktorjev
    - longitudinalne uporabniške študije so težko izvedljive
