# ■ Introduction to user modeling

#### Def.: Modeling users is the process of building a digital representation of the user for the use and understanding of his behavior, preferences, needs and purposes.

The user's profile is a central concept within the user of customized services.It is usually dynamically evolving, renewed periodically in various periods of periods, e.g.every week.

The simplest model of the user is simply the history of its use, which is a list of its activities with time stamps.It is usually a model of machine learning that is used when adjusting.E.g.For the recommender system, this is the Pa Matrix line belonging to a given user.

Unfortunately, modern user profiles are, in the vast majority, specially related to a very narrow -specific adaptation and is practically void in other areas of adaptation.This means that the user model for adapting media content is mostly useless e.g.when choosing films.

# ■ What is segmentation, why

##### Def.: : User segmentation is a process of separating users into different subgroups or segments based on common characteristics that determine their common needs.

User segmentation is a process of sharing a larger base of users into smaller, separate groups based on common characteristics, behavior, needs or wishes.These segments help businesses or service providers understand the variety of their users and allow them to adapt more efficiently to content and services.At the same time, they are the basis for marketing strategies.

##### Why segmentation

- the basis for understanding users
- Service planning: Services are planned by segments
- Improvement of services
- ** personalization and adaptation to users **
- acquiring new users and holding on existing ones: costs of probl.6-8: 1


#### Important aspects


1. **Demographic segmentation:** Users are classified by characteristics such as age, gender, income, education or location.
2. **Geographic segmentation:** segments users based on their geographical location, which can affect linguistic preferences, cultural values ​​and purchasing behavior.
3. **Needs based on needs**: important in supporting customers and product development.
4. **Behavioral segmentation:** Divides users based on their interactions with the product or service, including samples of use, purchases history or inclusion rate.
5. **Psychographic segmentation:** Segmentation: Treats lifestyle, interests, values ​​or personality traits, which often provides a deeper insight into what promotes users' decisions.



##### execution segmentation
- Today: questionnaires + user data
- Future: Analysis of user data with mechanical and statistical methods

# ■ What is the TGI segmentation

Target Group Index segmentation is a system of questionnaires for the production segmentation, see e.g.https://www.warc.com/parters/tgi.

TGI segmentation is a tool for profiling consumers (users are also consumers), developed by Kantar Media and is often used in market research for an in -depth understanding of consumer behavior and demography.TGI segmentation divides consumers into groups based on their views, interests, lifestyles, media consumption and purchasing habits.It gives brands insight into the special features of their target groups, which enables them to make thoughtful decisions in advertising, product development and media planning.

The techniques and results of the segmentation depend largely on the goals and the selected collapse.That is why we present a Slovenian case of segmentation after TGI approach.

The framework captures data for segmentation:
- Number of respondents n = 4000 a year
- Age: 15 to 75
- Sampling: A simple random pattern, representative for Slovenia by gender, age and regions

# ■ TGI targets - what does it offer



TGI segmentation goal is
- to determine general segments
- to collect the characteristics of segments that are useful for each segment




TGI research allows:
- understand the market situation;
- identify the correct target group;
- to get to know the target group in detail and personize it;
- get to know the client's and competitive users;
- understand consumer behavior on the market;
- to understand the use of brands and their competitive advantages
- discover the best way to communicate;
- strategic and tactical media planning;
- segmentations for different markets;
