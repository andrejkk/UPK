</div>## ■ Why ethical judgment

Protect users from abuse.

I have the following elements for ethical judgment:
1. _Os, application information: _ applicant, address, field of research, etc.
2. _UVOVO: _ The applicant provides the starting points of the research work, research issues, hypotheses and other relevant introductory information;
3. _Capes: _ Number and description of the participants, method of obtaining participants (a welcome indication of inclusion and exclusion criteria) and their potential benefit from participation;
4. _Rasian plan: _ Description of the research plan (eg survey, experiment, interview, focus, etc.) and indication of the dependent (results of the procedure) and independent variables (experiment control and more);
5. _Postopek: _ In addition to the steps of the steps, the applicant also gives a list of instruments and accessories (questionnaires, sensors, etc.), description of experimental tasks of participants and instructions to participants.
6. _PROGURE FOR COOPERATION: _ Completed written form for formal consent to participation participants, abandonment of written consent, possible interview on research and withdrawal of consent;
7. _Care and confidentiality of data: _ The applicant provides the protection of captured data in all stages from measurement to storage and processing, ie complete anonymity, pseudoanonymization of data (if necessary for further data processing) or full anonymization of data, and prevention of data recognition;
8. _Ocene potential risks and benefits of cooperation: _ risks for test participants must be clearly defined, as well as their potential benefits (the latter could affect the selection of participants and, consequently, the results themselves).
9. _Prilogs: _ necessary attachments such as consent for cooperation, etc.
