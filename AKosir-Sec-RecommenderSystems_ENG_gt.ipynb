## ■ What is the recommender system and its tasks

##### Def.: : Recommender system for the user evaluates and recommends entities (content, services, ...) that this user has not yet met.

##### from user point of view
Original recommender system solves ** Overcoming information ** when the user is available ** more content ** than the user within acceptable time and attention can:
- to review
- to evaluate according to his knowledge
- to evaluate according to his knowledge of his own preferences
- to evaluate according to his knowledge of the current wishes<br/>
qn##### from technology point of view: Data matrix

The basis for building recommender systems is ** Data matrix **
$$ D = [r_{uh}]_{uh}, \quad u\in U, h\in H, $$
where $ u $ u is a crowd of users and $ h \in H $ set entities (content, service, ...) and $ r_{uh}$ is an estimate of the entity $h$ user $u$.This is either directly acquired or indirect (eg the user has long viewed the content).<br/>

$$ D =
\begin{bmatrix}
 n.d. & \ldots & \ldots & \ldots & n.d. \cr
 \vdots & \vdots & \vdots & \vdots & \vdots  \cr
 \vdots & \vdots & r_{uh} & \vdots & \vdots  \cr
 \vdots & \vdots & \vdots & \vdots & \vdots  \cr
 n.d. & \ldots & \ldots & \ldots & n.d.
\end{bmatrix} 
$$

<br/>
Viewed **algorithmic** is the problem of recommender systems How unknown values ​​n.d (= no date) replace with good estimates $r_{uh}$, so the ratings that **would be assigned** by the user $u$ to the item $h$ if evaluated by user $u$.

## ■ Formal placement of the problem of recommender systems

For a multitude of users $ U $ and a crowd entitet $ H $ are known esteurs $ r_{uh} \in S$, where $S$ is a set of ratings ($S = \{1, \ldots, 5\}$ or $S = \{bad, good \}$).

The recommender system is mapping
$$ F: U \times H \to S, $$
which to unknown item $h\in inH$ for user $u$ assigns rating $f(u, h)$.

Recommender System for Active User $ U_A $ solves an optimization problem
$$ H^* = argmax_{h \in h \setminus h_{u_a}} f (u_a, h), $$
where $h\in H$ are items that the user $u$ user has already evaluated.

The recommender system therefore selects the optimal entity between the entities that the active user $u_a$ **has not yet evaluated**.

## ■ Advantages of a substantive recommender system

##### 1. Independence from other users
The profile of each user depends only on its ratings and not on the ratings of other users.


##### 2. Transparency and explanation of recommenders
As the recommenders depend only on the profile of the active user, they can be more reliably shaped, thereby increasing the confidence of users in the recommender system itself.


##### 3. Recommender of new entities
The substantive recommender system may also recommend entities that no user has already evaluated.

## ■ Disadvantages of a substantive recommender system

##### 1. The problem of structured entity presentation

If structured presentations of entities are poor, the recommenders of the content system can be pointless.Thus, substantive recommender systems are quite dependent on domain knowledge, since good structured presentations of entities depend on the specific properties of these.E.g.The recommender system for music must take into account the specific features of music such as a music genre, style, etc.


##### 2. Overspecialisation

The basic algorithms of the substantive recommender system recommend the closest content according to past estimates, thus not being able to recommend anything really new.Thus, there is too little diversity (this will be addressed in the sub -properties of the recommender systems).


##### 3. The problem of new users

Good recommenders depend on a good user profile that can only be built from a large number of users' estimates.Thus, the content recommender system for new users cannot make reliable recommenders.This is the problem of cold startup, which we will also be pushed later in this chapter.

## ■ Algorithm based on the user neighbours


Collaborative recommender systems based on the user neighbours ​​consists on the next simple process.


For an active user $u_a \in U$ and content $h \in H$
1. Measures the distance between $ u_a $ and other users
2. Choose the surroundings $N(u_a)$: list of the most (enough) similar users
3. Calculate $r_{u_a h}$ as a weighted combination of users' estimates from $N(u_a)$.


<br/><br/>We need to perform the above process
- **Measurement of distance or.Similarities between users**
- **Equation to calculate the weighted combination of ratings**

## ■ Calculate a weighted combination of rating

To calculate the score $r\in S$ of an individual entity $h\in H$ for a user $u\in U$, which will take into account the scores of similar users, we use a prediction function. A typical example of such a function is given below.

#### Prediction function

Based on the similarity between users $sim(u_1, u_2)$, we determine the rating of entity $h$ with the equation
$$ \hat{r}_{uh} = \overline{r}_u + \frac{\sum_{u_1\in N(u)} sim(u, u_1)(r_{u_1 h} - \overline{r}_{u_1 h})}{\sum_{u_1\in N(u)} sim(u, u_1)} $$

The elements of understanding the equation are:
- the starting point of the rating is the average rating of user $\overline{r}_u$ for a given set of entities. The second term only corrects this rating;
- the correction of the score with respect to $\overline{r}_u$ is the _normalized weighted sum_ of the deviations of the scores of similar users $u_1\in N(u)$ from the own average score $\overline{r}_{u_1 h}$, i.e. the deviations $(r_{u_1 h} - \overline{r}_{u_1 h})$;

There are several variants of improvements to this equation, but the basic idea remains the same.

## ■ Advantages of a collaborative recommender system

#### Simplicity

The simplicity of the process allows for a good understanding and thus better planning.The desired properties can be installed in the equation of similarity between users.



##### Explainable recommenders

The simplicity of the process also gives good opportunities for explaining recommenders.The reasons for good scores of similar rebels are used to compose understandable explanations.



##### Efficiency

Especially for the implementation based on the surrounding area, there is no complicated machine learning process and the algorithm of recommender is easy.

##### Stability

If all users are enough, the recommenders are stable in terms of the arrival of new entities and new users.The surroundings of each user $ n (u) $ typically experience few changes.Here is also a **cold start problem** - the surrounding area can also be determined for new users, and so they also receive solid recommenders.
