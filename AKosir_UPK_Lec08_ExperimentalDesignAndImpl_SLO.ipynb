## ■ Kaj je načrt poskusa

Elementi načrta poskusa so naslednji:
##### 1. Odvisna spremenljivka (odziv, angl. response) $c$: 
Kriterijska funkcija, indikator uspešnosti ipd, ki podaja glavni rezultat. V našem primeru uspešnost pogovora. 

##### 2. Kontroliran(i) faktor(ji) $F_e$: 
Neodvisne spremenljivke, ki jih v poskusu **kontroliramo**. V našem primeru je to komunikacijska naprava, torej ima kontroliran faktor vrednosti (nivoje) {PC, Alexa Echo}.   

##### 3. Moteči (angl. nuisance) faktorji: 
To so nezaželjeni vplivi na eksperimentalni rezultat, ki so neizogibni. Želimo jih nevtralizirat na več načinov;

##### 4. Vhodni šum $\varepsilon$:
To je šum, s katerim predstavimo nepoznane moteče faktorje in resnični šum, ki izhaja naključnosti človeškega obnašanja, šuma senzorjev ipd.

#### Def. Načrt poskusa (Design of experiment - DOE) je učinkovit postopek za načrtovanje eksperimentov, ki vključuje izbiro kriterija, načrtovanja postopkov v eksperimentu, in načrt analize dobljenih rezultatov, ki vodijo v veljavne in objektivne zaključke. 

Pri tem: 
- postopki vključujejo tudi določitev kriterijev za izbor tesnih oseb, potebno število testnih oseb ipd. Vključujejo tudi natančen načrt poteka eksperimenta
- pred poskusom izdelamo **statistični načrt** za analizo podatkov, ki bo odgovorila za zastavljena vprašanja
- vsak eksperiment pokriva le del območja - reprezentativen je le za območje, ki ga vhodni podatki (testne osebe) predstavljajo oz. reprezentirajo;

## ■ Ničelna hipoteza, p-vrednost in stopnja tveganja $\alpha$


Glede na primer definiramo:

#### Ničelna hipoteza $H_0$

Hipoteza $H_0$ predpostavlja, da **ni učinka**, torej da ni razlike med skupinami, napravami, algoritmi, da je kovanec raven itd. 


#### p-vrednost: 

$$ p = P[\text{Rezultat poskusa je za dobljen rezultat poskusa ali več oddaljen od ničelne hipoteze $H_0$}\big|_{H_0}]. $$


#### Stopnja tveganja $\alpha$ in sklep o hipotezi 

Izberemo stopnjo tveganja $\alpha$, tipčno $\alpha=0.05$, in je spodnja meja verjetnosti, ki jo še sprejmemo ob interpretaciji poskusa.  
Sklepamo takole:
- če je $p<\alpha$, hipotezo $H_0$ zavrnemo;
- če je $p\geq\alpha$, hipotezo $H_0$ sprejmemo;

Če ničelno hipotezo $H_0$ zavrnemo, je odklon (razlika) **statistično značilna**.

#### Alternativna hipoteza, eno in dvostranski test

Bistvena je določitev tudi glede alternative ničelni hipotezi, ki neposredno vpliva na določitev p vrednosti. 

Alternativne hipoteze so lahko
- $H_1$ je negacija $H_0$: ker je to navadno oblike $H_1=[X\ne x_0]$, je to dvostranski test
- $H_1$ je le del negacija $H_0$: ker je to navadno oblike $H_1=[X < x_0]$ ali $H_1=[X > x_0]$, je to enostranski test

## ■ Statistični test: t-test

#### Opis
t-test testira razlike povprečij dveh Normalno porazdeljenih spremenljivk $X_1$ in $X_2$. 

#### Ničelna hipoteza $H_0$

Ničelna hipoteza je 
$$ H_0=[E(X_1) = E(X_2)], $$ 
kjer je $E(X)$ matematično upanje (pričakovana vrednost) slučajne spemenljivke $X$.


#### Privzetki

Slučajni spremenljivki $X_1$ in $X_2$ sta porazdelejni Normalno. Preverimo z Lillieforsovim testom. 


#### Variante

Obstaja 
- t-test za odvisne vzorce: spremenljivki $X_1$ in $X_2$ predstavljate isti objekt (osebo, ...), ki je izmerjen v različnih časih; 
- t-test za neodvisne vzorce: spremenljivki $X_1$ in $X_2$ predstavljata različne objekte; 

#### Ekspermentalni podatki in p-vrednost

Eksperimentalni podatki sta vzorca slučajnih spremenljivk $X_1$ (torej $x_{11}, \ldots, x_{1n}$) in $X_2$ (torej $x_{21}, \ldots, x_{2n}$). 
Testna statistika je porazdeljena po Studentovi $t$ porazdelitvi. Za določanje p vrednosti glejte primer spodaj. 


<p style="margin-bottom:2cm;"></p>
<div style="width:100%;text-align:right;font-weight:bold;font-size:1.2em;"> 13 </div>

```python
## Title: T-test example
# Goal: to understand significant difference between means 
# Hint: change mu1 and mu2 to see when the difference is significant


# Data generation
import numpy as np
from scipy import stats
import random


## Independent samples

# Generate data
random.seed()
mu1, sg1 = 1, 0.8 # Params of normal distributions
mu2, sg2 = 1.2, 0.8 # Params of normal distributions
n1, n2 = 100, 120
x1_lst = np.zeros(n1)
x2_lst = np.zeros(n2)

la = 10.0
for i in range(n1):
    x1_lst[i] = random.gauss(mu1, sg1) + la*random.uniform(mu1-sg1, mu2+sg1)
for i in range(n2):
    x2_lst[i] = random.gauss(mu2, sg2) + la*random.uniform(mu2-sg2, mu2+sg2)

    
# Test are assumptions true
a1, s1 = np.mean(x1_lst), np.std(x1_lst)
a2, s2 = np.mean(x2_lst), np.std(x2_lst)
st1, p1 = stats.kstest((x1_lst-a1)/s1, 'norm')
st2, p2 = stats.kstest((x2_lst-a2)/s2, 'norm')
print ('X1 Normal?, p=', p1)
print ('X2 Noirmal?, p=', p2)
    
# Get p-value
t_stat, p_val = stats.ttest_ind(x1_lst, x2_lst)        
print ('Independent t-test, p-val =', p_val)



## Dependent samples = paired test

# Generate data
random.seed()
mu1, sg1 = 1, 0.8
mu2, sg2 = 1.2, 0.8
n = 100
x1_lst = np.zeros(n)
x2_lst = np.zeros(n)

for i in range(n1):
    x1_lst[i] = random.gauss(mu1, sg1)
    x2_lst[i] = random.gauss(mu2, sg2)    

# Get p-value
t_stat, p_val = stats.ttest_rel(x1_lst, x2_lst)        
print ('Dependent t-test, p-val =', p_val)


```

## ■ Statistični test: test deležev


#### Opis
Test deležev testira ali je eksperimentalno dobljen delež enak znani vrednosti $p_0$. 

#### Ničelna hipoteza $H_0$

Ničelna hipoteza je 
$$ H_0=[k/n = p_0], $$ 
kjer je $p_0$ znan delež in $k/n$ eksperimentalno dobljen delež ($n$ je število izvedb, $k$ število uspešnih izvedb).


#### Privzetki

Ni privzetkov. 



#### Ekspermentalni podatki in p-vrednost

Eksperimentalni podatek je dobljen delež $k/n$. 

<p style="margin-bottom:1cm;"></p>
<div style="width:100%;text-align:right;font-weight:bold;font-size:1.2em;"> 12 </div>
<img src="https://raw.githubusercontent.com/andrejkk/ORvTK_SlidesImgs/master/footer_full.jpg">

```python
## Tite: Test of proportions
# Goal: the implementations of the test of proportions
# Hint: change p1 to see when the deviation from p0 is significant



# Data generation
import numpy as np
from scipy import stats
import random


# Generate data
random.seed()
p0 = 0.32
p1 = 0.4
n = 100
x_lst = np.zeros(n)

for i in range(n):
    if random.random() < p1:
        x_lst[i] = 1

# Get p-value
n = len(x_lst)
k = np.sum(x_lst) 
p_val = stats.binom_test(k, n=n, p=p0)        
print ('Proportion test, p-val =', p_val)
```

## ■ Napake in statistična moč

### Napake statističnega testiranja

Pri sklepanju o ničelni hipotezi na podlagi eksperimenta se lahko zgodi napaka:

<table style="width:25%">
  <tr>
    <th width="16">  </th>
    <th> $\hat{H_0}$ </th>
    <th> $\neg\hat{H_0}$ </th>
  </tr>
  <tr>
    <th> $H_0$ </th>
    <th> OK </th>
    <th> nap. I. </th>
  </tr>
  <tr>
    <th> $\neg H_0$ </th>
    <th> nap. II. </th>
    <th> OK </th>
  </tr>
</table>

Napakam se **ni mogoče izogniti**. 


##### Napaka I. vrste
Verjetnost napak I. vrste je
$$ P[nap. I.] = \alpha, $$
torej ravno stopnja tvega $\alpha$. Ta verjetnost je presenetljivo **neodvisna od velikosti vzorca**.

Praktično to pomeni, da bi pri ponavljanju poskusov test približno $\alpha \cdot N$, kjer je $N$ število poskusov, zavrnil ničelno hipotezo ko ta drži. Npr. če bi pri $\alpha=0.05$ test ponovili $N=60$ krat, bi test zavrnil ničelno hipotezo približno $0.05\cdot 60 = 3$ krat. 

##### Napaka II. vrste
Verjetnost napak II. vrste je
$$ P[nap. II.] = \beta $$


### Statistična moč

Statistična moč je verjetnost, da ničelno hipotezo $H_0$ zavrnemo ko ta ne drži:
$$ \text{pw} = 1-\beta = P[\neg \hat{H_0}\vert_{\neg H_0}] $$ 

Intuitivno gledano gre za občutljivost statističnega testa v smislu, da zavrne hipotezo ko ta ne drži tudi tedaj, ko je odklon od ničelne hipoteze majhen.

Odvisna je od velikosti vzorca in preko nje določamo velikost vzorca. 

Poznamo apriori (vnaprej - za velikost vzorca) in posteriori (potem - doseženo s testiranjem) statistično moč.

## ■ Velikost učinka in velikost vzorca (števila testnih oseb)

### Velikost učinka

Velikost učinka (angl. effect size, es) je velikost normiranega odklona od ničelne hipoteze $H_0$, ki ima po definiciji velikost učinka $0$. Za vsak test je velikost učinka določena specifično:
- T-test - razlika sredin $\text{es}=\frac{\overline{x}_1 - \overline{x}_2}{\sigma_p}$
- korelacija: $\text{es}=|r|$
- ...

Navadno določimo meje za majhno, sredno in veliko velikost učinka. 

### Določanje velikosti vzorca (števila testnih oseb) 

Minimalno velikost vzorca lahko določimo na podlagi statističnega testa (npr. t-test ali shema ANOVA) in pričakovane velikosti učinka. 

Velikost vzorca določamo z apriori statistično analizo moči, kjer določimo:
- tip testa
- ocenjeno velikost učinka 

Zelo uporaben je program G*Power (https://www.psychologie.hhu.de/arbeitsgruppen/allgemeine-psychologie-und-arbeitspsychologie/gpower), kjer izberemo družino testov, tip statistične moči in ocenjeno velikost učinka.

## ■ Izbrana shema poskusa: načrt ANOVA 

$\def\ovr#1{{\overline{#1}}}$

#### Kontrolirani in moteči faktorji

Vrste:
- fiksen učinek (angl. fixed effect): eksperimentalni faktor kontroliramo
- slučajen (angl. random effect): vrednost eksperimentalnega faktorja je naključna
- model z mešainmi učinki (angl. mixed effect): imamo faktorje s fiksnimi in slučajnimi učinki


Zaloga vrednosti: 
- diskretne vrednosti
- dve ali več vrednosti

#### Pogoji

Spremenljivke so porazdeljene normalno. 



#### Izbrana terminologija


##### Glede kontrole faktorjev
- fiksen učinek (angl. fixed effect): eksperimentalni faktor kontroliramo 
- slučajen (angl. random effect): vrednost eksperimentalnega faktorja je naključna
- model z mešainmi učinki (angl. mixed effect): imamo faktorje s fiksnimi in slučajnimi učinki


##### Glede števila faktorjev 
 - enorazsežna (angl. one way): načrt ima en kontroliran faktor
 - dvorazsežna (angl. two way): načrt ima dva kontrolirana faktorja




#### Ničelna hipoteza in test

Pokrijemo le varianto **en faktor** s **fiksnim učinkom**.

Ničelna hipoteza:
$$ H_{0} = [\ovr{y}_{G_1} = \cdots = \ovr{y}_{G_I}],  $$

Test: F-test.

Pri tem je:

$\def\o#1{{\overline{#1}}}$
$\def\SS{{\text{SS}}}$
$\def\Var{{\text{Var}}}$
###### Oznake
- $n$ je število izvedb poskusa pri vsaki vrednosti od kontroliranega faktorja $F_e$. 
- $a$ je število vrednosti kontroliranega faktorja $F_e \in \{1, 2, \ldots, a\}$
- $y_{ij}$ je $j$-ti rezultat poskusa, $i=1,\ldots, n$, pri vrednosti faktorja $F_e=i$
- $y_{i.} = \sum_{j=1}^n y_{ij}$ vsote po poskusih
- $\o{y}_{i.} = y_{i.}/n$ povprečje po poskusih
- $y_{.j} = \sum_{i=1}^a y_{ij}$ vsote po vrednostih faktorja
- $\o{y}_{.j} = y_{.j}/n$ povprečje po vrednostih faktorja
- $y_{..} = \sum_{i,j} y_{ij}$ celotna vsota
- $\o{y}_{..} = y_{..}/n$ celotno povprečje

##### Vsote kvadratov
- totalna vsota kvadratov
$$ \SS_{tot} = \sum_{i=1}^a \sum_{j=1}^n (y_{ij} - \o{y_{..}})^2 $$
- vsota kvadratov faktorja
$$ \SS_{trt} = n \sum_{i=1}^a (\o{y}_{i.} - \o{y_{..}})^2 $$
- Tedaj je mogoče totalno vsoto kvadratov razdeliti na vsoto faktorja in napake $\SS_{err}$, to je
$$ \SS_{tot} = \SS_{trt} + \SS_{err} $$


##### Statistika F

Statistika
$$ F_0 = \frac{SS_{trt}/(a-1)}{SS_{err}/(n(a-1))} $$
je porazdeljena po porazdelitvi $F(a-1, n(a-1))$, od koder izračunamo $p$ vrednost. 

#### Variante
 - enorazsežna (angl. one way): načrt ima en kontroliran faktor
 - dvorazsežna (angl. two way): načrt ima dva kontrolirana faktorja

```python
# Define functions
def get_ANOVA_SS(data_df):
    a,n = data_df.shape
    N = a*n
    
    y_pp = 1.0*data_df.sum().sum()
    y_ip = 1.0*data_df.sum(axis=1)
    
    # SS total
    SS_T = (data_df**2).sum().sum() - (y_pp**2)/N

    # SS treatment
    SS_trt = (1.0/n)*(y_ip**2).sum() - (y_pp**2)/N
    
    # SS error
    SS_err = SS_T - SS_trt
   
    # Report
    return SS_trt, SS_err, SS_T
```

```python
## Title: ANOVA design
# Goal: undesrtand the implementaiton of the ANOVA experimental design 
# Hint: direct implementation

import numpy as np
import pandas as pd 
from scipy.stats import f

# Load data 
post_fn = 'https://raw.githubusercontent.com/andrejkk/UPK_DataImgs/master/ANOVAtestData31.csv'
data_df = pd.read_csv(post_fn, sep=';', encoding='utf8')


# Summs of squares
SS_trt, SS_err, SS_T = get_ANOVA_SS(data_df)

print (SS_trt, SS_err, SS_T)

# F-stat
a,n = data_df.shape
N = a*n
F_0 = (SS_trt/(a-1)) / (SS_err/(N-a))

# P-value
pVal = 1-f.cdf(F_0, a-1, N-a)

print ('p value = ', pVal)
```

## ■ Izbrana shema poskusa: načrt Latinski kvadrat

$\def\ovr#1{{\overline{#1}}}$

Načrt Latinski kvadrat dimenzij $N\times N$ (tipično $N$ med $2$ in $4$) je eksperimentalni načrt z enim kontroliranim faktorjem in dvema motečima faktorjema, katerih učinek želimo izločit. 

Latinski kvadrat je kvadratna shema, v kateri se vrednosti kontroliranih faktorjev $A, B, C, \ldots$ v vsaki vrstici in v vsakem stolpcu pojavijo natanko enkrat. Primer kvadrate za $3\times 3$ je

|        | $F_{11}$ | $F_{12}$ | $F_{13}$ |
|--------|-----|-----|-----|
| $F_{21}$ |  A  |  B  |  C  |
| $F_{22}$ |  C  |  A  |  B  |
| $F_{23}$ |  B  |  C  |  A  |

#### Kontrolirani in moteči faktorji

Zaloga vrednosti vseh (kontroliranega in motečih faktorjev) je natanko $N$
- Kontrolirani faktor: $F\in \{A, B, C\}$
- Dva moteča faktorja: $F_1$ in $F_2$


#### Primer
- Raziskovalno vprašanje: ali ima spol vpliv na izkušnjo kvalitete dvogovora s pametnim sistemom?
- Niželna hipoteza $H_0 = [\text{Spol nima vpliv na izkušnjo kvalitete dvogovora s pametnim sistemom.}]$
- Kontroliran faktor: $F\in \{\text{ženski, moški}\}$ => latinski kvadrat bo dvodimenzionalen
- Moteči faktor Izkušnje s tehnologijo": $F_1\in \{\text{malo, veliko}\}$
- Moteži faktor Afektivno stanje: $F_2\in \{\text{negativno, pozitivno}\}$


#### Pogoji

Pogoji na minimalno velikost vzorca. 


#### Ničelna hipoteza in test

Ničelna hipoteza:
$$ H_{0} = [\text{kontroliran faktor nima učinka}]  $$

Test: F-test.

#### Eksperimentalni načt: testiranje hipotez

Eksperimentalni načrt da podatke za **testiranje hipotez**
- iz **raziskovalnega vprašanja** oblikujemo **delovno hipotezo**
  - naš primer pogovora z Alexo: 
    - raziskovalno vprašanje: ali vmesnik (Echo ali simulator) vpliva da percepcijo kvalitete dvogovora?
    - delovna hipoteza: vmesnik vpliva na percepcijo kvalitete dvogovora
- na podlagi delovne hipoteze oblikujemo **ničelno hipotezo**
  - naš primer pogovora z Alexo:
    - ničelna hipoteza(e)
    $$ H_0 = [\overline{y}_S = \overline{y}_E], $$
    kjer je $\overline{y}_S$ povprečje rezultatov skupine s simulatorjem in $\overline{y}_E$ povprečje rezultatov skupine z napravo Echo.

#### Primer: enosmerna ANOVA s fiksnim učinkom

```python
# Define functions
def get_ANOVA_SS(data_df):
    a,n = data_df.shape
    N = a*n
    
    y_pp = 1.0*data_df.sum().sum()
    y_ip = 1.0*data_df.sum(axis=1)
    
    # SS total
    SS_T = (data_df**2).sum().sum() - (y_pp**2)/N

    # SS treatment
    SS_trt = (1.0/n)*(y_ip**2).sum() - (y_pp**2)/N
    
    # SS error
    SS_err = SS_T - SS_trt
   
    # Report
    return SS_trt, SS_err, SS_T
```

```python
# one-way ANOVA fixed factor
#speech_quality_est

## ANOVA
import pandas as pd 
from scipy.stats import f


latent_fs = ['UGO', 'CSB', 'SC', 'V']
groups = {}
groups['PC'] = [0, 1, 2, 3, 4, 5, 6, 7]
groups['Echo'] = [8, 9, 10, 11, 12, 13, 14, 15]

# For all vars
for fs_n in latent_fs:
    
    # Select groups
    sq_est_PC = anws_df.iloc[groups['PC']] # speech_quality_est[fs_n][groups['PC']]
    sq_est_Echo = anws_df.iloc[groups['Echo']] # speech_quality_est[fs_n][groups['Echo']]
    
    # Compute SS
    curr_df = pd.DataFrame([np.array(sq_est_PC), np.array(sq_est_Echo)]) #, columns = [1,2,3,4,5,6,7,8])
    SS_trt, SS_err, SS_T = get_ANOVA_SS(curr_df)
    

    # F-stat
    a,n = data_df.shape
    N = a*n
    F_0 = (SS_trt/(a-1)) / (SS_err/(N-a))
    
    # P-value
    p_val = 1-f.cdf(F_0, a-1, N-a)
    
    # Report
    print ('Latent var:', fs_n, 'p-val=', p_val)
    
    # => Accept H0 => no significant difference among PC and Alexa box 
    
```
