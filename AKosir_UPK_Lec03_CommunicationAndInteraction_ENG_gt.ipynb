## ■ Communication building blocks between user and smart systems

Essential building blocks between user and smart systems are key to understanding smart systems.The technological advances of recent years in the field of natural languages ​​and large language models, in the field of sensory and in the field of machine learning, has allowed the start of the construction of smart services as a smart home, a smart support application in recreational sports, etc.


##### Input modality: text, speech, touch, gesture, sensors.

##### Output modality: visual screens, voice, haptic feedback.

##### Multimodal interaction: combination of multiple input/output methods.

##### Ana processing of natural signals: understanding and generating natural language.

##### Awareness of context: capture and recognition of time, locations (space) and basic activity

##### Modeling user: Building user models in terms of machine and statistical learning models.

##### Adjustment and adjustment: adjusting content and interfaces.

##### Ective feedback mechanisms: providing explicit and implicit feedback outside the time released and in real time.

##### Trust and Transparency: Providing user control and system transparency.

##### Cognitive load management: simplification of interaction and reducing mental effort.The user (man) has limited cognitive abilities (working memory, information flow).

#### Mechanical learning and artificial intelligence: allows the user's behavior to be predicted and the flexibility of the system.

## ■ Feedback mechanisms

Feedback mechanisms allow users to understand the condition and responses of the system, thus closing the communication loop.

- Explicit feedback: Users provide direct feedback (eg estimates, likes or comments).
- implicit feedback: System of decision satisfaction based on actions such as staying time, frequency of interactions or avoidance.
- Treatment of errors: The system informs the user errors or misunderstandings and provides the options for correction or explanation (eg, "I didn't understand this. Can you repeat?").

## ■ Description of Donald Norman

This is ** communication / interaction from the user's point of view **.The model comes from the interaction Man of the machine and meets our needs - this is exposed to adapt important steps.

Seven States / Steps:

#### set a goal

1. The user determines the target
- What does he want to achieve using the service?
- it is a goal without thinking about how to achieve this goal - so in the form of a desire;

#### execution


2. The user forms a purpose
- the purpose is already tied to the performance, that is, what service and by what activities the user will achieve their goal;
- The purpose depends on the experience of the service itself


3. The user determines the activities of the interface
- User (consciously or unconsciously) draws up a plan for activities related to the service that, in his opinion, will bring him to the destination
- The plan of activity is often or even typically changing and adapting to the situation, content, unexpected disorders from other sources, etc.


4. The user performs the action
- The user performs the service plan on the service (activities with the user interface that provides the service)
- Performing an action may involve repetitions due to unexpected system responses

#### Evaluation
5. The user detects the status of the system
- The user at the time of interaction detects and follows the state of the system


6. The user explains the status of the system
- The user (consciously or unconsciously) explains the status of the system according to the responses of the user interface, the content and the quality of the service
- This interpretation of the entire service is an important part of the service itself


7. The user evaluates the status of the system according to the target
- The user establishes a service opinion, also consciously or unconsciously
- this rating can be very biased and for the service developer ** misleading. ** Therefore, measuring quality service is a very demanding process<p style="margin-bottom:1cm;"></p>

## ■ Burnshick Lentils

The Brunshik lens model is a framework for understanding how people choose a user interface and make decisions in complex environments, and offers insight into the benefits and restrictions of human judgment.Wider viewing, the model of Brunsy's lens developed by psychologist Egon Brunsy is a conceptual framework for understanding human judgment and decision -making.It shows how individuals use environmental information (called 'characters', cues) to make or decide.The model is especially important in areas such as cognitive psychology and interaction between man and smart sites.

A dilemma between two approaches to development
1. Study of patterns of behavior according to stimuli
2. Finding the laws of behavior

Egon Burnswick Spagener has argued that the right path is only Ad 1., because 2. It is not feasible or meaningful.

This is how the term ** Burnshick's lens **, which is the relationship between the environment and the behavior of the system in this environment, was created.Burnshick has advocated to describe an individual's behavior in relation to situations and stimuli.


#### ecological validity

An important contribution to Burnshick's lens is ecological validity, which refers to how good the sign corresponds to the actual environment.Not all signs are equally reliable, and ecological validity measures how much weight should be realistically in the process of judgment.The high ecological validity sign provides accurate information about the environmental status, while the low ecological validity sign is less reliable.



##### Subjective entry of characters
Subjective entitlement of characters refers to the meaning attributed by the individual to different signs in the formation of judgment and decision.Even if the sign has a high ecological validity, the individual may not recognize or attribute it to him or her appropriate weight, which leads to the wrong judgment.
The discrepancy between subjective use of characters and the actual ecological validity of characters can lead to decision -making errors.


##### judge as a random process
The Brunshik lens model emphasizes that the judgment by its nature is random.The environment often provides incomplete or ambiguous information, so individuals make decisions based on probability and not certainty.Decisions are therefore approximations that rely on how well a person can incorporate uncertain or probability clues.

## ■ Usability (Usability)

Usability answers the question "Can anyone use this interface?".
It is a purely practical usability.
For usability, they have developed measuring instruments that we will learn as part of user measurement.


Eight rules:
#### 1 Strive for consistency.
In similar situations, a consistent sequence of actions must be required: menus and screens with the help of the same terminology should be used, and consistent commands should be used at all times.

#### 2 For frequent users to use shortcuts.
By increasing the frequency of use, the user's wishes also increase the reduction of the number of interactions and the increase in the pace of interaction.An experienced user is very helpful with abbreviations, function keys, hidden commands and macros.

##### 3 Offer informative feedback.
There must be a system feedback for each action of the operator.In the case of frequent and smaller acts, the response can be modest, and in rare and major acts the response must be greater.

#### 4 Form an information channel to complete.
The sequence of actions should be organized into groups with the beginning, the middle and the end.At the conclusion of the Group of Actions, information feedback is given to the operators with achievement, a sense of relief, a signal that the plans of actions in unpredictable events and possibilities are released from their thoughts, and a sign that the path is clear for preparing for the next group of actions.

##### 5 Offering simple defects.
As far as possible, plan the system so that the user cannot make a serious error.If an error occurs, the system must be able to detect an error and offer simple and understandable mechanisms to handle the defect.

##### 6 Allow you to return to the previous phase.
This function relieves anxiety as the user knows that errors can be undone;This encourages exploration of unknown possibilities.

#### 7 Support Internal Supervision site.
Experienced users strongly want to feel that they are responsible for the system and that the system responds to their actions.Design the system so that users are the initiators of actions and not those who respond.

#### 8 Reduce short - term memory load.
The restriction of human processing of information in short -term memory requires that the display of simple, multilateral displays are concentrated, frequency of window movement reduced and sufficient time to get to know the codes and sequences of actions.
