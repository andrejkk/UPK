## ■ Uvod

#### Def.: Pogovorni sistemi so sistemi, ki so zmožni pogovora z uporabnikom. 


Multimedio ti bo storitev **_povedala_**. V naravnem socialnem dialogu, ko jo potrebuješ in na način ko jo potrebuješ.  

#### Problemi, izzivi:

- Le razpoznava govora in sinteza govora? Ne.
- Poleg razpoznave govora še razpoznava namena itd. 
- Kako sistem izluščiti bistvo? Predstavljajte si, da ti stroj bere elektronsko pošto ali pa zadetke pri iskanju na spletu?
- Ljudje ne vedo kako komunicirat s strojem
- Če prvič dvogovor ni uspešen, se uporabniki navadno ne vrnejo
- Dober časovni potek dvogovora je težko doseči
    - težavno ugotavljanje pravih trenutkov odzivov
    - prehitri in prepočasni odzivi so zelo moteči za človeka
- Sprotno potrjevanje sprejete informacije
- Govor je počasen - veliko počasnješi od branja
- Človek je zelo občlutljiv na govor:
    - Kodiranje videa: večja občutljivost na zvok
    - besede so bliže kognitivnim procesom kot slike
- Obravnava prekinjanja govora
- Kontekstualizacija, personalizacija

## ■ Zakaj pogovorni sistemi?


#### Razlogi za uporabo

1. Zmanjševanje kraje pozornosti
2. Uporabo sistema pripeljemo bližje naravni komunikaciji
3. Posebne razmere, ko uporabnik ne more uporabljat tipkovnice in miške (vožnja, merjenje, ...)
4. Pridobivanje podatkov o uporabniku v realnem času (npr. analiza govora za čustveno stanje)


#### Kdaj so pogovorni sitemi uporabni

1. Ko je prostoročna interakcija nujna:
   - Vmesnik v vozilu
   - podporni sistemi (pomočniki) z navodili
   - Vmesnik za upravljanje in nadzor
   - Učenje tujih jezikov
   - učenje v interaktivnih okoljih


2. Ko je govor lažji of tipkanja
    - Vmesnik za glasovno iskanje
    - Virtualni pomočnik (Siri, Google Asistent)


3. Človeški agenti 
    - Klicni centri
    - Glasovni vmesnik za pomoč uporabnikom (npr. davčna vida)

## ■ Vrste pogovornih sistemov


1. Pogovorni sistemi, **usmerjeni na naloge** (angl. Task Oriented) <br>

   1.1 Metode cevovodov (angl. pipelined) 
     - Razumevanje naravnega jezika 
     - Slednje stanja dialoga 
     - Načrtovanje naslednje aktivnosti 
     - Generiranje naravnega jezika 
     
   1.2 Metode od konca do konca (angl. end-to-end)
     - Sistem strojnega učenja se nauči izbrano domeno kot celoto 
   

<br> 

2. Pogovorni sistemi, ki **niso usmerjeni na naloge** (angl. Non-Task Oriented)
   
   2.1. Nevronski generativni modeli
     - Osnovni koncept: modeliranje zaporedij
     - vključevanje konteksta
     - diverzifikacija pogovora
   
   2.2 Na osnovi iskanja informacij
     - na osnovi posamezne zahteve uporabnika
     - na osnovi več zahtev uporabnika
   
   2.3 Hibridne metode
     - Kombinacija zgornjih metod

## ■ Kaj je razpoznava govora

#### Opis: razpoznava zvočnega signala govora v seznam besed

To še ni razumevanje vsebine ali pomena govora. Gre le za pretvorbo **govora** v **tekst**.

**Vhod:** govorni signal

**Rezultat:**: seznam besed


#### Zgodovina
- 1952: Bell labs, sistem Audry, razpoznava števil (prvi sistem)
- 1970: Carnige Mellon, sistem Happy, razpoznava 1011 besed
- 1980: Uvedba Skritih markovskih modelov

## ■ Razumevanje pomena besed

##### Določanje pomena posameznih besed na podlagi seznama besed

{Prosim, belo, kavo} $\to$ _"Prosim" "belo" "kavo"_. 

Razumevanje pomena posameznih besed, besednih zvez in stavkov - v smislu dobesednega prevoda. 

**Vhod**: seznam besed

**Rezultat**: interna predstavitev pomena


<br>
To je področje [Razumevanja govora](https://en.wikipedia.org/wiki/Natural-language_understanding) (angl. Speech understanding).

## ■ Pristopi k generiranju naravnega govora


#### Na osnovi vzorcev

- **vzorec:** _Na dan DATUM so odhodi IME DRUŽBE iz LETALIŠČE1 na LETALIŠČE2 ob ČAS1 [in ČAS2 in ČAS3]_.  

Besede z velikimi črkami so izmenljive. 

- **Primer dvogovora**
  
Uporabnik: "Potrebujem vozovnico iz Ljubljane v London" <br>
Sistem: »Kdaj želite potovati?« <br>
Uporabnik: "10. marec" <br>
Sistem: "Na dan 10. marec so odhodi Lufthanse iz Brnika na Heathrow ob 11:20 in 16:10." 

<br>

#### Na osnovi vsebine

Uporabnik:"Poiščite mi hotel v Ljubljani". <br>

Sistem poišče 16 hotelov in se nato odloči na osnovi rezultati kako predstavit zadevo:

Sistem: <br>
"Na voljo je 16 hotelov, ali imate dodatne zahteve?" <br>
"Na voljo je 16 hotelov, ali želite hotel blizu centra?" <br>
"Na voljo je 16 hotelov, ali želite hotel blizu železniške postaje?" <br>

## ■ Nekatere opombe k pogovornim sistemom

#### Algoritem "deep reinforcement learnig"

Zaradi kompleksnosti jezika s uveljavlja učenje, ki maksimizira končni učinek - deep reinforced learnig

**Predstavitev problema**
- Množica stanj okolja in stanj agenta S;
- Množica akcije agenta A
- Verjetnost prehoda iz stanja $s$ v stanje $s'$ ob izvedbi akcije $a\in A$ je
$$ P_a(s, s') = Pr(s_{t+1}=s'|_{s_t=s, a_t=s}) $$
- Nagrada (takojšna) za prehod med stanjema $s$ in $s'$ ob izvedbi akcije $a\in A$
$$ R_a(s,s') $$
- pravila, ki opisujejo informacije o okolju
 
Pri pogovornem sistemu je množica stanj množica odgovorov. 
 
#### Kontekstna občutljivost govora

Napaka pri določanju konsteksta je navadno usodna zapotek pogovora. Pomembni vidiki: 
- ljudje so zelo učinkoviti pri kontekstualizaciji in hkratnem razumevanju jezika
- izgovorjava je tudi odvisna od konteksta
- pri generiranju naravnega jezika potrebujemo kontekst dovogovora za
  - izboru relevantnih vidikov dvogovora, ki jih vključimo
  - opozarjanje sogovoronika na morebitno napačno razumevanje samge konteksta 



#### Dobre lastnosti

- potencial za rešitev minimalne kraje pozornosti
- rešitev naravne komunikacije
- rešitev komunikacije / interakcije za posebne razmere 


#### Slabosti

- napake posameznih modulov se usodno širijo do rezultata
- zelo zahtevna kontekstualizacija
- če je prva izkušnja uporabnika slaba, se v velikem delžu ne vrne več
