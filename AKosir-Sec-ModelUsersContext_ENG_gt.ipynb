# ■ Introduction and description of context

#### Def.: Context is any description of a situation that in any way influences a user's decisions or preferences regarding the content or method of using a service.

<br>
Context is a preliminary (aprior) information necessary for good service adjustment.Human adaptation of communication depends significantly to the context that is easily recognized by man in real time.

# ■ Phases considering context


Context in adjustment can be used in three places:

### pre-filtration
Depending on the context, we select the information that is relevant to the adjustment according to the context.E.g.When learning the recommender system, only books read by the reader at home - if the context of "at home" is known.

#### post-fluttering
After the adaptation, we follow only those recommenders that are relevant to the day known context.E.g.In the recommender system, we choose (fitrate) only books that the reader reads at home - if the context of "at home" is known.

#### Modeling Modeling
The context is taken into account in the adaptation model itself.E.g.In the recommender system for books, the context information is directly used as an entrance to the modeling algorithm (eg contextualized matrix factorization).
