## ■ Kaj je priporočilni sistem in njegove naloge

##### Def.: Priporočilni sistem za uporabnika ocenjuje in priporoča entitete (vsebine, storitve, ...), ki jih ta uporabnik še ni srečal. 

##### Iz stališča uporabnika
Izvorno priporočilni sistem rešuje **problem preobilja informacija**, ko je uporabniku na voljo **več vsebin** kot jih uporabnik v okviru sprejemljivega časa in pozornosti zmore:
- pregledati
- ovrednotiti glede na njegovo znanje
- ovrednotiti glede na njegovo poznavanje lastnih preferenc
- ovrednotiti glede na njegovo poznavanje trenutnih želja



<br>

##### Iz stališča tehnologije: podatkovna matrika

Osnova za gradnjo priporočilnih sistemov je **podatkovna matrika** 
$$ D = [r_{uh}]_{uh}, \quad u\in U, h\in H, $$
kjer je $U$ množica uporabnikov in $h\in H$ množica entitet (vsebin, storitev, ...) in $r_{uh}$ je ocena entitete $h$ uporabnika $u$. To je ali neposredno pridobljena ocena ali posredna (npr. uporabnik si je dolgo ogledoval vsebino).

<br>

$$ D =
\begin{bmatrix}
 n.d. & \ldots & \ldots & \ldots & n.d. \cr
 \vdots & \vdots & \vdots & \vdots & \vdots  \cr
 \vdots & \vdots & r_{uh} & \vdots & \vdots  \cr
 \vdots & \vdots & \vdots & \vdots & \vdots  \cr
 n.d. & \ldots & \ldots & \ldots & n.d.
\end{bmatrix} 
$$


<br>

Gledano **algoritmično** je problem priporočilnih sistemov kako neznane vrednosti n.d (=no data) nadomestiti z dobrimi ocenami $r_{uh}$, torej ocenami, ki **bi jih uporabnik $u$ dodelil entiteti $h$, če bi jo ocenil**.

## ■ Formalna postavitev problema priporočilnih sistemov

Za množico uporabnikov $U$ in množico entitet $H$ so znane oceen $r_{uh}\in S$, kjer je $S$ množica ocen (tipično $S=\{1,2,3,4,5\}$ ali $S=\{slabo, dobro\}$). 

Priporočilni sistem je preslikava 
$$ f : U\times H \to S, $$
ki neznani entiteti $h$ za uporabnika $u$ priredi oceno $f(u,h)$.  

Priporočilni sistem za aktivnega uporabnika $u_a$ rešuje optimizacijski problem
$$ h^* = argmax_{h\in H\setminus H_{u_a}} f(u_a, h),  $$
kjer je $H_u$ množica entitet, ki jih je uporabnik $u$ že ocenil.  

Priporočilni sistem torej izbere optimalno entiteto med entitetami, ki jih aktivni uporabnik $u_a$ **še ni ocenil**.

## ■ Osnovni pristop

##### Def.: Predpostavka vsebinskega priporočilnega sistema je, da se uporabnikove preference skozi čas dovolj malo spreminjajo in tako priporoča entitete, ki so podobne entiteam, ki jih je uporabnik želel v preteklosti.

V poensotavljeni obliki algoritem za priporočanje deluje tako, da najbolje rangira entitete, ki so temu uporabniku ustrezale v preteklosti. 

**Bistveni element** vsebinskega priporočilnega sistema je očitno merjenje **razdalje med entitetami**.

## ■ Opombe k zgradbi vsebinskega priporočilnega sistema



<br><br>
Oblike **eksplicitne povratne informacije** so lahko 
1. Všečkanje: mi je všeč, mi ni všeč
2. Ocenjevanje $r$ na izbrani lestvici, tipično $\{1,\ldots, 5\}$
3. Tekstovni opis v prostem tekstu 
4. ...

<br><br>
Oblike **implicitne povratne informacije** so lahko 
1. Uporabnik si je ogledal film
2. Predčasen zaključek ogleda filma, preskakovanje delov vsebine filma
3. Razpoznava sentimenta iz teksta uporabnikovega sporočil
4. ...


<br>


Glavna naloga strojnega učenja v sodelovalnem priporočilnem sistemu je **gradnja profilov uporabnikov**. Pri tem:
- vhod v učenje profila uporabnika $u$ je **učna množica**, ki je sestavljena iz parov
$$ \{(I_h, r_h): h\in H, \text{uporabnik $u$ je ocenil vsebino $h$}\}, $$
kjer je $I_h$ strukturirana predstavitev entitete $h$ in $r_h$ njena ocena; 
- ob novih entitetah, ki jih je (implicitno ali eksplicitno) ocenil uporabnik, učno množico dopolnimo in **periodično ponavljamo gradnjo profila uporabnika**. To je neprekinjeno učenje (angl. continious learning).

## ■ Prednosti vsebinskega priporočilnega sistema

##### 1.  Neodvisnost od ostalih uporabnikov

Profil vsakega uporabnika je odvisen le od njegovih ocen in ne od ocen ostalih uporabnikov. 


##### 2. Transparentnost in pojasnjevanje priporočil

Ker so priporočila odvisna samo od profila aktivnega uporabnika, jih je mogoče bolj zanesljivo oblikovati in s tem se poveča zaupanje uporabnikov v sam priporočilni sistem.


##### 3. Priporočanje novih entitet

Vsebinski priporočilni sistem lahko priporoča tudi entitete, ki jih še ni ocenil noben uporabnik.

## ■ Slabosti vsebinskega priporočilnega sistema

##### 1. Problem strukturirane predstavitve entitet

Če so strukturirane predstavitve entitet slabe, so priporočila vsebinskega sistema lahko nesmiselna. Tako so vsebinski priporočilni sistemi precej odvisni od domenskega znanja, saj so dobre strukturirane predstavitve entitet odvisne od specifičnih lastnosti le teh. Npr. priporočilni sistem za glasbo mora upoštevati specifične lastnosti glasbe kot je glasbeni žanr, stil ipd.  


##### 2. Prevelika specializiranost

Osnovni algoritmi vsebinskega priporočilnega sistema priporočajo najbližje vsebine glede na pretekle ocene in tako ne morejo priporočiti ničesar zares novega. Tako pride do premajhne raznolikosti (to bomo obravnavali v podpoglavju Lastnosti priporočilnih sistemov).


##### 3. Problem novih uporabnikov

Dobra priporočila so odvisna od dobrega profila uporabnika, ki ga je mogoče zgraditi le iz večjega števila ocen uporabnikov. Tako vsebinski priporočilni sistem za nove uporabnike ne more dajati zanesljivih priporočil. To je problem hladnega zagona, ki ga bomo prav tako pbravnavali kasneje v tem poglavju.

## ■ Algoritem na osnovi okolic uporabnikov


Sodelobvalni priporočilni sistemi osnovi okolic uporabnikov temeljijo na naslednjem preprostem postopku. 


Za aktivnega uporabnika $u_a\in U$ in vsebino $h\in H$
1. Izmeri razdaljo med $u_a$ in ostalimi uporabniki
2. Izberi okolico $N(u_a)$: seznam najbolj (dovolj) podobnih uporabnikov
3. Izračunaj oceno $r_{u_a h}$ kot uteženo kombinacijo ocen uporabnikov iz okolice $N(u_a)$. 


<br><br>
Za izvedbo zgornjega postopka potrebujemo
- **merjenje razdalje oz. podobnosti med uporabniki**
- **enačbo za izračun utežene kombinacije ocen**

## ■ Izračun utežene kombinacije ocen

Za izračun ocene $r\in S$ posamezne entitete $h\in H$ za uporabnika $u\in U$, ki bo upoštevala ocenje njemu podobnih uporabnikov, uporabimo funkcijo predikcije. Tipičen primer take funkcije je naveden spodaj.

##### Funkcija predikcije

Na podlagi podobnosti med uporabniki $sim(u_1, u_2)$ določimo oceno entitete $h$ z enačbo
$$ \hat{r}_{uh} = \overline{r}_u + \frac{\sum_{u_1\in N(u)} sim(u, u_1)(r_{u_1 h} - \overline{r}_{u_1 h})}{\sum_{u_1\in N(u)} sim(u, u_1)} $$

Elementi razumevanja enačbe so:
- izhodišče ocene je pocprečna ocena uporabnika $\overline{r}_u$ za dan nabor entitet. Drugi člen le korigira to oceno;
- korekcija ocene glede na $\overline{r}_u$ je _normirana utežena vsota_ odklonov ocen podobnih uporabnikov $u_1\in N(u)$ od lastne povprečne ocene $\overline{r}_{u_1 h}$, torej odklonov $(r_{u_1 h} - \overline{r}_{u_1 h})$;

Obstaja več variant izboljšav te enačbe, osnovni razmislek pa ostaja enak.

## ■ Specifikacija omejitev in rešitve

#### Def.: Omejitve so pravila za dovoljene vrednosti opisov entitet , na podlagi katerih sistem lahko izbere skupino entitet za uporabnika. 


Formalna oblika takih pravil je **problem zadovoljevanja omejitev** (angl. constraint satisfaction problem) in je predstavlje kot par
$$ (V, C), $$
kjer je
- $V=V_U \cup V_{PROD}$ je množica spremenljivk;
- $C=C_R \cup C_F \cup C_{PROD}$ je omejitev, ki določajo dovoljene kombinacije vrednosti spremenljivk; 


Za razumevanje si izberimo nakup fotoaparata. 

##### Lastnosti uporabnikov $V_U$
Za priporočanje lastnosti uporabnikov. Primeri so _nivo poznavanja fotografije_, _finančne zmnožnosti_ itd.


##### Lastnosti entitet $V_{PROD}$
Lastnosti produktov so lahko tehnične specifikacije kot _Ločljivost slike_, _Podprti protokoli komunikacije, npr. micro USB_, _cena_, itd.


##### Omejitve $C_R$
Omejitve so splošne omejitve na področju npr. _ekspertiza uporabnika = nizka -> avtomatske nastavitve so na voljo_, ipd.


##### Pogoji filtriranja $C_F$
Dodatne omejitve pri izboru, npr _visoka ločljivost -> cena >_, itd


##### Omejitve entitet $C_{PROD}$
Omejitve entitet, ki držijo v splošnem, npr. _območje zaslonk -> osvetlitveni časi_ itd. 



<br>

**Rešitev problema zadovoljevanja omejitev** je **naloga priporočilnega sistema**: prireditev vrednosti spremenljivkam iz $V$ tako, da so izpolnjene omejitev it $C$.

## ■ Natančnost napovedi (angl. Prediction accuracy) (1)


#### Def. Natančnost napovedi merimo z ujemanjem priporočil sistema s "pravimi vrednostmi". 

Obstaja več različih mer natančnosti, ki so uporabne ob različnih predpostavkah. Navajamo najpomembnejše. 


#### Natančnost napovedi ocen

Za znane vrednosti ocen $r_{uh}$ parov $(u,h)\in T$ in (s priporočilnim sistemom) ocenjene vrednosti ocen $\hat{r}_{uh}$ izmerimo napako z naslednjimi merami. 


$\def\RMSE{{\text{RMSE}}} 
\def\MAE{{\text{MAE}}}$
1. **Srednja kvadratična napaka (angl. Root Mean Squared Error, RMSE)**: 
Srednja kvadratična napaka je dana z enačbo
$$ \RMSE(r,\hat{r}) = \sqrt{\frac{1}{|T|}\sum_{(u,h)\in T} (\hat{r}_{uh}-r_{uh})^2}.  $$ 


2. **Srednja absolutna napaka (angl. Mean Absolte Error, MAE)**:
Srednja absolutna napaka dana z enačbo
$$ \MAE(r,\hat{r}) = \sqrt{\frac{1}{|T|}\sum_{(u,h)\in T} |\hat{r}_{uh}-r_{uh}|}.  $$

## ■ Natančnost napovedi (angl. Prediction accuracy) (2)




#### Natančnost napovedi uporabe

Natančnost napovedi uporabnikove ocene, ki bi jo ta podal če bi dano entiteto ocenil, ni vedno indikator dobrega priporočila. Pomembnen vidik je tudi **napoved uporabe**, to je ali bo uporabnik dano entiteto res "uporabil" ali ne. 

Imamo naslednje štiri možnosti:

|                | Priporočene entitete | Ne priporočene entitete |
|----------------|:-----------:|:--------------:|
| Uporabljene entitete    |      TP     |       FN       |
| Ne uporabljene entitete |      FP     |       TN       |


Kratice pomenijo:
- TP: resnično pozitivno (angl. true postitive)
- FN: napačno negativno (angl. false negative)
- FP: napačno pozitivno (angl. false positive)
- TN: resnično negativno (angl. trune negativen)


<img style="float: center; width: 60%;" src="https://raw.githubusercontent.com/andrejkk/UPK_DataImgs/master/PrecisionRecallAcuracy_SLO.png">


<br>

Od tod izpeljemo mere


- **točnost (angl. precision)**: Kakšen delež priporočenih dokumentov je bil uporabljen?
$$ P = \frac{TP}{TP+FP} $$


- **priklic (angl. recall)**: Kakšen delež priporočenih dokumentov je bil uporabljen?
$$ R = \frac{TP}{TP+FN}  $$


- **natančnost (angl. accuracy)**: Kakšen delež dokumentov je pravilno priporočenih?
$$ A = \frac{TP+TN}{TP+FN+FP+TN}  $$

## ■ Pokritost (angl. Coverage)

Pokritost pomeni vključitev tipičnih predstavnikov uporabnikov in entitet. Zato jo delimo glede na naslednje vidike.

##### Pokritost entitet
Merimo ga z deležem entitet dane domene, ki jih priporočilni sistem lahko priporoči. Npr. v primeru filmov je to lahko delež filmov, ki jih pokriva baza IMDB. V priumeru produktov trgovine je to delež produktov glede na prodajni katalog. 

Če nas poleg deleža zanima tudi porazdelitev priporočanih entitet, potem uporabimo **Indeks Gini**. Za entitete $h_i\in H$, $n=|H|$  označimo delež, v katerem jih kupci izberejo, s $p_(h_i)$. Indeks Gini je dan z
$$ G = \frac{1}{n-1}\sum_{i=1}^n (2i-n-1) p(h_i) \in [0,1], $$
kjer so $p(h_i)$ urejeni v naraščajočem vrstnem redu. Index $G=0$ ko so vse entitete izbrane enako pogosto in $G=1$ ko je vedno izbrana ista entiteta. 

Enakomernost porazdelitve meri tudi **Entropija**, dana z
$$ H_e = -\sum_{i=1}^n  p(h_i) \log(h_i). $$
Entropija je $H=0$ ko je vedno izbrana ista entiteta in maksimalno vrednost $H=\log(n)$ ko je so vse entitete izbrane enako pogosto, to je $p(h_i)=1/n$ za vse $h_i\in H$.



##### Pokritost uporabnikov

Gre za delež uporabnikov dane množice, ki jim priporočilni sistem lahko priporoči entitete. 

Pokritost uporabnikov merimo ali 
- z deležem uporabnikov, ki jim sistem lahko priporoča
- s pokritostjo uporabniških profilov: ali so tipične uporabniške skupine zastopane s svojimi uporabniškimi profili;

##### Problem hladnega zagona

Nov uporabnik še nima zgodovine alim zelo revno zgodovino in tako mu priporočilni sistem ne more ustrezno priporočati. Če so entitete dobro zastopane pri podobnih uporabnikih, problem delno rešuje skupinski pristop. 

Problem hladnega zagona se pojavi tudi pri novih entitah v primerih, ko se priporočilni sistem zanaša na ocene uporabnikov (skupinski in vsebinski priporočilni sistem).

Lastnost raznolikosti je, da priporočene entitete **niso podobne med seboj**. Njen pomen je v tem, da uporabniku z danim naborom priporočenih entitet predstavimo **širši nabor entitet** in da s tem uporabnika hitreje pripeljemo do zanj optimalnega izbora. 

Algoritmično dovolj veliko raznolikost dosežemo z nadzorovanjem podobnosti entitet. Iz nabora najboljših (najbolje ocenjenih) entitet tako izberemo med sabo **dovolj različne**.

## ■ Uporabnost (angl. Utility)

Kriterij uporabnosti pri priporočilnih sistemih je **odvisen od domene priporočanja**. Primeri so:
- trgovina: prodaja, dobiček, ...
- turistične destinacije: ocene gostov, ...

Pri uporabnosti je pomembno, da vemo, da pri celotnem ekosistemu, katerega del je priporočilni sistem, nastopa več strank z **različnimi interesi**. Npr. pri Amazon Books:
- kupec: ustrezna knjiga, nizka cena
- tretji podnudniki knjig: visoka cena zanje
- Amazon: dobiček pri prodaji

**Določitev ustrezne mere uporabnosti** pri priporočilnem sistemu je torej najprej **strateška odločitev** in šele nato tehnična.

## ■ Robustnost (angl. Robustness)


Robustnost priporočilnega sistema se nanaša na **stabilnost sistema** glede na **lažne informacije**. Navadno gre za lažne ocene entitet, ki jih dajejo lažni uporabniki. 


Očitno obstajajo veliki poslovni in drugačni interesi za vplivanje na priporočilne sisteme. Priporočila npr. sistema _Trip Advisor_ ima resen vpliv na poslovne izide ponudnikov.    

Robustnost je tako povezana z **varnostjo podatkov** v sistemih za ocenjevanje in priporočanje.
