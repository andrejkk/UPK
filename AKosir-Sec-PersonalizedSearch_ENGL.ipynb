## ■ Kja je personalizirano iskanje, pomembni vidiki


Personalized search tailors search results to individual users based on their preferences, context, and behavior. Below are the most important aspects of personalized search:

1. User Profile and Preferences
Historical Data: Personalized search relies on the user's search history, click-through data, and interaction patterns to infer preferences.
Explicit Preferences: Some systems allow users to set preferences for topics, languages, or sources, improving relevance.
2. Context Awareness
Location: Search results often adapt based on the user's physical location (e.g., "pizza" search in New York shows nearby pizzerias).
Time: Recent or time-sensitive results may be prioritized based on when the query was made.
Device Type: Results may differ based on whether the user is on a desktop, mobile device, or smart assistant.
3. Query Understanding
Intent Recognition: Understanding whether the query is informational, navigational, or transactional.
Synonyms and Semantics: Mapping queries to related concepts (e.g., "buy phone" aligns with "smartphone deals").
Natural Language Processing (NLP): Captures subtle nuances in user queries to enhance personalization.
4. Behavior-Based Adaptation
Click Behavior: Tailoring results based on what the user clicked in previous searches.
Session Context: Adjusting recommendations based on the current browsing session.
Engagement Metrics: Measuring time spent on results to refine relevance further.
5. Collaborative Filtering
Social Signals: Incorporating preferences and behaviors of similar users.
Recommendations: Suggesting results based on trends in a similar demographic or user group.
6. Privacy and Transparency
Data Protection: Ensuring user data is handled securely and complies with privacy laws (e.g., GDPR).
User Control: Allowing users to manage or disable personalization features.
Bias Mitigation: Preventing overly narrow result sets (filter bubbles) by incorporating diverse perspectives.
7. Multi-Modal Data Integration
Search History: Integrating browsing, shopping, and social media activity.
Behavior on Other Platforms: Leveraging data from services like YouTube or Gmail in the case of integrated ecosystems.
8. Feedback Mechanisms
Allowing users to explicitly rate or refine results improves personalization accuracy over time.
9. Dynamic and Continuous Learning
Machine Learning Models: Continuously updating recommendations using algorithms like collaborative filtering, reinforcement learning, or neural networks.
Real-Time Adaptation: Immediate adjustment of results based on recent actions.
Challenges in Personalized Search
Privacy Concerns: Balancing data collection and user trust.
Bias and Fairness: Avoiding overfitting to specific behaviors and missing out on diverse content.
Relevance vs. Exploration: Balancing results that match user preferences with exposure to new information.
Personalized search enhances user satisfaction but must be carefully designed to ensure relevance, transparency, and fairness. Let me know if you'd like a deeper dive into any of these aspects!

- Vsak logični operator se da predstaviti z Booleanovimi operatorji AND, OR, NOT

Luka Dončić je v statistiko vpisal 13 točk, šest skokov ter po dve podaji in ukradeni žogi v 25 minutah igre, s čimer je bil tretji strelec moštva.

#### Logični operatorji

Poizvedba je logični izraz s ključnimi besedami

#### Operatorji bližine

Logičnio operatorji so grobi v smislu natančnejšega presejanja rezultatov, kar vodi v ali preveliko število rezultatov ali v prazen rezultat. 

Dodatne specifikacije podamo z **operatorji bližine**
- "NEAR n" dovoljuje največ n-1 vmesnih besed med dvema terminoma
- "WITH" zahteva, da sta termina zaporedna (v pravem zaporedju)
- ...

Uporaba teh operatorjev zahteva natančnejše predstavitve dokumentov, npr. poleg prisotnosti še pozicije besed. 

Primer: 

<center>  
(Luka **OR** Dončić) **(NEAR 6)** košarka **AND** (**NOT** Dragić)

## ■ Uteževanje elementov (besed)

Uteževanje besed po pomenu je **ključni mehanizem za regulacijo iskalnih reultatov**. Gre za dodeljevanje uteži besedam za **skupino vsebin**, ne za posamezno vsebino. 

Želene lastnosti:
- Beseda, ki se v dokumentu pogosto pojavi, naj ima veliko utež;
- Besede, ki pa se pojavljajo v veliko dokumentih, naj imajo nizko vrednost uteži;

#### TF.IDF uteževanje

Kratica: TF = "term frequency", IDF = "inverse document frequency"

Za element (besedo) $e_i$ v vsebini $h_j$ določimo utež
$$ w_{ij} = \text{tf}_{ij} \log\frac{N}{n_i},  $$
kjer je $w_{ij}$ utež elementa $e_i$ v vsebini $h_j$, $\text{tf}_{ij}$ število pojavitev elementa $e_i$ v vsebini $h_j$, $N$ število vseh dokumentov in $n_i$ število dokumentov, ki vsebujejo element $e_i$.

## ■ Poizvedbe v vektorskem modelu


#### Osnovni pristop
Poizvedba je skupek elementov (besed) in poizvedba poteka po naslednjih korakih:

## ■ Iskanje v verjetnostnem modelu



Za vsak dokument določimo verjetnosti model $M_D$ (verjetnosti pojavitev posameznih besed). 

Iskane poteka tako, da **dokumente rangiramo** glede na **verjetnost**, da je model dokumneta $M_D$ **generirala poizvedba** $e=(e_1, ...)$, to je
$$ P(M_D|_{e}). $$


<br>
Po Bayesovi formuli velja

$$ P(M_D|_{e}) = \frac{P(e|_{M_D}) P(M_D)}{P(e)}.  $$

Ker sta verjetnosti $P(e)$ in $P(M_D)$ znotraj posamezne poizvedbe enaki, dokumente rangiramo po verjetnosti 
$$ P(e|_{M_D}). $$
Te verjetnosti pa ocenimo neposredno iz verjetnostnega modela dokumenta.

## ■ Algoritem PageRank

#### Cilj algoritma: rangiranje strani na podlagi numeričnih vrednosti

Prikažemo poenostavljen algoritem, ki je osnova vseh nadaljnjih razvojev. 

##### Osnovni pojmi: 
- seznam s ppovezavami povezanih strani: $A_1, A_2, \ldots, A_n$
- graf povezav med stranmi: $G = (V,E)$, 
$V(G) = \{ A_i: i=1,\ldots, n \}$, 
$E(G)=\{(i,j): \text{obstaja povezava iz $A_i$ v $A_j$} \}$
- Rangiranje: verjetnost, da bo naključno klikanje pripelalo do strani $P[A]$: $Pr[A]$
- osnonvi razmislek (brez dušenja), kjer je $L(A_i, A_j)$ število povezav med stranema $A_i$ in $A_j$:
$$ P[A_i] = \sum_{(ij)\in E(G)} \frac{P[A_j]}{L(A_i, A_j)} $$
- dušenje (angl. dumping): $d=P[\text{naključno klikanje se nadaljuje}]$. Rešuje problem napačne predpostavke, da se tako klikanje zares konča. 
- dopolnjena formula za verjetnost 
$$ P[A_i] = \frac{1-d}{n} + d \sum_{(ij)\in E(G)} \frac{P[A_j]}{L(A_i, A_j)} $$


##### Algoritem

Poenostavlje algoritem: ignorira povezave znotraj dokumenta in vzporedne povezave šteje kot eno samo.

Zakaj iteracija: ker so formule iterativne.

Vhod: graf dokumentov $G=(V, E)$

Rezultat: točkovanje (scores) strani $P[A]$

Koraki
1. $P[A_i] = \frac{1}{n}$
2. While ($\Delta (P[A_i]) < \varepsilon$):
3. $\quad$ for ($i\in V(G)$): 
4. $\quad\quad$ $P[A_i] = \frac{1-d}{n} + d \sum_{(ij)\in E(G)} \frac{P[A_j]}{L(A_i, A_j)}$
5. EndWhile

Tipično število iteracij: $50$

```python
# Primer PageRank algoritma v Pythonu
# Obravnava preprost primer štirih dokumentov


import networkx as nx
import matplotlib.pyplot as plt
```

```python
import numpy as np


page_lst = ["A1", "A2", "A3", "A4"]
pageG = nx.DiGraph()
pageG.add_nodes_from(page_lst)
pageG.add_edges_from([('A1', 'A2'), ('A1', 'A3'), ('A1', 'A4'), ('A3', 'A4'), ('A2', 'A4'),('A4', 'A3'),])
nx.draw(pageG, with_labels=True)
plt.show()
```

```python
prs = nx.pagerank(pageG)
print ('Page rank scores: ', prs)
```
