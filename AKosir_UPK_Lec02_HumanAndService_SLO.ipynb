Če želimo **storitev prilagajati uporabniku**, potrebujemo
- njegove osnovne lastnosti: osebnost
- trenutno stanje uporabnika: razpoloženje, čustva, izbrane socialne signale

Bistvena lastnosti lastnosti uporabnika je njena **hitrost spreminjanja v času**:
- osebnost: se ne spreminja
- razpoloženje: spreminja se počasi (skozi minute)
- čustva: spreminja se hitro (skozi desetinke sekunde)

Način in **pogostnost merjenja** je odvisna od hitrosti spreminjanja.

##  ■ Uporabnik, informacija in kognitivni procesi


Uporabnik med uporabo naprave z njo **komunicira**, npr. ko bere knjigo in poizveduje za v knjigi omenjenimi osebami. Med tem **procesira informacije**, in informacije tečejo v obe smeri - to je komunikacija.  

#### Komponente procesiranja informacij so

1. Občutenje: zaznavanje dražljajev iz okolja.

2. Zaznavanje: interpretacija čutnih informacij.

3. Pozornost: osredotočanje na določene dražljaje.

4. Delovni spomin: Kratkoročno shranjevanje in upravljanje informacij.

5. Dolgoročni spomin: Shranjevanje informacij za daljše časovno obdobje.

6. Kodiranje informacije: pretvorba informacij v spomin.

7. Priklic: Dostop do shranjenih informacij.

8. Kognitivna obremenitev: Učinek: miselni napor, ki se uporablja pri obdelavi informacij.

9. Sprejemanje odločitev: Odločanje: Izbira ukrepov na podlagi informacij.

10. Reševanje problemov: Reševanje problemov: iskanje rešitev za reševanje problemov.

11. Metakognicija: Prepoznavanje: zavedanje in uravnavanje lastnega mišljenja.



#### Nekateri vidiki

1. Informacije v obe smeri
    - vizualni, zvočni, haptični, gibanje, ...
    

2. Informacije, shranjene v pomnilniku
    - senzorni spomin, kratkoročni spomin, dolgoročni spomin

3. Obdelava in uporaba informacij
    - sklepanje, reševanje problemov, spretnost, napake   
    - odzivni in reakcijski čas <br>

    **odzivni čas = reakcijski čas + čas gibanja za izvedbo akcije**

    - človek zmore približno 120 bit/sek. Človeški govor je pribl. 60 bit/sek.

##  ■ Uvod


##### Def.: Osebnost je karekterističen nabor obnašanj, kognicije in čustvenih vzorcev. Tvori osebne lastnosti posameznika. 


- Osebnost opisuje časovno nespremenljive lastnosti uporabnika. Dovolj jo je "izmeriti" enkrat na nekaj let. 

- Namen uporabe osebnosti v prilagajanju je upoštevanje relevantnih lastnosti uporabnika. Npr. dimenziji nevroticizem in odprtost dobro napovedujeta ali bo posameznik odprl blog ali ne.  


##### Zakaj osebnost: 

Osebnostne lastnosti kot so introvertiranost / ekstrovertiranost so **morda koristna informacija za prilagajanje**. 

Izraz koristna informacija pomeni, da sta optimalno prilagojeni storitvi za dva **uporabnika z raličnima osebnostnima različni**. 


<br>

_Literatura:_ Vinciarelli, A., and Mohammadi, G. A survey of personality computing. IEEE Transactions on Affective Computing, 5 (3). pp. 273-291, 2014.

##  ■ Model osebnosti Velikih 5 (angl. Big5)


Model osebsnoti Velikih 5 je **najbolj uporabljan model v interakciji človeka z napravami**. 

#### Značilnosti - dimenzije
- _Sprejemljivost:_ ohranjanje pozitivnih odnosov z drugimi. Sposobnost razumevanja z drugimi osebami.
- _Ekstravertnost:_ Ekstravertne osebe se v veliki meri zanimajo za družabne dogodke, so zelo aktivne in energične
- _Odprtost:_ osredotoča se na posameznikovo kulturno sofisticiranost in posameznikovo odprtost za nove izkušnje.
- _Vestnost:_ odgovornost, naklonjenost pravilom, vztrajnost, pozornost…
- _Nevroticizem:_ visoko izražen nevroticizem pomeni manjšo sposobnost obvladovati čustva, več anksioznosti

## ■ Kaj je razpoloženje


##### Def.: "Razpoloženje je čutstveno stanje". V nasprotju s čustvi je manj specifično določeno in relacije do vzrokov, ki ga povzročajo, so manj znane. Navadno govorimo le o pozitivnem in o negativnem razpoloženju, ki ima manjšo ali večjo amplitudo. 

Razpoloženje navadno merimo le kot negtivno in pozitivno, kar pomeni, da za uporabno oceno **zadoščajo že precej revni vhodni podatki** kot npr. del teksta, ki ga je uporabnik napisal ali kratek govorni posnetek. 

Na drugi strani pa je za prilagajanje storitev **dovolj le ta podatek o predznaku** in je amplituda tega razpoloženja manj pomembna. 

##### Merjenje razpoloženja
Zaradi vloge razpoloženja navadno ugotavljamo (merimo) le pozitivno oz. negativno razpoloženje. Tako je merjenje razpoloženja v veliki meri izpeljano iz merjenja 
- razpoloženje iz teksta
- razpoločenje iz socialnih omrežij
- afektivnega stanja 
- ...


<br>

#####  Sentiment in razpoloženje:
Nekonsistentno izrazoslovje. Bolje definirano je razpoloženje (angl. mood). Sentiment rezerviramo za binarno operacijo - odnos uporabnika do izbranih vsebin in/ali storitev. Več kasneje.

## ■ Kaj je afektivno in kaj čustveno stanje


##### Def.: Afektivno stanje je odziv organizma na vzbujanje.

Pri tem imamo v mislih nizkonivojski biološki odziv, na nivoju samih biokemičnih signalov v telesu. Ali je to ustrezen opis ali ne, se stroka **ne strinja**.


##### Def.: Čustveno stanje je zavedni del izkustva. Znanstvene definicije so zelo različne in tudi nasprotujoče si. 



V rabi so različne teorije čustvenih stanj - med drugim: 
- James-Lange: čustva so naša interpretacija psihološkega odziva na stimuli 
- Cannon: čustva so naš psihološki odziv na stimuli
- Schacter-Singer: čustva so rezultat ovrednotenja psihološkega odziva na stimuli, upoštevaje celotno situacijo


Čustva vključujejo tako kognitivne kot fizične odzive na stimuli.



##### Uporabna postavitev za prilagajanje uporabnikom
- merimo afektivno stanje: VAD prostor (podrobnosti kasneje)
- po potrebi (v postopku prilagajanja) to preslikamo v čustveno stanje. Procesno gledano ni potrebno, koristno je zaradi interpretacije

## ■ Modeli čustvenih stanj

Tako kot pri afektivnih stanjih, tudi pri čustvenih stanjih obstaja več modelov. Zaradi zgodovinskega razvoja je teh modelov veliko in jih razumemo na podlagi razlogov, zaradi katerih so nastali.

Izbrani modeli čustvenih stanj so
- _Circumpleksni model_ : Preslikava iz VAD prostora (**najvažnejši za naš okvir**)
- _Vectorski model_ : 2D model smeri čustev
- _Positivna aktivacija – negativna aktivacija (PANA) model_ : negativen afekt in pozitiven afekt sta samostojna pojava
- _Plutchikov model_ : vključuje tako osnovna čustva kot dimenzijski model
- _PAD model_ : 3D model z osmi zadovoljstvo, vzburjenje in dominantnost 
- _Lövheimova kocka čustev_ : preslikava bio substanc dopamin, noradrenalin in serotonin v osem osnovnih čustev
- _Cowen & Keltner_ : 27 kategorij na podlagi analize odzivov posameznikov, ki so skupaj gledali 2185 video posnetkov

## ■ Ocenjevalna teorija čustev 

Ta teorija trdi, da čustva **niso neposreden biološki odziv na dražljaj**, ampak rezultat **višjenivojskega ovrednotenja teh dražlajev**.  

Ima več pomembnih posledic za področje prilagajanja:
- vloga čustev v komunikaciji
- vloga čustev pri ocenjevanju stanja uporabnika



Privzetek ocenjevalne teorije čustev (pri čustvenem odzivu sodelujejo tudi višje kognitivne sposobnosti) za prilagajanje pomeni
- kompleksnejše rešitve: trenutno stanje razvoja jih še ne omogoča 
- izboljšane možnosti za pridobivanje podatkov o uporabniku

Več faz ocenjevaja:
1. Primarna ocena: ocena učinka dogodka - koristen, škodljiv, nevaren, ...
2. Sekundarna ocena: ocena okoliščin, posledic in možnosti za odziv
3. Generiranje čustev: generiranje čustvenega stanja
4. Ponovna kognitivna ocena

- ocenjevalne (angl. appraisal) - najbolj primerno za tehnologijo

##  ■ Kaj je sentiment?


##### Def.: Sentiment je naravnanost uporabnika glede na entitete, ki so posamezniki, organizacije, produkti, dogodki, tematike, vsebine ter njegova naravnanost do njihovih lastnosti. Torej je sentiment uporabnika vezan na določeno entiteto. 

Taka entiteta je tudi multimedijska vsebina, njen avtor, storitev ki jo prinaša (npr. Facebook), itd. 

Uporabnik ima lahko **različne sentimente do različnih entitete** ob **istem času**. Take lastnosti npr. čustveno stanje nima - to je v vsakem trenutku določeno eno samo in je torej stanje uporabnika in ni nujno vezano na eno samo entiteto ali sploh katero entiteto. 


Ta naravnanost je povezana z afektivnimi in čustvenimi stanji in razpoloženjem, z mnenjem itd. 


Sentiment navadno merimo le v predznaku, torej **pozitiven** in **negativen**. 


Analiza sentimenta se navadno nanaša na **_dokument_**: tekst, govor, zaporedje tweetov, itd

Že iz opisa posameznih pojmov (spodaj) je jasno, da je pri uporabniku prilagojeni komunikaciji kognicija pomembna. Brez kognitivnih virov uporabnik ne more konzumirati vsebine niti upravljati s storitvijo, ki jo ta prinaša. 

#### Deljena pozornost

Uporabniku omogoča, da istočasno opravljao več nalog. Seveda za ceno manjše uspešnosti posamezne naloge. Stroka se še vedno ne strinja o veljavnem modelu deljene pozornosti. Pri komunikaciji s pametnimi sistemi gre za deljeno pozornost predvsem med upravljanjem sistema (nezaželjen del) in samo storitvijo (konzumiranje MM vsebine; zaželjen del).


#### Selektivna pozornost

Selektivna pozornost uporabniku omogoča, da se osredotoči na tisto dejavnost, ki jo je sam izbral. Gre za sposobnost, da kljub motečim dejavnikom uporabnik s pozornostjo uporabnik vztraja na zastavljenem cilju (izbranem cilju ali najvažnejšem cilju v danem trenutku).


#### Delovni spomin

Delovni spomin uporabniku omogoča, da informacije preko senzorike začasno shrani z namenom organizacije v strukturo in nato sočasno uporabo pri kognitivnih procesih. Ker sta glavna vira informacij akustične informacije in vidno-prostorske informacije, navadno tudi delovni spomin delimo na ta dva dela. Razvoj novih interakcijskih možnosti pa zahteva (bo zahteval) prilagoditev teh modelov. 

V okviru delovnega spomina so posamezni podprocesi sorazmerno dobro znani. Glavna vidika sta
* kapaciteta: $7 \pm 2$ enoti. Velike razlike med posamezniki, zadeva je merljiva; 
* čas hrambe: $10 - 15$ sekund, lahko do ene minute. Razlike med posamezniki in razlike med razmerami (prisotnost motenj, utrujenost, ...)

##  ■ Teorija večih virov in preklapljanje

#### Katere vire si delimo?

##### Avdio (jezik) in video (vidno):

V osnovi so viri za avdio in vidno informacijo v večji meri ločeni. Tako naloge v zvezi z jezikom (branje, poslušanje, ...) tekmujejo za avdio vire, naloge v zvezi z vidno percepcijo pa za vizualne vire.

## ■ Kaj je storitev in kaj naprava


##### Def.: Naprava je strojna oprema in strojno-programska oprema (angl. firmware), ki omogoča njeno delovanje.
Ta opis je zelo blizu pojmu v običajnem pogovornem jeziku. Naprava je torej računalnik, tablica, bralnik, pametni telefon, prikazovalnik v avtu itd. 



##### Def.  Storitev sestavljajo funkcionalnosti (npr. prikaz multimedijskih vsebin) in naprava, ki tak prikaz omogoča (npr. bralnik knjig). 
Storitev je celotna ponudba, naprava je njen nujni del, ki omogoča funkcionalnosti. Storitev je torej prikaz elektronskih knjig na bralniku, ki omogoča še brskanje po spletu za npr. omenjenimi osebami v knjigi.
