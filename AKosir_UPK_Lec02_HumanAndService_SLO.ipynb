##  ■ Uvod



Če želimo **storitev prilagajati uporabniku**, potrbujemo
- njegove osnovne lastnosti: osebnost
- trenutno stanje uporabnika: razpoloženje, čustva, izbrane socialne signale
- kontekst

V tem poglavju bomo spoznali vse razen konteksta in socialnih signalov (posebna poglavja).



Bistvena lastnosti lastnosti uporabnika je njena **hitrost spreminjanja v času**:
- osebnost: se ne spreminja
- razpoloženje: spreminja se počasi (skozi minute)
- čustva: spreminja se hitro (skozi desetinke sekunde)

Način in **pogostnost merjenja** je odvisna od hitrosti spreminjanja. 



##### Lastnost ali stanje uporabnika, ki ni merljivo / določljivo, ni uporabno pri prilagajanju.

- Obdelava in uporaba informacij
    - sklepanje, reševanje problemov, spretnost, napake

##  ■ Uvod


##### Def.: Osebnost je karekterističen nabor obnašanj, kognicije in čustvenih vzorcev. Tvori osebne lastnosti posameznika. 


- Osebnost opisuje čaosvno nespremenljive lastnosti uporabnika. Dovolj jo je "izmeriti" enkrat na nekaj let. 

- Namen uporabe osebnosti v prilagajanju je upoštevanje relevantnih lastnosti uporabnika. Npr. dimenziji nevroticizem in odprtost dobro napovedujeta ali bo posameznik odrl blog ali ne.  


##### Zakaj osebnost: 

Osebnostne lastnosti kot so introvertiranost / ekstrovertiranost so **morda koristna informacija za prilagajanje**. 

Izraz koristna informacija pomeni, da sta optimalno prilagojeni storitvi za dva **uporabnika z raličnimima osebnostnima različni**. 


<br>
_Literatura: _ Vinciarelli, A., and Mohammadi, G. A survey of personality computing. IEEE Transactions on Affective Computing, 5 (3). pp. 273-291, 2014.

##  ■ Kaj je psihološki konstrukt - merilni instrument?


##### Def.: 

Construct, also called hypothetical construct or psychological construct, in psychology, a tool used to facilitate understanding of human behaviour. 

In a sense, a psychological construct is a label for a cluster or domain of covarying behaviours. 



##### Dimenzije:
Pojem, ki ga psihološki konstrukt opisuje, lahko zajema več (vsaj delno) neodvisnih značilnosti. Imenujemo jih dimenzije. Psihološki konstrukt ima dimenzije 
$$ d_1, \ldots, d_N, $$
ki imajo svojo interpretacijo. 

Npr. pri modelu osebnosti Velikih 5 je $N=5$ dimenzij te dimenzije so
$$ d_1=\mbox{Sprejemljivost},\quad\ldots\quad d_5=\mbox{Nevroticizem}. $$

Dimenzije konstrukta navadno niso zares neodvisne. 


##### Kdaj je uporaben:

Psihološki konstrukt na podlagi vprašalnika je uporaben predvsem za časovno nespreminjajoče se lastnosti uporabnika. Zato ga lahko izmerimo (določimo) le enkrat oz. zelo redko, enkratna nekaj let. 

<br><br>
**Pomembno:** sprememba domene, sprememba jezika vprašanj ipd pomeni, da je **intruments potrebno ponovno ovrednotiti in potrditi** psihometrične karakteristike.

##  ■ Model osebnosti Velikih 5 (angl. Big5)


Model osebsnoti Velikih 5 je **najbolj uporabljan model v interakciji človeka z napravami**. 

#### Značilnosti 
- _Sprejemljivost:_ ohranjanje pozitivnih odnosov z drugimi. Sposobnost razumevanja z drugimi osebami.
- _Ekstravertnost:_ Ekstravertne osebe se v veliki meri zanimajo za družabne dogodke, so zelo aktivne in energični
- _Odprtost:_ osredotoča se na posameznikovo kulturno sofisticiranost in posameznikovo odprtost za nove izkušnje.
- _Vestnost:_ odgovornost, naklonjenost pravilom, vztrajnost, pozornost…
- _Nevroticizem:_ visoko izražen nevroticizem pomeni manjšo sposobnost obvladovati čustva, več anksioznosti

```python
## Zajem osebnosti Velikih 5
import ipywidgets as widgets
import numpy as np


# Definicija vprašalnika
m = 50
tab_contents = ["Q" + str(i) for i in range(1,m+1)]

# Vprašanja
list_of_questions = ["Am the life of the party", "Feel little concern for others.", "Am always prepared.", "Get stressed out easily.",
             "Have a rich vocabulary.", "Don't talk a lot.", "Am interested in people.", "Leave my belongings around.",
             "Am relaxed most of the time.", "Have difficulty understanding abstract ideas.", "Feel comfortable around people.",
             "Insult people.", "Pay attention to details.", "Worry about things.", "Have a vivid imagination.",
             "Keep in the background.", "Sympathize with others' feelings.", "Make a mess of things.", "Seldom feel blue.",
             "Am not interested in abstract ideas.", "Start conversations.", "Am not interested in other people's problems.",
             "Get chores done right away.", "Am easily disturbed.", "Have excellent ideas.", "Have little to say.",
             "Have a soft heart.", "Often forget to put things back in their proper place.", "Get upset easily.",
             "Do not have a good imagination.", "Talk to a lot of different people at parties.",
             "Am not really interested in others.", "Like order.", "Change my mood a lot.", "Am quick to understand things.",
             "Don't like to draw attention to myself.", "Take time out for others.", "Shirk my duties.", "Have frequent mood swings.",
             "Use difficult words.", "Don't mind being the center of attention.", "Feel others' emotions.", "Follow a schedule.",
             "Get irritated easily.", "Spend time reflecting on things.", "Am quiet around strangers.", "Make people feel at ease.",
             "Am exacting in my work.", "Often feel blue.", "Am full of ideas."]

# Možni odgovori
options = ["Very Inaccurate", "Moderately Innacurate", "Neither Accurate nor Innacurate", "Moderately Accurate", "Very Accurate"]
list_of_options = [options] * m

 

# "radio" gumbi
children = [widgets.RadioButtons(options=opt) for opt in list_of_options]


# Implementacija vprašalnika - uporaba widgeta
# sestava: labela(vprašanje, widget (radio gumbi))
questions = [widgets.HBox([widgets.Label(name), child])
            for name, child in zip(list_of_questions, children)]


# Poimenuj vprašanja za lažjo orientacijo
tab = widgets.Tab()

tab.children = questions
for i in range(len(questions)):
    tab.set_title(i, tab_contents[i])
```

```python
## Izvedba merjenja
tab
```

```python
## Definicija modela osebnosti - določitev uteži za posamezne lastnosti

# Pretvorba opisnih odgovorov v številske
map_dict = {k:v for k, v in zip(options, [0, 1, 2, 3, 4])}
aQ = [map_dict[x.value] for x in children]


# Definicija uteži - to je modela
Sprejemljivost_d1 = (100 / 40) * (aQ[0] + 4 - aQ[5] + aQ[10] + 4 - aQ[15] + aQ[20] + 4 - aQ[25] + aQ[30] + 4 - aQ[35] + aQ[40] + 4 - aQ[45])
Ekstravertnost_d2 = (100 / 40) * (aQ[6] + 4 - aQ[1] + aQ[16] + 4 - aQ[11] + aQ[26] + 4 - aQ[21] + aQ[36] + 4 - aQ[31] + aQ[46] + 4 - aQ[41])
Odprtost_d3 = (100 / 40) * (aQ[2] + 4 - aQ[7] + aQ[12] + 4 - aQ[17] + aQ[22] + 4 - aQ[27] + aQ[32] + 4 - aQ[37] + aQ[42] + 4 - aQ[47])
Vestnost_d4 = (100 / 40) * (aQ[8] + 4 - aQ[3] + aQ[18] + 4 - aQ[13] + aQ[28] + 4 - aQ[23] + aQ[38] + 4 - aQ[33] + aQ[48] + 4 - aQ[43])
Nevroticizem_d5 = (100 / 40) * (aQ[4] + 4 - aQ[9] + aQ[14] + 4 - aQ[19] + aQ[24] + 4 - aQ[29] + aQ[34] + 4 - aQ[39] + aQ[44] + 4 - aQ[49])
```

```python
## Izpis
print ("Izmerjene dimenzije Velikih 5 (lastnosti) so:")
print ("Sprejemljivost:", Sprejemljivost_d1) 
print ("Ekstravertnost:", Ekstravertnost_d2)
print ("Odprtost:", Odprtost_d3)
print ("Vestnost:", Vestnost_d4)
print ("Nevroticizem:", Nevroticizem_d5)
```

## ■ Kaj je razpoloženje


##### Def.: "Razpoloženje je čutstveno stanje". V nasprotju s čustvi je manj specifično določeno in relacije do vzrokov, ki ga povzročajo, so manj znane. Navadno govorimo le o pozitivnem in o negativnem razpoloženju, ki ima manjšo ali večjo amplitudo. 

Razpoloženje navadno merimo le kot negtivno in pozitivno, kar pomeni, da za uporabno oceno **zadoščajo že precej revni vhodni podatki** kot npr. del teksta, ki ga je uporabnik napisal ali kratek govorni posnetek. 

Na drugi strani pa je za prilagajanje storitev **dovolj le ta podatek o predznaku** in je amplituda tega razpoloženja manj pomembna. 

##### Merjenje razpoloženja
Zaradi vloge razpoloženja navadno ugotavljamo (meirmo) le pozitivno oz. negativno razpoloženja. Tako je merjenje razpoloženja v veliki meri izpeljano iz merjenja 
- razpoloženje iz teksta
- razpoločenje iz socialnih omrežij
- afektivnega stanja 
- ...


Zato posebnih metod za merjene razpoloženja ne bomo obravnavali. 

<br>

######  Sentiment in razpoloženje:
Nekonsistentno izrazoslovje. Bolje definirano je razpoloženje (angl. mood). Sentiment rezerviramo za binarno opreacijo - odnos uporabnika do izbranih vsebin in/ali storitev. Več kasneje.

## ■ Kaj je afektivno in kaj čustveno stanje


##### Def.: Afektivno stanje je odziv organizma na vzbujanje.

Pri tem imamo v mislih nizkonivojski biološki odziv, na nivoju samih biokemičnih signalov v telesu. Ali je to ustrezen opis ali ne, se stroka **ne strinja**.


##### Def.: Čustveno stanje je zavedni del izkustva. Znanstvene definicije so zelo različne in tudi nasprotujoče si. 



V rabi so različne teorije čustvenih stanj - med drugim: 
- James-Lange: čustva so naša interpretacija psihološkega odziva na stimuli 
- Cannon: čustva so naš psihološki odziv na stimuli
- Schacter-Singer: čustva so rezultat ovrednotenja psihološkega odziva na stimuli, upoštevaje celotno situacijo


Čustva vključujejo tako kognitivne kot fizične odzive na stimuli.



##### Uporabna postavitev za prilagajanje uporabnikom
- merimo afektivno stanje: VAD prostor (podrobnosti kasneje)
- po potrebi (v postopku prilagajanja) to preslikamo v čustveno stanje. Procesno gledano ni potrebno, koristno je zaradi interpretacije

## ■ Modeli afektivnega stanja

Obstaja **več modelov afektivnega stanja**. Za nas je najpomembnejši model **VAD**.

#### Model afekta VAD (Valence, arousal, dominance):

To je dimenzionalni model, kjer dimenzije pomenijo
- _valenca_ : pozitivno ali negativno čustvo
- _vzburjenje_ : pomirjajoče ali razburljivo
- _dominantnost_ : dominantnost čustva
   
   

Za prostor VAD obstaja največ senzorske in programske opreme za določanje stajna v realnem času. Npr. Noldus Face Reader, ki ga predstavljamo kasneje.

## ■ Modeli čustvenih stanj

Tako kot pri afektivnih stanjih, tudi pri čustvenih stanjih obstaja več modelov. Zaradi zgodovinskega razvoja je teh modelov veliko in jih razumemo na podlagi razlogov, zaradi katerih so nastali.

Izbrani modeli čustvenih stanj so
- _Circumpleksni model_ : Preslikava iz VAD prostora (**najvažnejši za naš okvir**)
- _Vectorski model_ : 2D model smeri čustev
- _Positivna aktivacija – negativna aktivacija (PANA) model_ : negativen afekt in pozitiven afekt sta samostojna pojava
- _Plutchikov model_ : vključuje tako osnovna čustva kot dimenzijski model
- _PAD model_ : 3D model z osmi zadovoljstvo, vzburjenje in dominantnost 
- _Lövheimova kocka čustev_ : preslikava bio substanc dopamin, noradrenalin in serotonin v osem osnovnih čustev
- _Cowen & Keltner_ : 27 kategorij na podlagi analize odzivov posameznikov, ki so skupaj gledali 2185 video posnetkov

## ■ Računski modeli čustev

##### Def.: Izraz **računski model** izhaja iz psihologije in označuje mehanizme, ki vhodne dražljaje (dogodke, signale) prevede v čustven odziv. 

Računski modeli so opisi mehanizmov, ki se dogajajo v človeku na vseh nivojih, in rezultirajo v čustveni odziv.

##### Izraz računski model uporabljamo tudi za modeliranje socialnih signalov kot so oklevanje, nezaupanje ipd.

Računski modeli se delijo na
- ocenjevalne
- dimenzijske 
- anatomske
- racionalne

Razlikujejo se glede **osnovnih teoretičnih privzetkov**. 


Za potrebe prilagajanja uporabnikom sta najvažnejši skupini računskih modelov **ocenjevalni** in **dimenzijski**. Razlog je v tem, da sta - čeprav daleč - vendarle najbližja realni implmenetaciji v okviru prilagajanja storitev.


<br>

_Literatura_ : S. C. Marsella, J. Gratch: EMA: A process model of appraisal dynamics, _Cognitive Systems Research_ , 2009.

## ■ Oris implementacije modela afektivnega in čistvenega stanja (1)

##### Osnovni razmislek:
- ker gre za časovno hitro spreminjajoč pojav, ga lahko ocenjujemo (modeliramo) iz senzorjev, ki imajo dovolj veliko vzorčno frekvenco. 
- v poštev pridejo vizualni signali in psihofiziološki senzorji

##### Kodomena modeliranja - kako opišemo stanje
- _afektivna stanja_ : lahko jih opisujemo na 
    - intervalnem nivoju (zvezno): npr. $V, A, D\in [0, 1]$
    - razredih velikosti: {"nizko", "visoko"} ali {"nizko", "srenjde", "visoko"}
    
- _čustvena stanja_ : lahko jih opisujemo na
    - originalnih opisih, odvisno od modela, npr cirkumpleksni model
    - združenih opisih - razredih: sorodna čustven stanja združimo v nov razred



Glede na zgornje ločimo modele:
- zvezni nivo: regresijski modeli
- diskretni razredi: klasifikacijske metode strojenga učenja

##  ■ Uvod


##### Def.: Sentiment je naravnanost uporabnika glede na entitete, ki so posamezniki, organizacije, produkti, dogodki, tematike, vsebine ter njegova naravnanost do njihovih lastnosti. Torej je sentiment uporabnika vezan na določeno entiteto.

Uporabnik ima lahko **različne sentimente do različnih entitete** ob **istem času**. Take lastnosti npr. čustveno stanje nima - to je v vsakem trenutku določeno eno samo in je torej stanje uporabnika in ni nujno vezano na eno samo entiteto ali sploh katero entiteto. 


Ta naravnanost je povezana z afektivnimi in čustvenimi stanji in razpoloženjem, z mnenjem itd. 


Sentiment navadno merimo le v predznaku, torej **pozitiven** in **negativen**.

##  ■ Analiza sentimenta iz uporabniško generirane vsebine

##### Def.: Uporabniško generirana vsebina je vsebina, ki jo generira sam uporabnik med uporabo storitve - med komunikacijo, med komentiranjem vsebine, med načinom branja vsebine itd 

Tipični primeri uporabniško generiranih vsebin so
- objave na Facebook
- twitti 
- kratka sporočila (SMS)
- govor (npr. Skype)

## ■ Model kognitivnega procesiranja uporabnika (2)

Že iz opisa posameznih pojmov (spodaj) je jasno, da je pri uporabniku prilagojeni komunikaciji kognicija pomembna. Brez kognitivnih virov uporabnik ne more konzumirati vsebine niti upravljati s storitvijo, ki jo ta prinaša. 

#### Deljena pozornost

Uporabniku omogoča, da istočasno opravljao več nalog. Seveda za ceno manjše uspešnosti posamezne naloge. Stroka se še vedno ne strinja o veljavnem modelu deljene pozornosti. Pri komunikaciji s pametnimi sistemi gre za deljeno pozornost predvsem med upravljanjem sistema (nezaželjen del) in samo storitvijo (konzumiranje MM vsebine; zaželjen del).


#### Selektivna pozornost

Selektivna pozornost uporabniku omogoča, da se osredotoči na tisto dejavnost, ki jo je sam izbral. Gre za sposobnost, da kljub motečim dejavnikom uporabnik s pozornostjo uporabnik vztraja na zastavljenem cilju (izbranem cilju ali najvažnejšem cilju v danem trenutku).


#### Delovni spomin

Delovni spomin uporabniku omogoča, da informacije preko senzorike začasno shrani z namenom organizacije v strukturo in nato sočasno uporabo pri kognitivnih procesih. Ker sta glavna vira informacij akustične informacije in vidno-prostorske informacije, navadn tudi delovni spomin delimo na ta dva dela. Razvoj novih interakcijskih možnosti pa zazteva (bo zahteval) prilagoditev teh modelov. 

V okviru delovnega spomina so posamezni podprocesi sorazmerno dobro znani. Glavna vidika sta
* kapaciteta: $7 \pm 2$ enoti. Velike razlike med posamezniki, zadeva je merljiva; 
* čas hrambe: $10 - 15$ sekund, lahko do ene minute. Razlike med posamezniki in razlike med razmerami (prisotnost motenj, utrujenost, ...)
