## ■ Osnove Booleanovega modela iskanja

#### Osnovne značilnosti 
- vsebine so opisane s **prisotnostjo atomov** kot so besede, torej kot $0$ (ni) in $1$ (je). To je **binarna predstavitev**
- iskanje poteka na logičnem nivoju, kjer so vhodne spremenljivke prisotnosti atomov
- Vsak logični operator se da predstaviti z Booleanovimi operatorji AND, OR, NOT
- Sistem vrne vse dokumente, ki ustrezajo logičnemu opisi. Dokumenti **niso rangirani** po ustreznosti

## ■ Binarna predstavitve vsebin

#### Def.: Binarna predstavitev vsebin je tabela prisotnosti izbranih elementov vsebin, tipično besed v tekstovnem elementu. Prosotnost je označena z $1$, odsotnost pa z $0$.

##### Dokument 1
Dallas je v obračunu z Utahom v drugem polčasu prikazal ples mačke z mišjo. Jazz je v drugih 24 minutah zbral pičlih 22 točk in doživel najhujši poraz po marcu 1979, ko jih je še kot New Orleans za 56 točk premagal Milwaukee (102:158). Luka Dončić je v statistiko vpisal 13 točk, šest skokov ter po dve podaji in ukradeni žogi v 25 minutah igre, s čimer je bil tretji strelec moštva. 
Celotno zadnjo četrtino je Dončić presedel, po tekmi je opravil preventivno slikanje rame, saj je med tekmo zadel steber koša po trku z Jaejem Crowderjem, a poškodbe ni dobil. Najbolj razpoložen v domači vrsti je bil Harrison Barnes z 19 točkami, s 14 točkami mu je sledil JJ Barea, vsaj 4 točke pa so prispevali prav vsi izmed 13 igralcev Mavericksov.

##### Dokument 2
Luka Dončić je vknjižil 11 točk v prvi gostujoči zmagi Dallasa v sezoni. Mavsi so s 103:98 slavili v Chicagu. ## ■ Poizvedba pri Booleanovi predstavtvi

#### Logični operatorji

To so **AND, OR, NOT**

Velja: **vsak logični izraz** lahko izrazimo s temi tremi izrazi


<br>

#### Poizvedba 

Poizvedba je logični izraz s ključnimi besedami

Primer: 

<center>
*(Luka **OR** Dončić) **AND** košarka **AND** (**NOT** Dragić)*  
</center>

<br>

#### Operatorji bližine

Logičnio operatorji so grobi v smislu natančnejšega presejanja rezultatov, kar vodi v ali preveliko število rezultatov ali v prazen rezultat. 

Dodatne specifikacije podamo z **operatorji bližine**
- "NEAR n" dovoljuje največ n-1 vmesnih besed med dvema terminoma
- "WITH" zahteva, da sta termina zaporedna (v pravem zaporedju)
- ...

Uporaba teh operatorjev zahteva natančnejše predstavitve dokumentov, npr. poleg prisotnosti še pozicije besed. 

Primer: 

<center>  
(Luka **OR** Dončić) **(NEAR 6)** košarka **AND** (**NOT** Dragić)
</center>

## ■ Uteževanje elementov (besed)

Uteževanje besed po pomenu je **ključni mehanizem za regulacijo iskalnih reultatov**. Gre za dodeljevanje uteži besedam za **skupino vsebin**, ne za posamezno vsebino. 

Želene lastnosti:
- Beseda, ki se v dokumentu pogosto pojavi, naj ima veliko utež;
- Besede, ki pa se pojavljajo v veliko dokumentih, naj imajo nizko vrednost uteži;

#### TF.IDF uteževanje

Za element (besedo) $e_i$ v vsebini $h_j$ določimo utež
$$ w_{ij} = tf_{ij} \log\frac{N}{n_i},  $$
kjer je $w_{ij}$ utež elementa $e_i$ v vsebini $h_j$, $tf_{ij}$ število pojavitev elementa $e_i$ v vsebini $h_j$, $N$ število vseh dokumentov in $n_i$ število dokumentov, ki vsebujejo element $e_i$.

## ■ Algoritem PageRank

#### Cilj algoritma: rangiranje strani na podlagi numeričnih vrednosti

Prikažemo poenostavljen algoritem, ki je osnova vseh nadaljnjih razvojev. 

##### Osnovni pojmi: 
- seznam s ppovezavami povezanih strani: $A_1, A_2, \ldots, A_n$
- graf povezav med stranmi: $G = (V,E)$, $V(G) = \{A_i: i=1,\ldots, n\}$, $E(G)=\{(i,j): \mbox{obstaja povezava iz $A_i$ v $A_j$}\}$
- verjetnost, da bo naključno klikanje pripelalo do strani $Pr[A]$: $Pr[A]$
- osnonvi razmislek (brez dušenja)
$$ Pr[A_i] = \sum_{(ij)\in E(G)} \frac{Pr[A_j]}{L(A_i, A_j)} $$
- dušenje (angl. dumping): $d=Pr[\mbox{naključno klikanje se nadaljuje}]$. Rešuje problem napačne predpostavke, da se tako klikanje zares konča. 
- dopolnjena formula za verjetnost 
$$ Pr[A_i] = \frac{1-d}{n} + d \sum_{(ij)\in E(G)} \frac{Pr[A_j]}{L(A_i, A_j)} $$


##### Algoritem

Poenostavlje algoritem: ignorira povezave znotraj dokumenta in vzporedne povezave šteje kot eno samo.

Zakaj iteracija: ker so formule iterativne.

Koraki
1. $Pr[A_i] = \frac{1}{n}$
2. While ($\Delta (Pr[A_i]) < \varepsilon$):
3. $\quad$ for ($i\in V(G)$): 
4. $\quad\quad$ $Pr[A_i] = \frac{1-d}{n} + d \sum_{(ij)\in E(G)} \frac{Pr[A_j]}{L(A_i, A_j)}$
5. EndWhile

Tipično število iteracij: $50$

Obstaja matrična rešitev. 


<p style="margin-bottom:2cm;"></p>
<div style="width:100%;text-align:right;font-weight:bold;font-size:1.2em;"> 122 </div>
<img src="https://raw.githubusercontent.com/andrejkk/ORvTK_SlidesImgs/master/footer_full.jpg">

```python
# Primer PageRank algoritma v Pythonu
# Obravnava preprost primer štirih dokumentov


import networkx as nx
import matplotlib.pyplot as plt
import numpy as np


page_lst = ["A1", "A2", "A3", "A4"]
pageG = nx.DiGraph()
pageG.add_nodes_from(page_lst)
nx.draw(pageG, with_labels=True)
plt.show()
```

```python
# Add links - uniform weights 

pageG.add_weighted_edges_from([("A1", "A2", 1), ("A2", "A3", 2),("A3","A4",1),("A4","A1",1),("A2", "A4", 1)]) 
nx.draw(pageG,with_labels=True)
plt.show()
```

```python
# Set initial ranking values
pr=nx.pagerank(pageG, alpha=0.9)
pr
```

```python
rank_vector=np.array([[*pr.values()]])
best_node=np.argmax(rank_vector)
print("The most popular website is {}".format(page_lst[best_node]))
```

## ■ Nekatere opombe k pogovornim sistemom

#### Algoritem "deep reinforcement learnig"

Zaradi kompleksnosti jezika s uveljavlja učenje, ki maksimizira končni učinek - deep reinforced learnig

**Predstavitev problema**
- Množica stanj okolja in stanj agenta S;
- Množica akcije agenta A
- Verjetnost prehoda iz stanja $s$ v stanje $s'$ ob izvedbi akcije $a\in A$ je
$$ P_a(s, s') = Pr(s_{t+1}=s'|_{s_t=s, a_t=s}) $$
- Nagrada (takojšna) za prehod med stanjema $s$ in $s'$ ob izvedbi akcije $a\in A$
$$ R_a(s,s') $$
- pravila, ki opisujejo informacije o okolju

#### Kontekstna občutljivost govora

Napaka pri določanju konsteksta je navadno usodna zapotek pogovora. Pomembni vidiki: 
- ljudje so zelo učinkoviti pri kontekstualizaciji in hkratnem razumevanju jezika
- izgovorjava je tudi odvisna od konteksta
- pri generiranju naravnega jezika potrebujemo kontekst dovogovora za
  - izboru relevantnih vidikov dvogovora, ki jih vključimo
  - opozarjanje sogovoronika na morebitno napačno razumevanje samge konteksta 



#### Dobre lastnosti

- potencial za rešitev minimalne kraje pozornosti
- rešitev naravne komunikacije
- rešitev komunikacije / interakcije za posebne razmere

- napake posameznih modulov se usodno širijo do rezultata
- zelo zahtevna kontekstualizacija
- če je prva izkušnja uporabnika slaba, se v velikem delžu ne vrne več
