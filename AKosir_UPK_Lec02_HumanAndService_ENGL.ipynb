##  ■ User, information and cognitive processes


The user while using the device ** communicates **, e.g.When he reads a book and inquires for the books mentioned in the book.In the meantime, ** is processing information ** that runs in both directions.

#### The components of the processing of information are

1. Feeling: sensing stimuli from the environment.

2. Perception: Interpretation of sensual information.

3. Attention: Focusing on certain stimuli.

4. Working memory: Short -term storage and information management.

5. Long -term memory: storing information for a longer period of time.

6. Coding Information: Information conversion to memory.

7. Call: Access to stored information.

8. Cognitive load: Effect: A mental effort used in the processing of information.

9. Decision-making: making informed choices.

10. Problem solving: problem solving: Finding solutions to solve problems.

11. Metacognition: Recognition: Awareness and regulation of one's thinking.



#### Some aspects

1. Information in both directions
- Visual, sound, haptic, movement, ...


2. Information stored in memory
- Sensory memory, short -term memory, long -term memory

3. Processing and using information
- concluding, solving problems, skill, mistakes   
- Response and reaction time <br>

**Response time = reaction time + movement time to carry out action**

- Humans can do about 120 bit/sec of attentive information processing. Human speech is approx. 60 bit/sec.

## ■ What is mood


##### Def.: "Mood is a feeling state". In contrast to emotions, it is less specifically defined and the relationships to the causes that give rise to it are less well known. We usually speak only of positive and negative moods, which have a smaller or larger amplitude. 

Mood is usually measured only as negative and positive, which means that even quite poor input data** such as a piece of text written by the user or a short speech recording is sufficient for a useful assessment. 

On the other hand, for customisation services **only this information about the mood** is sufficient and the amplitude of this mood is less important. 

##### Mood measurement
Due to the role of mood, we usually only identify (measure) positive or negative mood. Thus, the measurement of mood is largely derived from the measurement of

## ■ What is an affective state and what is an emotional state


##### Def.: An affective state is an organism's response to arousal.

Here we mean a low-level biological response, at the level of the body's own biochemical signals. Whether or not this is an adequate description, the profession **disagrees**.


##### Def.: An emotional state is a conscious part of experience. Scientific definitions vary widely and are also contradictory. 



Different theories of emotional states are in use - including: 
- James-Lange: emotions are our interpretation of our psychological response to stimuli. 
- Cannon: emotions are our psychological response to stimuli
- Schacter-Singer: emotions are the result of evaluating our psychological response to a stimulus, taking into account the whole situation


Emotions include both cognitive and physical responses to stimuli.



##### A useful layout for personalisation
- Measures affective state: VAD space (details later)
- if necessary (during the adaptation process), map this to the affective state. Procedurally not necessary, it is useful for the sake of interpretation

## ■ Models of Emotional States

As with affective states, there are several models of emotional states. Due to historical development, these models are numerous and we understand them based on the reasons for which they arose.

Selected models of emotional states are
- _Circumplex model_ : Mapping from VAD space (**most important for our framework**)
- _Vector model_ : 2D model of emotion direction
- _Positive activation – negative activation (PANA) model_ : negative affect and positive affect are independent phenomena
- _Plutchik model_ : includes both basic emotions and a dimensional model
- _PAD model_ : 3D model with eights pleasure, arousal and dominance
- _Lövheim's cube of emotions_ : mapping of biosubstances dopamine, noradrenaline and serotonin into eight basic emotions
- _Cowen & Keltner_ : 27 categories based on analysis of responses of individuals who watched a total of 2185 video clips

<br>

## ■ Appraisal theory of emotions

This theory claims that emotions **are not a direct biological response to stimuli**, but the result of **higher-level evaluation of these stimuli**.

It has several important implications for the field of adaptation:
- the role of emotions in communication
- the role of emotions in assessing the user's state

The assumption of an evaluative theory of emotions (higher cognitive abilities also participate in emotional response) for adaptation means
- more complex solutions: the current state of development does not yet allow them
- improved possibilities for obtaining data about the user

Several phases of assessment:
1. Primary assessment: assessment of the effect of the event - beneficial, harmful, dangerous, ...
2. Secondary assessment: assessment of circumstances, consequences and possibilities for response
3. Generation of emotions: generation of an emotional state
4. Re-cognitive assessment

<br>

## ■ What is sentiment?

##### Def.: Sentiment is the user's attitude towards entities, which are individuals, organizations, products, events, topics, content and his attitude towards their properties. Therefore, the user's sentiment is tied to a specific entity.

Such an entity is also multimedia content, its author, the service it provides (e.g. Facebook), etc.

A user can have **different sentiments towards different entities** at **the same time**. For example, an emotional state does not have such properties - it is determined at any given moment by a single one and is therefore the user's state and is not necessarily tied to a single entity or any entity at all.

This attitude is related to affective and emotional states and mood, to opinion, etc.

Sentiment is usually measured only in sign, i.e. **positive** and **negative**.

Sentiment analysis usually refers to **_document_**: text, speech, sequence of tweets, etc.

## ■ Sentiment Analysis Approaches

Sentiment analysis includes two main approaches:
- **_Lexicon-based_**: based on pre-labeled pieces of sentiment data. Typically, some words in the text are labeled with negative sentiment, some words with positive sentiment, and the rest (most) are without sentiment.
- **_Machine learning-based_**: supervised and unsupervised machine learning, typically based on user feedback (questionnaires) or indirect sentiment cues (e.g., complaint text has negative sentiment).

## ■ User Cognitive Processing Model (2)

It is clear from the description of individual concepts (below) that cognition is important in user-adapted communication. Without cognitive resources, the user cannot consume content or manage the service it provides.

#### Divided attention

It allows the user to perform multiple tasks simultaneously. Of course, at the cost of lower performance of individual tasks. The profession still does not agree on the valid model of divided attention. In communication with smart systems, attention is divided primarily between system management (the unwanted part) and the service itself (consuming MM content; the desired part).

#### Selective attention

Selective attention allows the user to focus on the activity that he or she has chosen. It is the ability to persist with attention on a set goal (the selected goal or the most important goal at a given moment) despite distracting factors.

#### Working memory

Working memory allows the user to temporarily store information via the senses for the purpose of organizing it into a structure and then simultaneously using it in cognitive processes. Since the main sources of information are acoustic information and visual-spatial information, working memory is usually divided into these two parts. The development of new interaction options requires (will require) the adaptation of these models.

Within the framework of working memory, individual subprocesses are relatively well known. The main aspects are
* capacity: $7 \pm 2$ units. Large differences between individuals, the matter is measurable;
* retention time: $10 - 15$ seconds, possibly up to one minute. Differences between individuals and differences between conditions (presence of distractions, fatigue, ...)

## ■ Multiple Source Theory and Switching

###### Def.: Multiple source theory assumes that tasks that use different resources interfere less with tasks that use different resources than tasks that use the same resources.

#### Which resources do we share?

##### Audio (language) and video (visual):

Basically, the resources for audio and visual information are largely separate. Thus, language-related tasks (reading, listening, ...) compete for audio resources, while tasks related to visual perception compete for visual resources.

So, we can perform one auditory task and one visual task in parallel.

#### Upper limit on the amount of information

There is also an upper limit on the amount of information that can be processed, which is estimated at $120$ bits / second. For orientation, human speech contains approximately $60$ bits per second.
