### Cilj poglavja in učni izidi

#### Cilj:
Cilj poglavja Mehanizmi prilagajanja, kontekst in prilagojene storitve je spoznavanje osnovnih pojmov samega prilagajanja storitev. Spoznali jih bomo skupaj s pomembnimi primeri prilagojenih storitev.  
- spoznati pristope k priporočanju vsebin in storitev
- razumeti delovanje in obnašanje osnovnih algoritmov


#### Učni izzidi
Razumevanje in aktivna uporaba pojmov prilagajanja, razdalje med uporabniki in vsebinami, rangiranja, ter posameznih prilagojenih storitev.
