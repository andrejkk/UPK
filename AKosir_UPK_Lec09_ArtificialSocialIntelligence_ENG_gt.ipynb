## ■ Why social signals and usability

##### Why social signals

Social signals are additional information for the **listener**, so that in addition to the content of the message itself (speech, image, ..., i.e. multimedia) he also understands necessary **additional information** such as **the purpose of the message**, **the situation (context)**, possible **consequences**, **the expectations of the communicator**, etc.

Social signals are additional information for the **speaker**, so that in addition to the content of the message he communicates, he receives information from the listener (previous paragraph), which is essential for **continuing communication**.

##### Usability

Social signals are useful for
- **real-time communication**: significant enrichment of content
- **adaptation to the user**: understanding the user's state in real time

## ■ Expression of emotions

#### Basic insights

1. A machine or service can express emotions **with similar channels as a human:** graphics similar to a human face, voice color
2. A machine or service can also express emotions **with completely new channels** such as the color of light, etc.

#### Emotional bandwidth

Property of a communication channel regarding the ability to express emotions:
- higher bandwidth: more emotions
- lower bandwidth: less emotions

Emotional bandwidths:
- text: small
- voice: larger
- visual information: largest

Approaches with **virtual reality** can have **higher bandwidth** than human-to-human communication.

## ■ Step 5: Determining and verifying the true values ​​(3)

#### Verifying the true value

##### Why?
If the **ground truth** is derived from annotation, a **necessary condition** for validity is a sufficiently high **rater agreement**.

If the rater agreement is not high enough, the obtained true values ​​are **not valid**.

##### Consistency Check Measures
$\def\o#1{{\overline{#1}}}$

Consistency check measures are
1. Cohen's and Fleiss's Kappa <br>
- removes random matches
- minimum values ​​are known


|$\kappa$| Consistency |
|---|---|
| $ < 0.01$ | very poor  |
| $0.01 – 0.20$ | poor  |
| $0.21 – 0.40$ | medium  |
| $0.41 – 0.60$ | good  |
| $0.61 – 0.80$ | very good  |
| $0.81 – 1.00$ | excellent  |




2. Correlation coefficients <br>
- Depending on the types of variables being evaluated, we use Pearson's $r$, Kendall's $\tau$ or Spearman's $\rho$ correlation coefficient.
- We calculate pairwise correlations between evaluators, which must be **high enough**.

3. Intra-class correlation coefficient
- the most favorable for our needs
