## ■ What is experimental design

The elements of the experimental design are as follows:
##### 1. Response (output, dependent) variable $c$:
Criteria function, performance indicator, etc., which gives the main result. In our case, the MM exposure.

##### 2. Controlled factor (s) $F_e$:
Independent variables that we ** control ** in the experiment. In our case, this is a communication device, so it has a controlled value factor (levels) {PC, Alexa Echo}.

##### 3. Nuisance factors:
These are adverse effects on the experimental result that are unavoidable. We want to neutralize them in several ways;

##### 4. Input noise $ \ varepsilon $:
This is the noise with which we present the unknown disturbing factors and the real noise resulting from the randomness of human behavior, the noise of the sensors, etc.

- before the experiment, we make a **statistical design** for data analysis, which will answer the questions

## ■ Experimental in nonexperimental design

#### Experimental design

This is an experiment where we **compare** two or more variants
- user groups
- system variant
- situations
- ... <br>
with each other. To do this, we need to control at least one of the **factors**, that is, to control its value (level).

It is important to consider other ** factors ** (influences on the result) appropriately.


#### Nonexperimental design

This is an experiment in which we observe the results **without factor control** and do not interfere in the course of the factor control. We call it **observational study**.

## ■ Design ANOVA  

$\def\ovr#1{{\overline{#1}}}$


#### Controlled and nuisance factors

Types:
- fixed effect: the experimental factor is controlled
- random effect: the value of the experimental factor is random
- mixed effects: we have factors with fixed and random effects


Stock value:
- discrete values
- two or more values

#### Conditions

The variables are distributed normally.



#### Zero hypothesis and test

We cover only the option **one factor** with **fixed effect**.

Zero hypothesis:
$$ H_{0}=[\ovr{y}_{G_1}=\cdots=\ovr{y}_{G_I}], $$

Test: F-test.


$\def\o#1{{\overline{#1}}}$
$\def\SS{{\mbox{SS}}}$
$\def\Var{{\mbox{Var}}}$

###### Notations
- $n$ is the number of experiments performed at each value of the controlled factor $F_e$.
- $a$ is the number of values of the controlled factor $F_e \in \{1, 2, \ldots, a \}$
- $y_{ij}$ is the $j$ th result of the experiment, $i=1, \ldots, n$, for the value of the factor $F_e=i$
- $y_{i.}=\sum_{j=1}^n y_{ij}$ sums per experiment
- $\o{y}_{i.}=y_{i.} / n$ average per experiment
- $y_{. j}=\sum_{i=1}^n y_{ij}$ sums by factor values
- $\o{y}_{. j}=y_{. j} / n$ average by factor values
- $y_{..}=\sum_{i, j} y_{ij}$ total
- $\o{y}_{..}=y_{..} / n$ total average

##### Sum of squares
- total sum of squares
$$\SS_{tot}=\sum_{i=1}^a \sum_{j=1}^n (y_{ij} - \o{y_{..}}) ^ 2$$
- the sum of the squares of the factor
$$\SS_{trt}=n \sum_{i=1}^a (\o{y}_{i.} - \o{y_{..}}) ^ 2$$
- Then the total sum of squares can be divided by the sum of the factor and the error $\SS_{err}$, i.e.
$$\SS_{tot}=\SS_{trt} + \SS_{err}$$


##### Statistics F

Statistics
$$F_0=\frac{SS_{trt} / (a-1)}{SS_{err} / (n (a-1))}$$
is distributed after the distribution of $F (a-1, n (a-1))$, from which we calculate the $p$ value.

#### Variants
 - one way: the design has one controlled factor
 - two way: the design has two controlled factors

<div style="width:100%;text-align:right;font-weight:bold;font-size:1.2em;"> 17 </div>

```python
# Define functions
def get_ANOVA_SS(data_df):
    a,n = data_df.shape
    N = a*n
    
    y_pp = 1.0*data_df.sum().sum()
    y_ip = 1.0*data_df.sum(axis=1)
    
    # SS total
    SS_T = (data_df**2).sum().sum() - (y_pp**2)/N

    # SS treatment
    SS_trt = (1.0/n)*(y_ip**2).sum() - (y_pp**2)/N
    
    # SS error
    SS_err = SS_T - SS_trt
   
    # Report
    return SS_trt, SS_err, SS_T
```

```python
## Title: ANOVA design
# Goal: undesrtand the implementaiton of the ANOVA experimental design 
# Hint: 

import numpy as np
import pandas as pd 
from scipy.stats import f

# Load data 
post_fn = 'https://raw.githubusercontent.com/andrejkk/UPK_DataImgs/master/ANOVAtestData31.csv'
data_df = pd.read_csv(post_fn, sep=';', encoding='utf8')


# Summs of squares
SS_trt, SS_err, SS_T = get_ANOVA_SS(data_df)

print (SS_trt, SS_err, SS_T)

# F-stat
a,n = data_df.shape
N = a*n
F_0 = (SS_trt/(a-1)) / (SS_err/(N-a))

# P-value
pVal = 1-f.cdf(F_0, a-1, N-a)

print ('p value = ', pVal)
```

```python
## Get scores - standard formula
import numpy as np

# Get factor analysis scores 
def get_FA_scores(X, La):
    
    C = np.cov(X.T)
    invC = np.linalg.pinv(C)
    
    F = X @ invC @ La
    
    return F
    
    
```

## ■ Zanesljivost

#### Def.: Zanesljivost se nanaša na konsistentnost mere, to je ali bi isti pojem/pojav ob večkratnem merjenju izmerila dovolj podobno.


#### Vrste zanesljivosti
1. Zanesljivost v času - zanesljivost test-retest: ali večkratno testiranje da podobne rezultate (*ni uporabno za primer dvogovora z Alexo*)
2. Notranja konsistentost: ali so odgoovri med vprašanji dovolj povezani (*bistveno za primer dvogovora z Alexo* - glejte spodaj)
3. Konsistentnost med raziskavami (raziskovalci) (*ni uporabno za primer dvogovora z Alexo*)



#### Notranja konsistentost - metoda paralelnih oblik (angl. split half)

Odgovori na vprašanje **Ali so odgovori/podatki stabilni?**

##### Potek določanja paralelnih oblik
$\def\s{{\sigma}}$
Pri tej metodi za dane podatke:
1. Testne osebe naključno razdelimo na dve polovici
2. Izračunamo korelacije med njimi $r_{12}$
3. Po Spearman-Brownovi formuli izračunamo koeficient zanesljivosti
   - za privzete enake standardne odklone
  $$ r_{tt} = \frac{2 r_{12}}{1 + r_{12}} $$
   - za različne standardne odklone
  $$ r_{tt} = \frac{4 \s_1 \s_2 r_{12}}{\s_1^2 + \s_2^2 + 2 \s_1 \s_2 r_{12}}, $$
  kjer sta $\s_1$ in $\s_2$ standardna odklona posameznih podatkov. 



#### Notranja konsistentost - Cronbach alfa

Odgovori na vprašanje **Ali vsa vprašanja/podatki skupaj merijo koherenten pojem?**

##### Potek določanja Cronbach alfa
Rezultate (odgovore) na $K$ vprašanj $X_i$ seštejemo v 
$$ Y = X_1 + \cdots + X_K, $$
izračunamo standardne odklone posameznih rezultatov $\s_{X_i}$ in vsote $Y_{Y}$, ter izračunamo
$$ \alpha = \frac{K-1}{K} \left(1-\frac{\sum _{i=1}^K\s_{X_i}^2}{\s_Y^2}\right) $$

```python
# Define functions
def get_ANOVA_SS(data_df):
    a,n = data_df.shape
    N = a*n
    
    y_pp = 1.0*data_df.sum().sum()
    y_ip = 1.0*data_df.sum(axis=1)
    
    # SS total
    SS_T = (data_df**2).sum().sum() - (y_pp**2)/N

    # SS treatment
    SS_trt = (1.0/n)*(y_ip**2).sum() - (y_pp**2)/N
    
    # SS error
    SS_err = SS_T - SS_trt
   
    # Report
    return SS_trt, SS_err, SS_T
```

## ■ ANOVA

$\def\o#1{{\overline{#1}}}$
$\def\SS{{\mbox{SS}}}$
$\def\Var{{\mbox{Var}}}$
#### Def.: ANOVA primerja pričakovane vrednosti med dvema (to je t-test) ali več podgrupama in testira ničelno hipotezo, da so pričakovane vrednosti vseh grup enake. 

Ime **analiza avriance** (angl ANalysis Of VAriance, ANOVA) je nekoliko zavajajoče in ničelna hipoteza se ne tiče varianc, ampak pričakovanih vrednosti experimentalh grup. 

#### Vsota kvadratov (angl. sums of squares, SS)

Jedro razmisleka statističnih experimentalh načtov je zapis variance rezultata $Y$ z vsotami kvadratov **na dva načina**, kjer drugi način vključuje odklone od pričakovanih vrednosti po experimentalh grupah.   

1. Variance rezultata
$$ \Var_y = \frac{1}{n-1} \sum_{i=1}^n (y_i - \o{y})^2 = \frac{1}{n-1} \SS_{tot},  $$
kjer je $n$ število vseh meritev (rezultatov), $\Var_y$ celotna varianca rezultata, $\SS_{tot}$ pa celotna vsota kvadratov.


2. Vzemimo, da imamo eksperimentalne grupe $G_1, \ldots, G_a$, v katerih so pričakovane vrednosti $E(y)|_{G_j} = \o{y}_{G_j}$, $j=1,\ldots,a$. Vsoto kvadratov se da preurediti tako, da ločemo zapišemo vsoto kovadratov celote, znotraj skupin in napako
$$ \SS_{tot} = \SS_{treat} + \SS_{err}, $$
kjer je $\SS_{tot}$ pa celotna vsota kvadratov, $\SS_{treat}$ vsota kvadratov faktorja (varianca med skupinami) in $\SS_{err}$ vsota kvadratov napak (varianca znotraj skupin). 


#### Ničelna hipoteza $H_0$ in F-porazdelitev

Izkaže se, da če velja ničelna hipoteza
$$ H_{0} = [\o{y}_{G_1} = \cdots = \o{y}_{G_a}], $$
potem je statistika
$$ \frac{\SS_{treat}/(I-1)}{\SS_{err}/(n-a)} $$
porazdeljena po $F(I-1, n-a)$ porazdelitvi s prostostnimi stopnjami $a-1$ in $n-a$. 

To je osnova za izračun $p$ vrednosti.
