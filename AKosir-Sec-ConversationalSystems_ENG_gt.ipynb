## ■ Why conversationsl systems?


#### Reasons to use

1. Reducing theft of attention
2. We bring the system to a closer to natural communication
3. Special conditions when a user cannot use a keyboard and mice (ride, measure, ...)
4. Getting a user's data in real -time (eg analysis of speech for emotional state)


#### When to use it

1. When hands-free interaction is essential:
- an interface in a vehicle
- Support systems (assistants) with instructions
- management and control interface
- Learning foreign languages
- Learning in interactive environments


2. When the speech is lighter of typing
- Voice search interface
- Virtual assistant (Siri, Google Assistant)


3. Human agents
- Call Centers
- Voice Interface to help users (eg tax vision)

## ■ What is the recognition of speech

# Description: recognition of the audio signal of speech into a list of words

This is not yet an understanding of content or the importance of speech. It's just a conversion **speech** into **text**.

**Input:** Speech signal

**Result:** List of words


#### History
- 1952: Bell Labs, Audry system, recognizing numbers (first system)
- 1970: Carnige Mellon, Happy System, recognizing 1011 words
- 1980: Introduction of hidden Markov models

## ■ approaches to the Natural speech synthesis

### Two phases of synthesis
1. Preprocessing
- the division of the text into sentences
- Labeling parts of the sentences
- deciding the ambiguity of words
- the breakdown of words
- Determination of rhythm and intonation of speech



2. Synthesizing - see the following point
